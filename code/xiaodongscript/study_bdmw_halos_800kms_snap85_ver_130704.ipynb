## Sec 1. Useful Definitions & Basic Information of MDR1 

```python
import matplotlib.pyplot as plt
import numpy as np
#import os
import mpl_toolkits.mplot3d.axes3d as Axes3D
%pylab inline
import random
import scipy.integrate
```

```python
# Useful Definiations

const_c = 299792.458 #unit: km/s

# get theta, phi according to x,y. 2D
def getangle(x, y):
    theta = np.arcsin(y/np.sqrt(x**2+y**2))
    if x<0:
        theta = pi-theta
    if x>0 and y<0:
        theta = 2*pi+theta
    return theta

# get r, theta, phi given x,y,z. 3D
def getSC(x, y, z):
    r=np.sqrt(x**2+y**2+z**2)
    theta=getangle(z,np.sqrt(x**2+y**2))
    phi=getangle(x,y)
    return r, theta, phi

# Find Root: f(x) == f0
def FindRoot( f, f0, xl, xr, tol = 1e-6, maxstep=100 ):
    fl = f(xl); fr = f(xr);
    if abs(fl-f0) < tol:
        return xl, fl-f0, 0, 'S'
    if abs(fr-f0) < tol:
        return xr, fr-f0, 0, 'S'
    if fl > f0 and fr > f0:
        print "Error! f(xl)/f(xr) = ", fl, fr, "both larger than ", f0, "!"
        return 0, 0, 0, 'E' # E means error
    if fl < f0 and fr < f0:
        print "Error! f(xl)/f(xr) = ", fl, fr, "both smaller than ", f0, "!"
        return 0, 0, 0, 'E'
    
    xm = xl+(xr-xl)*(f0-fl)/(fr-fl); fm = f(xm)
    step = 0
    while abs(fm-f0) > tol and step <= maxstep:
        if (fm > f0 and fl > f0) or (fm < f0 and fl < f0):
            xl = xm; fl = fm
        else:
            xr = xm; fr = fm
        xm = xl+(xr-xl)*(f0-fl)/(fr-fl); fm = f(xm)
        step +=1
    return xm, fm-f0, step, 'S' # S means success 
    
```

```python
# MDR1 cosmological parameters (dft means default)

h_dft=0.70; om_dft=0.27; omb_dft=0.0469; ns_dft=0.96; sigma8_dft=0.82; w_dft = -1.0;

# Hubble parameter in unit of km/s/Mpc


def Hz(omegam, w, h, z):
    return 100*h*np.sqrt(omegam*(1.0+z)**3.0 + (1.0-omegam)*(1.0+z)**(3.0*(1+w)))

# comoving r in unit of Mpc/h
# in fact the result is independt on the input variable h

def comov_r(omegam, w, h, z):
     x, y = scipy.integrate.quad(lambda x: 1.0/Hz(omegam, w, h, x), 0, z)
     return const_c * x * h 

# invert of the above function: get z according to comoving_r (in unit of Mpc/h)

def get_z(omegam, w, h, given_comov_r, zl=0.0, zr=5.0):
    z, deltar, num_step, flag = FindRoot(lambda z: comov_r(omegam, w, h, z), given_comov_r, zl, zr)
    return z

def km_to_Mpctoh(v, h):
    return v / (3.08567758e19) * h

# plottings of H(z)

X=np.linspace(0,1,100) 
Y=[ Hz(om_dft,  w_dft, h_dft, X[i]) for i in range(len(X)) ]    
plt.plot( X, Y, 'k', lw=2)
Y=[ Hz(om_dft, -2.0, h_dft, X[i]) for i in range(len(X)) ]  
plt.plot( X, Y, 'g:', lw=2)
Y=[ Hz(om_dft, -0.5, h_dft, X[i]) for i in range(len(X)) ]  
plt.plot( X, Y, 'b--', lw=2)
plt.xlabel("redshift"); plt.ylabel("H(z) [km/s/Mpc]"); plt.title("H(z) of MDR1 Cosmology");
plt.legend(('w = -1 (LCDM)', 'w = -2.0', 'w = -0.5'), 'upper left', shadow=True)
plt.show()

# Test whether get_z can get the right z

print 'Test of the get_z routine:\n'
testz=0.0
while testz < 3.0:
    testr=comov_r(om_dft, w_dft, h_dft, testz)
    print 'input z, r(z) (Mpc/h), get_z(r(z)) = ', testz, testr, get_z(om_dft, w_dft, h_dft, testr)
    testz +=0.5
```

## Sec 3. RSD & AP Effect on Single Halos

```python
# This class describes the full information of a single halo: 
# position, velocity, redshift, RSD effect, AP effect, ...

# The position of the observer looking at the halos is x,y,z=0,0,0. 
#  This will be used to calculate the observed redshift z and comoving r

# All lengths are in unit Mpc/h; All velocities are in unit km/s

# The structure is named as "point" rather than "halo", since it can be anything (e.g., particles, galaxies, ...)

class point:
    x =0.0;  y =0.0; z =0.0
    vx=0.0;  vy=0.0; vz=0.0
    theta=0.0; phi=0.0 
    unit_vector = [0.0,0.0,0.0]
    vlos=0.0 
    
    z_real = 0.0
    z_obs = 0.0
    z_obs_diff = 0.0
    z_obs_discrep = 0.0
    
    r =0.0
    r_obs = 0.0
    r_obs_diff = 0.0
    r_obs_discrep = 0.0
    xyz_obs = [0,0,0]
    
    Hubble = 0.0
    comov_dist = 0.0
    D_A = 0.0
    fidu_cosmo = [[0.0,0.0,0.0]]
    num_fidu_cosmo = 1
    r_inferred_no_RSD = [0.0]
    r_inferred_no_RSD_diff = [0.0]
    r_inferred_no_RSD_discrep = [0.0]
    xyz_inferred_no_RSD = [[0,0,0]]
    r_inferred_with_RSD = [0.0]
    r_inferred_with_RSD_diff = [0.0]
    r_inferred_with_RSD_discrep = [0.0]
    xyz_inferred_with_RSD = [[0,0,0]]
    
    # initialize 
    
    def __init__(self, x,y,z, vx,vy,vz, fidu_cosmo = [[om_dft,w_dft,h_dft]] ): 
        
        # basic information
        self.x, self.y, self.z = x,y,z
        self.vx, self.vy, self.vz = vx,vy,vz
        self.r, self.theta, self.phi = getSC(self.x,self.y,self.z)
        self.unit_vector = [ self.x/self.r, self.y/self.r, self.z/self.r ]
        self.vlos = np.dot(self.unit_vector, [self.vx, self.vy, self.vz]) 
        #los velocity in comoving space, in unit of Mpc/h/s        
        
        # cosmological information
        self.z_real = get_z(om_dft, w_dft, h_dft, self.r)
        self.Hubble = Hz(om_dft, w_dft, h_dft, self.z_real)
        self.D_A = self.r / (1.0 + self.z_real)
        
        # observed quantities (considering RSD)
        self.z_obs = self.z_real + self.vlos / const_c * (1.0 + self.z_real) # observed, apparent redshift 
        self.z_obs_diff = self.z_obs - self.z_real 
        self.z_obs_discrep = self.z_obs_diff / self.z_real
        self.r_obs = comov_r(om_dft, w_dft, h_dft, self.z_obs)
        self.r_obs_diff = self.r_obs - self.r
        self.r_obs_discrep = self.r_obs_diff / self.r
        ratio = self.r_obs / self.r
        self.xyz_obs = [ ratio*self.x, ratio*self.y, ratio*self.z ] # apparent x,y,z
        
        # inferred information (with assumed fiducial cosmology)
        self.fidu_cosmo = fidu_cosmo
        self.num_fidu_cosmo = len(fidu_cosmo)
        self.r_inferred_no_RSD = range(self.num_fidu_cosmo)
        self.r_inferred_with_RSD = range(self.num_fidu_cosmo)
        self.r_inferred_no_RSD_diff = range(self.num_fidu_cosmo)
        self.r_inferred_with_RSD_diff = range(self.num_fidu_cosmo)
        self.r_inferred_no_RSD_discrep = range(self.num_fidu_cosmo)
        self.r_inferred_with_RSD_discrep = range(self.num_fidu_cosmo)
        self.xyz_inferred_no_RSD = [ [0.0, 0.0, 0.0] for row in range(self.num_fidu_cosmo) ]
        self.xyz_inferred_with_RSD = [ [0.0, 0.0, 0.0] for row in range(self.num_fidu_cosmo) ]
        for i in range(self.num_fidu_cosmo):
            om_fd, w_fd, h_fd = self.fidu_cosmo[i][0], self.fidu_cosmo[i][1], self.fidu_cosmo[i][2]
            self.r_inferred_no_RSD[i] = comov_r(om_fd, w_fd, h_fd, self.z_real)
            self.r_inferred_no_RSD_diff[i] = self.r_inferred_no_RSD[i] - self.r
            self.r_inferred_no_RSD_discrep[i] = (self.r_inferred_no_RSD[i] - self.r) / self.r
            ratio = self.r_inferred_no_RSD[i] / self.r
            self.xyz_inferred_no_RSD[i] = [ ratio*self.x, ratio*self.y, ratio*self.z ]
            self.r_inferred_with_RSD[i] = comov_r(om_fd, w_fd, h_fd, self.z_obs)
            self.r_inferred_with_RSD_diff[i] = self.r_inferred_with_RSD[i] - self.r
            self.r_inferred_with_RSD_discrep[i] = (self.r_inferred_with_RSD[i] - self.r) / self.r
            ratio = self.r_inferred_with_RSD[i] / self.r
            self.xyz_inferred_with_RSD[i] = [ ratio*self.x, ratio*self.y, ratio*self.z ]
    
    # print information    
    def speak(self):
        print '  Information of a point object (distances Mpc/h, velocities km/s)\n'
        print '     x, y, z                         =   ', self.x,self.y,self.z
        print '     comoving r                      =   ', self.r
        print '     theta, phi                      =   ', self.theta,self.phi
        print '     vx, vy, vz                      =   ', self.vx, self.vy, self.vz
        print '     line of sight speed             =   ', self.vlos
        print '     cosmological redshift z         =   ', self.z_real
        print '     Hubble (km/s/Mpc)               =   ', self.Hubble
        print '     D_A (Mpc/h)                     =   ', self.D_A
        print '\n  Observed quantities (considering RSD; discrep = (obs - real) / real)\n'
        print '     apparent redshift z (discrep)   =   ', self.z_obs, '  (', self.z_obs_discrep, ')'
        print '     apparent comoving r (discrep)   =   ', self.r_obs, '  (', self.r_obs_discrep, ')'
        
        for i in range(self.num_fidu_cosmo):
            print '\n  Inferred r  (assuming (om, w, h) =', self.fidu_cosmo[i][0], \
            self.fidu_cosmo[i][1], self.fidu_cosmo[i][2], '; discrep = (obs - real) / real) \n'
            print '     inferred comov r based on  cosmological  z (discrep)   =   ', \
            self.r_inferred_no_RSD[i], '  (', self.r_inferred_no_RSD_discrep[i] , ')'
            print '     inferred comov r based on observed RSDed z (discrep)   =   ', \
            self.r_inferred_with_RSD[i], '  (', self.r_inferred_with_RSD_discrep[i] , ')'

# print the information of the first halo, as an example
            
i = 0
x, y, z = halo_data[i,0], halo_data[i,1], halo_data[i,2]
vx, vy, vz = halo_data[i,3], halo_data[i,4], halo_data[i,5]
p0 = point( x,y,z, vx,vy,vz, assumed_cosmo  )       
print 'Informatiion of the first BMDW halo: \n'
p0.speak()
```

## Sec 4. Calculating BSkeletons

```python
########################################################################

### Useful Functions 

### Name of files; Where to save them.

########################################################################


# This defines the rule of data_name.
# There are four kinds of beta skleton:
#   with/without RSD; with/without AP effect (assumed wrong cosmology)
# Different kinds of beta skleton have differnt names

def generate_data_name(basic_name, RSD, AP):
    if AP=='F' and RSD == 'F':
        return basic_name
    if AP=='F' and RSD == 'T':
        return basic_name+'_RSD'
    if RSD == 'T':
        return basic_name+'_RSD_AP'+str(AP)
    if RSD == 'F':
        return basic_name+'_noRSD_AP'+str(AP)   

# File which saves the x,y,z scatter points...
def generate_data_file_name(basic_name, RSD, AP):    
    return output_dir+generate_data_name(basic_name, RSD, AP)

# some strings related with a beta value
#  DIR: directory saving the files; 
#  index_file: file saving index of halo pairs (generated by ngl-beta)
#  basic_info_file: file saving basic information (coordinates and velocities) of halo pairs

def get_strings(data_name, beta):
    char_beta='%.1f'%beta
    DIR=output_dir+data_name+'_beta_'+char_beta
    index_file = DIR+'/pair_index_'+char_beta
    # In clean index file half of the over counted halo pairs are removed
    # Usually it is done by requiring the first halo has index smaller than the second
    clean_index_file = index_file+'.clean'
    basic_info_file=DIR+'/pair_basic_info_'+char_beta
    return DIR, index_file, clean_index_file, basic_info_file


# processing the pairs
# will do two things 
#  1) remove redudant pairs in index_file
#  2) generate files saving coordinates / velocities of halo pairs, based on index_file and data file (DF)

def proc_pair_info(data_file, index_file, clean_index_file, basic_info_file):
    A=np.loadtxt(data_file)
    B=np.loadtxt(index_file)
    # Basic information (3-d position and velocities, totally 6 numbers) of halo pairs
    C=[[0 for col in range(3)] for row in range(len(B))]
    
    i=0; j=0; 
    INDEX_FILE = open(clean_index_file, 'w')
    while i<len(B):
        i1 = B[i,0]; i2 = B[i,1]
        if i1 < i2: # the pairs are double-counted by ngl-beta; only keep half of them
            z1=A[i1,2]; z2=A[i2,2];
            INDEX_FILE.write(str(int(i1+0.001))+'   '+str(int(i2+0.001))+"\n")
            if z1<z2: #keep all vetors pointing "up" (from low-z to large-z)
                C[2*j] = A[i1]
                C[2*j+1] = A[i2]
            else:
                C[2*j+1] = A[i1]
                C[2*j] = A[i2]  
            j+=1
        i = i+1
    np.savetxt(basic_info_file,C)
    INDEX_FILE.close()
    
```

```python
##########################################

### Useful defination 1. 

#      Quickly locating files

##########################################

# Quickly locate the scatter file, pair index file, pair coord file,
#  based on the keywords of RSD, AP, beta.

def locate_scat_file(RSD, AP):
    return generate_data_file_name(halo_data_name, RSD, AP)

def locate_BSke_index_file(RSD, AP, beta):
    return get_strings(generate_data_name(halo_data_name, RSD, AP), beta)[2]

def locate_BSke_coord_file(RSD, AP, beta):
    return get_strings(generate_data_name(halo_data_name, RSD, AP), beta)[3]

print 'beta = 3.0 skeleton of RSD considered, AP effect 0 (cosmo par = ',assumed_cosmo[0],\
      ') sample is saved in the file \n  ', \
    locate_BSke_coord_file('T', 0, 3.0)
```

```python
##########################################

### Useful defination 2. 

#  Select out the scatter & pair data 
#   inside some rectangular region

##########################################

def select_scat_data(inputdata, xmin, xmax, ymin, ymax, zmin, zmax):
    index =  np.where((inputdata[:,0]<xmax) & (inputdata[:,0]>xmin) &
                   (inputdata[:,1]<ymax) & (inputdata[:,1]>ymin) &
                   (inputdata[:,2]<zmax) & (inputdata[:,2]>zmin))
    return inputdata[index], index

def select_pair_data(PairData, xmin, xmax, ymin, ymax, zmin, zmax):
    tmp1 = [[0 for col in range(3)] for row in range(len(PairData))]
    tmp2 = [0 for row in range(len(PairData)/2)]
    i=0; j=0;
    while i<len(PairData):
        x1 = PairData[i][0]; y1 = PairData[i][1]; z1 = PairData[i][2];
        x2 = PairData[i+1][0]; y2 = PairData[i+1][1]; z2 = PairData[i+1][2];
        if xmin<x1<xmax and ymin<y1<ymax and zmin<z1<zmax and xmin<x2<xmax and ymin<y2<ymax and zmin<z2<zmax:
            tmp1[j][0:3] = [x1, y1, z1]
            tmp1[j+1][0:3] = [x2, y2, z2]
            tmp2[j/2] = i/2
            j+=2
        i+=2
    selected_data = [[0 for col in range(3)] for row in range(j)]
    index = [0 for row in range(j/2)]
    for i in range(j):
        selected_data[i][0:3]=tmp1[i][0:3] 
    for i in range(j/2):
        index[i]=tmp2[i]
    return selected_data, index

#### Example.

AP = 'F'
RSD = 'F'
beta = 3.0
scat_data = np.loadtxt(locate_scat_file(RSD, AP))
BSke_coord_data = np.loadtxt(locate_BSke_coord_file(RSD, AP, beta))
BSke_index_data = np.loadtxt(locate_BSke_index_file(RSD, AP, beta))

print select_scat_data(scat_data, 0, 100, 0, 50, 0, 50)
print select_pair_data(BSke_coord_data, 0, 100, 0, 100, 0, 100)
```

```python
##########################################

### Useful defination 2. 

#  3-d and 2-d plottings.

##########################################

def list_plot3Dske(ScatData_List, PairData_List, Title_List=[],\
                   xmin=-10000, xmax=10000, ymin=-10000, ymax=10000, zmin=-10000, zmax=10000):
    num_scat_data = len(ScatData_List)
    num_pair_data = len(PairData_List)
    
    if(num_scat_data != num_pair_data):
        print 'ERROR! num_scat_data ', num_scat_data , 'and num_pair_data',num_pair_data,'not match!'
        return
    if(len(title_list) != num_scat_data and len(title_list) != 0):
        print 'ERROR! num_scat_data ', num_scat_data , 'and num_title',len(title_list),'not match!'
        return
    
    fig_row =  num_scat_data /3 + min(1,np.mod(num_scat_data,3))
    fig=plt.figure(figsize=(21,6*fig_row));
    
    for k in range(num_scat_data):    
        ax=fig.add_subplot(fig_row,3,k+1,projection='3d');
        ax.set_xlabel('x'); ax.set_ylabel('y'); ax.set_zlabel('z');
        if len(Title_List) != 0:
            ax.set_title(Title_List[k])
        
        # Plotting scatter data
        SData, index_array = select_scat_data(ScatData_List[k], xmin, xmax, ymin, ymax, zmin, zmax)
        NSP = len(SData)
        X= [SData[col][0] for col in range(NSP)]
        Y= [SData[col][1] for col in range(NSP)]    
        Z= [SData[col][2] for col in range(NSP)]
        ax.scatter(X,Y,Z,color='black');
        
        # Plotting pair lines
        LData, index_array = select_pair_data(PairData_List[k], xmin, xmax, ymin, ymax, zmin, zmax)
        NLP = len(LData)
        for i in range(0,NLP,2):
            X=[LData[i][0], LData[i+1][0]]
            Y=[LData[i][1], LData[i+1][1]]
            Z=[LData[i][2], LData[i+1][2]]
            ax.plot(X, Y, Z, color='black', lw=1)
            
    plt.show()   

    
def list_plot2Dske(ScatData_List, PairData_List, Title_List=[],\
                   xmin=-10000, xmax=10000, ymin=-10000, ymax=10000, zmin=-10000, zmax=10000):
    num_scat_data = len(ScatData_List)
    num_pair_data = len(PairData_List)
    
    if(num_scat_data != num_pair_data):
        print 'ERROR! num_scat_data ', num_scat_data , 'and num_pair_data',num_pair_data,'not match!'
        return
    if(len(title_list) != num_scat_data and len(title_list) != 0):
        print 'ERROR! num_scat_data ', num_scat_data , 'and num_title',len(title_list),'not match!'
        return
    
    fig_row =  num_scat_data /3 + min(1,np.mod(num_scat_data,3))
    fig=plt.figure(figsize=(21,6*fig_row));
    
    for k in range(num_scat_data):    
        ax=fig.add_subplot(fig_row,3,k+1);
        ax.set_xlabel('x'); ax.set_ylabel('y')
        if len(Title_List) != 0:
            ax.set_title(Title_List[k])
        
        # Plotting scatter data
        SData, index_array = select_scat_data(ScatData_List[k], xmin, xmax, ymin, ymax, zmin, zmax)
        NSP = len(SData)
        X= [SData[col][0] for col in range(NSP)]
        Y= [SData[col][1] for col in range(NSP)]    
        ax.scatter(X,Y,color='black');
        
        # Plotting pair lines
        LData, index_array = select_pair_data(PairData_List[k], xmin, xmax, ymin, ymax, zmin, zmax)
        NLP = len(LData)
        for i in range(0,NLP,2):
            X=[LData[i][0], LData[i+1][0]]
            Y=[LData[i][1], LData[i+1][1]]
            ax.plot(X, Y, color='black', lw=1)
            
    plt.show() 
    
```
