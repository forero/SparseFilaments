```python
#./calc_grad_chisqs chisqlike/smnum30_newdrop_sigis0.50__fixnuminx60__test 
#chisqlike/smnum30_newdrop_sigis0.50__fixnuminx60__test

import matplotlib.pyplot as plt
import numpy as np
import commands
import mpl_toolkits.mplot3d.axes3d as Axes3D
%pylab inline
import os
import random
import scipy.integrate
import gc
import re

#################################
# Useful Definiations
#################################

FIG_X_SIZE = 18
FIG_Y_SIZE = 6
FIG_FONT_SIZE = 18

plt.rc('xtick', labelsize=20) 
plt.rc('ytick', labelsize=20)

OM_DFT = 0.26
W_DFT = -1.0

CONST_C = 299792.458 #unit: km/s

def w_kernel(q):
    if q < 0:
        return 0
    if q < 1.0:
        return  (0.25*(2.0-q)**3.0 - (1.0-q)**3.0)
    if q < 2.0:
        return  0.25*(2.0-q)**3.0
    return 0

def der_w_kernel(q):
    if q < 0:
        return 0
    if q < 1.0:
        return  (-0.75*(2.0-q)**2.0 + 3.0*(1.0-q)**2.0)
    if q < 2.0:
        return  -0.75*(2.0-q)**2.0
    return 0

def W_kernel(r, h):
    q = r / (h+0.0)
    return w_kernel(q) / (h+0.0)

def der_W_kernel(r, h):
    q = r / (h+0.0)
    return der_w_kernel(q) / (h+0.0)

def ilist(i, di, imin, imax):
    i1 = i - di
    i2 = i + di
    if i1 < imin and i2 > imax:
        return range(imin, imax+1)
    if i1 < imin:
        return range(imin, imin + 2*di +1)
    if i2 > imax:
        return range(imax - 2*di, imax +1)
    return range(i1, i2+1)

def do_average(gv_chisq_array, av_width = 0.01):
    n = len(gv_chisq_array)
    chisq_array = [[gv_chisq_array[j][0], gv_chisq_array[j][1]] for j in range(n)]
    X = [chisq_array[j][0] for j in range(n)]
    Y = [chisq_array[j][1] for j in range(n)]
    if True:
        for i in range(n):
            x0 = chisq_array[i][0]
            av_chisq_list = [chisq_array[i][1]]
            # left side search
            j = i-1; 
            while j >= 0:
                x = chisq_array[j][0]
                if abs(x-x0) < av_width:
                    av_chisq_list.append(chisq_array[j][1])
                    j -= 1
                else:
                    break
            # right side search
            j = i+1; 
            while j < len(chisq_array):
                x = chisq_array[j][0]
                if abs(x-x0) < av_width:
                    av_chisq_list.append(chisq_array[j][1])
                    j += 1
                else:
                    break
            # averaging
            chisq_array[i][1] = sum(av_chisq_list) / len(av_chisq_list)
    return chisq_array

def find_mean_var(n_list):
    n = len(n_list)
    n_mean = sum(n_list) / (n+0.0)
    n_var = sum([(n_list[row]-n_mean)**2.0 for row in range(n)]) / (n-1.0)
    return n_mean, n_var

###################
## smoothing curve
###################
def do_smooth(gv_chisq_array, sm_width = 0.01, sm_weight = 10, keep_symmetry = True, fast_dx_alg = False, \
              spline_kernel=True, rescale_sm_width = False, print_info=False):
    n = len(gv_chisq_array)
    chisq_array = [[gv_chisq_array[j][0], gv_chisq_array[j][1]] for j in range(n)]
    X = [chisq_array[j][0] for j in range(n)]
    Y = [chisq_array[j][1] for j in range(n)]
    xmin = min(X); xmax = max(X);
    if fast_dx_alg:
        dx = X[1] - X[0]
        m = int((sm_width/dx)+0.5) 
        if m<1:
            if print_info:
                print 'found sm_width < dx: ',sm_width, dx, '. Will not do smoothing...'
            return gv_chisq_array
        #print 'sm_width, dx, m, dx*m = ', sm_width, dx, m, dx*m
        if spline_kernel:
            if rescale_sm_width and print_info:
                print 'rescalling sm_with to ', m*dx, '...'
            else:
                dx = sm_width / (m+0.0)
            h = dx*m / 2.0
            sm_kernel_list = [W_kernel(dx*row, h) for row in range(m+2)]
        else:
            sm_kernel_list = [np.exp(-(row/(m+0.0))**2.0/sm_weight) for row in range(m+2)]
        # plot to test
        #X = [dx*row for row in range(len(sm_kernel_list))]
        #YX = sm_kernel_list
        #plot(X,YX)
        for i in range(n):
            lefti = max(0,i-m); leftdi = i - lefti;
            righti = min(i+m,n-1); rightdi = righti -i;
            #print 'i, lefti, leftdi, righti, rightdi = ', i, lefti, leftdi, righti, rightdi
            sum_kernel = sm_kernel_list[0]; sum_y = Y[i]*sm_kernel_list[0];
            if keep_symmetry:
                di = min(leftdi,rightdi);
                leftdi = i-di; rightdi = i+di
                #print '(1,di+1) = ', range(1,di+1)
                for j in range(1,di+1):
                    sum_kernel += 2*sm_kernel_list[j]
                    #if i-j <0 or i+j > n-1:
                    #    continue
                    #print j, i-j, i+j, len(sm_kernel_list)
                    sum_y += sm_kernel_list[j]*Y[i-j]
                    sum_y += sm_kernel_list[j]*Y[i+j]
            else:
                for j in range(1,leftdi+1):
                    #if i-j <0:
                    #    continue
                    sum_kernel += sm_kernel_list[j]
                    sum_y += sm_kernel_list[j]*Y[i-j]
                for j in range(1,rightdi+1):
                    #if i+j > n:
                    #    continue
                    sum_kernel += sm_kernel_list[j]
                    sum_y += sm_kernel_list[j]*Y[i+j]
            chisq_array[i][1] = sum_y / sum_kernel
    else:        
        for i in range(n):
            x0 = chisq_array[i][0]
            # left side search
            x_list = []; y_list = [];
            j = i-1; 
            left_dx = 0
            while j >= 0:
                x = chisq_array[j][0]
                if abs(x-x0) < sm_width:
                    x_list.append(x)
                    y_list.append(Y[j])
                    left_dx = max(left_dx, abs(x-x0))
                    j -= 1
                else:
                    break
            # right side search
            j = i+1; 
            right_dx = 0
            while j < n:
                x = chisq_array[j][0]
                if abs(x-x0) < sm_width:
                    x_list.append(x)
                    y_list.append(Y[j])
                    right_dx = max(right_dx, abs(x-x0))
                    j += 1
                else:
                    break
            dx = sm_width
            if keep_symmetry:
                dx = min(sm_width, left_dx*1.1, right_dx*1.1)
                for j in range(len(x_list)):
                    x = x_list[j]
                    if abs(x-x0) > dx:
                        y_list[j] = 0.0
            m = len(x_list)
            # do smooth
            if m == 0:
                continue
            else:
                if spline_kernel:
                    h = dx / 2.0
                    sm_kernel_list = [W_kernel(abs(x0-x_list[row]), h) for row in range(m)]
                else:
                    sm_kernel_list = [np.exp(-((x0-x_list[row])/dx)**2.0/sm_weight) for row in range(m)]
                weighted_y_list = [y_list[row]*sm_kernel_list[row] for row in range(m)]
                y = sum(weighted_y_list)/sum(sm_kernel_list)
                chisq_array[i][1] = y
    return chisq_array

def get_like_from_chisq(chisq_array):
    n = len(chisq_array)
    Y = [chisq_array[j][1] for j in range(n)]
    Ymin = min(Y)
    Y = [exp(-0.5*(chisq_array[j][1]-Ymin)) for j in range(n)]
    like_array = [[chisq_array[j][0],Y[j]] for j in range(n)]
    return like_array

def rescale_to_norm(A):
    n = len(A)
    Y = [A[j][1] for j in range(n)]
    Ymax = max(Y)
    return [[A[j][0],(Y[j]/Ymax)] for j in range(n)]

def plot_distributs(chisq_array_list, legend_list, title='', style_list = [],xlabel='x',ylabel='y', text='',\
                    textlocx=0.6, textlocy=0.2, set_xylim=False, xleft=0.1,xright=0.6,\
                    savfig=False,filename='distribution'):
    n = len(chisq_array_list)
    if style_list == []:
        style_list = ['-' for row in range(n)]
    if(len(legend_list) != n or len(style_list) != n):
        print 'ERROR! # of legend/style shall be ', n
        return
    figure = plt.figure(figsize=(FIG_X_SIZE,FIG_Y_SIZE));
    ax = figure.add_subplot(111);
    ax.set_title(title, fontsize = FIG_FONT_SIZE+5);
    ax.set_xlabel(xlabel, fontsize=FIG_FONT_SIZE+5); ax.set_ylabel(ylabel, fontsize=FIG_FONT_SIZE+5);
    if set_xylim:
        ax.set_xlim(xleft, xright);
        #ax.set_ylim(yleft, yright);
    for i in range(n):
        X = [chisq_array_list[i][j][0] for j in range(len(chisq_array_list[i]))]
        Y = [chisq_array_list[i][j][1] for j in range(len(chisq_array_list[i]))]
        if i == 1:
            if not set_xylim:
                xmax = max(X); xmin = min(X); ymax = max(Y); ymin = min(Y)
            else:
                xmax = xleft; xmin = xright; ymax = max(Y); ymin = min(Y)
        ax.plot(X,Y,style_list[i], marker = 'o', lw=2);
    leg = ax.legend(legend_list, 'best');
    for label in leg.get_texts():
        label.set_fontsize(FIG_FONT_SIZE)
    #leg.legend(fontsize=FIG_FONT_SIZE)
    leg.get_frame().set_alpha(0.5);
    ax.text(xmin+textlocx*(xmax-xmin),ymin+textlocy*(ymax-ymin),text,color='red',fontsize=FIG_FONT_SIZE-2);
    if savfig:
        figure.savefig(filename,format='pdf')
    plt.show()
    
    
def gen_ini(fortran_ini_file, smnum, num_in_x, auto_num_in_x=False, \
        check_boundary=True, print_info=False,  cb_adjust_ratio=1.0, \
        remov_dist_ratio=1.0,  use_num_density=True, nbins_av = 10, use_intpl_rho=True, numdrop=0,  \
        num_changenuminx = 10, amp_changenuminx = 0.1, num_om_w=0, \
        dropval=[], lowdropvalratio=[], highdropvalratio=[],
        dropdval=[], lowdropdvalratio=[], highdropdvalratio=[],
        om_w_list=[] ):
    f = open(fortran_ini_file, "w")
    if len(dropval) != numdrop or len(dropdval)!= numdrop \
        or len(lowdropvalratio)!= numdrop or len(lowdropdvalratio)!= numdrop \
        or len(highdropvalratio)!= numdrop or len(highdropdvalratio)!= numdrop:
            print 'ERROR! # of drops mismatches with ', numdrop
            return
    if len(om_w_list) != num_om_w:
            print 'ERROR! # of om_w mismatches with ', num_om_w
            return
    # General info
    nowstr = str(smnum)+'  '+str(num_in_x)+'   '+str(auto_num_in_x)+'   '+str(check_boundary)+'   '\
        +str(print_info)+'   '+str(cb_adjust_ratio)+'   '+str(remov_dist_ratio)+'   '\
        +str(use_num_density)+'   '+str(nbins_av)+'   '+str(use_intpl_rho)+'   '+str(numdrop)+'   '\
        + str(num_changenuminx)+'   '+str(amp_changenuminx)+'   '+str(num_om_w)+'\n'
    f.write(nowstr)
    ## Many info about drops
    for i in range(numdrop):
        nowstr = str(dropval[i])+'  '+str(lowdropvalratio[i])+'  '+str(highdropvalratio[i])\
                +'  '+str(dropdval[i])+'  '+str(lowdropdvalratio[i])+'  '+str(highdropdvalratio[i])+'\n'
        f.write(nowstr)
    # om_w_list
    for i in range(len(om_w_list)):
        nowstr=str(om_w_list[i][0])+'  '+str(om_w_list[i][1])+'\n'
        f.write(nowstr)
    f.close()
    return

def gen_chisq_list(pmin, pmax, num_chisqs, om_or_w='om', om_dft = OM_DFT, w_dft = W_DFT):
    dp = (pmax-pmin) / (num_chisqs - 1.0)
    if om_or_w == 'om':
        om_w_list = [ [pmin + dp*row, w_dft] for row in range(num_chisqs)] 
    else:
        om_w_list = [ [om_dft, pmin + dp*row] for row in range(num_chisqs)] 
    return om_w_list

def rm_nearby_element(A, tol):
    for i in range(len(A)):
        x = A[i]
        if x == 'NONE':
            continue
        for j in range(i+1,len(A)):
            y = A[j]
            if y =='NONE' :
                continue
            if abs(x-y) < tol:
                A[j] = 'NONE'
    A = filter(lambda x: x != 'NONE', A)
    return A

def get_margomw(data):
    tol = 1e-7;
    margomlist = []
    margwlist = []
    margwchisqlist = []
    for i in range(len(data)):
        om, w, chisq = data[i,0:4]
        margomlist.append(om)
        margwlist.append(w)
    margomlist = list(set(margomlist))
    margwlist = list(set(margwlist))
    #print len(margomlist), len(margwlist)
    # remove close elements
    margomlist = rm_nearby_element(margomlist, tol);
    margwlist = rm_nearby_element(margwlist, tol);
    margomlist.sort()
    margwlist.sort()
    #print len(margomlist), len(margwlist)
    nom = len(margomlist)
    nw = len(margwlist)
    margomchisqlist = [[0,0] for row in range(nom)]
    margwchisqlist = [[0,0] for row in range(nw)]
    for i in range(len(data)):
        om, w, chisq = data[i,0:4]
        for iom in range(nom):
            margom = margomlist[iom]
            if abs(margom - om) < tol:
                margomchisqlist[iom][0] += chisq
                margomchisqlist[iom][1] += 1
                break
        for iw in range(nw):
            margw = margwlist[iw]
            if abs(margw - w) < tol:
                margwchisqlist[iw][0] += chisq
                margwchisqlist[iw][1] += 1
                break
    for iom in range(nom):
        margomchisqlist[iom] = margomchisqlist[iom][0] / ( margomchisqlist[iom][1] + 0.0)
    for iw in range(nw):
        margwchisqlist[iw] = margwchisqlist[iw][0] / (margwchisqlist[iw][1] + 0.0)
    return [[margomlist[row], margomchisqlist[row]] for row in range(nom)], \
            [[margwlist[row], margwchisqlist[row]] for row in range(nw)]

# Marginalized chisq
def plot_marg_chisq_list(file_list, base_legend_list, \
                    sum_up_bins, smooth_chisq, smooth_like, sm_width, sm_weight,\
                    average_chisq, average_like, av_width, usemargchisq = True, \
                    spline_smooth=True, keep_symmetry=False,  plot_chisq=True, plot_like=True,\
                    fast_dx_alg=True, print_info = True, text='', textlocx=0.6, textlocy=0.2,\
                    set_xylim=False, xleft=0.1,xright=0.6, \
                    set_wlim=True, wleft=-3,wright=0, \
                    savfig=False,basefigname='distribution'):   
    w_om_sm_ratio = 5.0
    om_chisq_array_list = []
    w_chisq_array_list = []
    om_like_array_list = []
    w_like_array_list = []
    legend_list = []
    style_list = []
    for i_file in range(len(file_list)):
        data = np.loadtxt(file_list[i_file])
        margomchisq, margwchisq = get_margomw(data)
        if print_info:
            print '\tfile_index, # of data = ', i_file, len(margomchisq), '...'
        n = len(margomchisq)
        if True:
            legend_list.append(base_legend_list[i_file])
            if mod(i_file,2) == 0:
                style_list.append('-o')
            else:
                style_list.append('--o')
            if smooth_chisq:
                margomchisq = do_smooth(margomchisq, sm_width=sm_width, spline_kernel=spline_smooth, \
                    print_info=print_info, sm_weight=sm_weight, keep_symmetry=keep_symmetry, fast_dx_alg=fast_dx_alg)
                margwchisq = do_smooth(margwchisq, sm_width=sm_width*w_om_sm_ratio, spline_kernel=spline_smooth, \
                    print_info=print_info, sm_weight=sm_weight, keep_symmetry=keep_symmetry, fast_dx_alg=fast_dx_alg)
            if average_chisq:
                margomchisq = do_average(margomchisq, av_width=av_width)
                margwchisq = do_average(margwchisq, av_width=av_width*w_om_sm_ratio)
            om_chisq_array_list.append(margomchisq)
            w_chisq_array_list.append(margwchisq)
            if usemargchisq:
                omlike = get_like_from_chisq(margomchisq)
                wlike = get_like_from_chisq(margwchisq)
            else:
                ndat = len(data)
                Y = [data[j][2] for j in range(ndat)]
                Ymin = min(Y)
                Y = [exp(-0.5*(data[j][2]-Ymin)) for j in range(ndat)]
                for row in range(ndat):
                    data[row][2] = Y[row]
                omlike, wlike = get_margomw(data)
                omlike = rescale_to_norm(omlike)
                wlike = rescale_to_norm(wlike)
            if smooth_like:
                omlike = do_smooth(omlike, sm_width=sm_width, spline_kernel=spline_smooth, print_info=print_info,\
                              sm_weight=sm_weight, keep_symmetry=keep_symmetry, fast_dx_alg=fast_dx_alg)
                omlike = rescale_to_norm(omlike)
                wlike = do_smooth(wlike, sm_width=sm_width*w_om_sm_ratio, spline_kernel=spline_smooth, 
                    print_info=print_info, sm_weight=sm_weight, keep_symmetry=keep_symmetry, fast_dx_alg=fast_dx_alg)
                wlike = rescale_to_norm(wlike)
            if average_like:
                omlike = do_average(omlike, av_width=av_width)
                omlike = rescale_to_norm(omlike)
                wlike = do_average(wlike, av_width=av_width)
                wlike = rescale_to_norm(wlike)
            om_like_array_list.append(omlike)
            w_like_array_list.append(wlike)
    if plot_chisq:
        plot_distributs(om_chisq_array_list, legend_list, 'Marginalized $\chi^2$ distributioin of $\Omega_m$', \
                        style_list,\
                            '$\Omega_m$', '$\chi^2$', text, textlocx, textlocy, set_xylim, xleft, xright,\
                            savfig, basefigname+'__chisq_of_omegam.pdf')
        plot_distributs(w_chisq_array_list, legend_list, 'Marginalized $\chi^2$ distributioin of w', style_list,\
                            'w', '$\chi^2$', text, textlocx, textlocy, set_wlim, wleft, wright,\
                            savfig, basefigname+'__chisq_of_w.pdf')
    if plot_like:
        legend_list.append('Default Cosmology')
        style_list.append('-')
        om_like_array_list.append([[OM_DFT, row/100.0] for row in range(100)])
        w_like_array_list.append([[W_DFT, row/100.0] for row in range(100)])
        plot_distributs(om_like_array_list, legend_list, 'Marginalized Likelihood distribution of $\Omega_m$', 
                        style_list,\
                            '$\Omega_m$', 'Likelihood', text, textlocx, textlocy, set_xylim, xleft, xright,\
                            savfig, basefigname+'__like_of_omegam.pdf')

        plot_distributs(w_like_array_list, legend_list, 'Marginalized Likelihood distribution of w', style_list,\
                            'w', 'Likelihood', text, textlocx, textlocy, set_wlim, wleft, wright,\
                            savfig, basefigname+'__like_of_w.pdf')
        
def list_plot_marg_chisq_list(sm_width_list, text='', textlocx=0.5, textlocy=0.4, savfig=False, \
                        basefigname='distribution',\
                        smooth_chisq = True, smooth_like = True, \
                         average_chisq = False, average_like = False, plot_chisq = True, plot_like = True, 
                         set_xylim=False, xleft=0.1, xright=0.6, ignoreRSD = False):
        
    spline_smooth=True
    sm_width = 0.1;    sm_weight = 100;    keep_symmetry = True;
    av_width = 0.1;    sum_up_bins = False;    fast_dx_alg = True;        print_info = False;   
    
    if not ignoreRSD:
        file_list = [inifile+'_rho.txt', inifile+'_rhoRSD.txt', inifile+'_delta.txt', \
             inifile+'_deltaRSD.txt', inifile+'_ndelta.txt', inifile+'_ndeltaRSD.txt']
        base_legend_list = ['$n$', '$n$ (RSD)', '$\delta$', '$\delta$ (RSD)', \
                        '$\delta/\sqrt{<\delta\delta>}$', '$\delta/\sqrt{<\delta\delta>}$ (RSD)']
    else:
        file_list = [inifile+'_rho.txt', inifile+'_delta.txt', inifile+'_ndelta.txt']
        base_legend_list = ['$n$','$\delta$',  '$\delta/\sqrt{<\delta\delta>}$']

    for sm_width in sm_width_list:
        av_width = sm_width
        print sm_width
        plot_marg_chisq_list(file_list, base_legend_list,  \
                    sum_up_bins=sum_up_bins, smooth_chisq=smooth_chisq, \
                    smooth_like=smooth_like, sm_width=sm_width, sm_weight=sm_weight,\
                    average_chisq=average_chisq, average_like=average_like, av_width=av_width, \
                    spline_smooth=spline_smooth, keep_symmetry=keep_symmetry,  plot_chisq=plot_chisq, \
                    plot_like=plot_like, fast_dx_alg=fast_dx_alg, print_info = print_info, text=text,\
                    textlocx=textlocx, textlocy=textlocy, set_xylim=set_xylim, xleft=xleft, xright=xright,
                    savfig=savfig, basefigname=basefigname) 

def plot_chisq_list(file_list, base_legend_list, om_or_w, \
                    sum_up_bins, smooth_chisq, smooth_like, sm_width, sm_weight,\
                    average_chisq, average_like, av_width, \
                    spline_smooth=True, keep_symmetry=False,  plot_chisq=True, plot_like=True,\
                    fast_dx_alg=True, print_info = True, text='', textlocx=0.6, textlocy=0.2,\
                    set_xylim=False, xleft=0.1,xright=0.6,savfig=False,basefigname='distribution'):    
    chisq_array_list = []
    like_array_list = []
    legend_list = []
    style_list = []
    for i_file in range(len(file_list)):
        data = np.loadtxt(file_list[i_file])
        if print_info:
            print '\tfile_index, # of data = ', i_file, len(data), '...'
        n = int(len(data))
        m = len(data[0])
        if True:
            X = []; 
            legend_list.append(base_legend_list[i_file])
            if mod(i_file,2) == 0:
                style_list.append('-')
            else:
                style_list.append('--')
            for row in range(n):
                if om_or_w=='om':
                    x = data[row][0]
                else:
                    x = data[row][1]
                if sum_up_bins:
                    chisq = sum(data[row][2:m+1]) / (m-2)
                else:
                    chisq = data[row][2]
                    X.append([x, chisq])
            if smooth_chisq:
                X = do_smooth(X, sm_width=sm_width, spline_kernel=spline_smooth, print_info=print_info,\
                              sm_weight=sm_weight, keep_symmetry=keep_symmetry, fast_dx_alg=fast_dx_alg)
            if average_chisq:
                X = do_average(X, av_width=av_width)
            chisq_array_list.append(X)
            X = get_like_from_chisq(X)
            if smooth_like:
                X = do_smooth(X, sm_width=sm_width, spline_kernel=spline_smooth, print_info=print_info,\
                              sm_weight=sm_weight, keep_symmetry=keep_symmetry, fast_dx_alg=fast_dx_alg)
                X = rescale_to_norm(X)
            if average_like:
                X = do_average(X, av_width=av_width)
                X = rescale_to_norm(X)
            like_array_list.append(X)
    if plot_chisq:
        if om_or_w == 'om':
            plot_distributs(chisq_array_list, legend_list, '$\chi^2$ distributioin of $\Omega_m$', style_list,\
                            '$\Omega_m$', '$\chi^2$', text, textlocx, textlocy, set_xylim, xleft, xright,\
                            savfig, basefigname+'__chisq_of_omegam.pdf')
        else:
            plot_distributs(chisq_array_list, legend_list, '$\chi^2$ distributioin of w', style_list,\
                            'w', '$\chi^2$', text, textlocx, textlocy, set_xylim, xleft, xright,\
                            savfig, basefigname+'__chisq_of_w.pdf')
    if plot_like:
        legend_list.append('Default Cosmology')
        style_list.append('-')
        if om_or_w == 'om':
            like_array_list.append([[OM_DFT, row/100.0] for row in range(100)])
        else:
            like_array_list.append([[W_DFT, row/100.0] for row in range(100)])
        if om_or_w == 'om':
            plot_distributs(like_array_list, legend_list, 'Likelihood distribution of $\Omega_m$', style_list,\
                            '$\Omega_m$', 'Likelihood', text, textlocx, textlocy, set_xylim, xleft, xright,\
                            savfig, basefigname+'__like_of_omegam.pdf')
        else:
            plot_distributs(like_array_list, legend_list, 'Likelihood distribution of $\Omega_m$', style_list,\
                            'w', 'Likelihood', text, textlocx, textlocy, set_xylim, xleft, xright,\
                            savfig, basefigname+'__like_of_w.pdf')   

def list_plot_chisq_list(sm_width_list, text='', textlocx=0.5, textlocy=0.4, savfig=False, basefigname='distribution',\
                         om_or_w = 'om', smooth_chisq = True, smooth_like = True, \
                         average_chisq = False, average_like = False, plot_chisq = True, plot_like = True, 
                         set_xylim=False, xleft=0.1, xright=0.6,ignoreRSD=False):
        
    spline_smooth=True
    sm_width = 0.1;    sm_weight = 100;    keep_symmetry = True;
    av_width = 0.1;    sum_up_bins = False;    fast_dx_alg = True;        print_info = False;   
    
    if not ignoreRSD:
        file_list = [inifile+'_rho.txt', inifile+'_rhoRSD.txt', inifile+'_delta.txt', \
             inifile+'_deltaRSD.txt', inifile+'_ndelta.txt', inifile+'_ndeltaRSD.txt']
        base_legend_list = ['$n$', '$n$ (RSD)', '$\delta$', '$\delta$ (RSD)', \
                        '$\delta/\sqrt{<\delta\delta>}$', '$\delta/\sqrt{<\delta\delta>}$ (RSD)']
    else:
        file_list = [inifile+'_rho.txt', inifile+'_delta.txt',inifile+'_ndelta.txt']
        base_legend_list = ['$n$',  '$\delta$', '$\delta/\sqrt{<\delta\delta>}$']


    for sm_width in sm_width_list:
        av_width = sm_width
        print sm_width
        plot_chisq_list(file_list, base_legend_list, om_or_w=om_or_w,  \
                    sum_up_bins=sum_up_bins, smooth_chisq=smooth_chisq, \
                    smooth_like=smooth_like, sm_width=sm_width, sm_weight=sm_weight,\
                    average_chisq=average_chisq, average_like=average_like, av_width=av_width, \
                    spline_smooth=spline_smooth, keep_symmetry=keep_symmetry,  plot_chisq=plot_chisq, \
                    plot_like=plot_like, fast_dx_alg=fast_dx_alg, print_info = print_info, text=text,\
                    textlocx=textlocx, textlocy=textlocy, set_xylim=set_xylim, xleft=xleft, xright=xright,
                    savfig=savfig, basefigname=basefigname)

def plot_rho_chisq(name, data_index_list = [1], om_or_w = 'om', text = ''):
    filename = inifile + '_rho.txt'
    data = np.loadtxt(filename)
    ndata= len(data)
    if om_or_w == 'om':
        X = [data[row,0] for row in range(ndata)]
    else:
        X = [data[row,1] for row in range(ndata)]
    
    fig = plt.figure(figsize=(FIG_X_SIZE, FIG_Y_SIZE) )
    ax = fig.add_subplot(111, )
    for data_index in data_index_list:
        Y = [data[row, data_index+1] for row in range(ndata)]
        ax.plot(X, Y, '-ko')#,marker = '+')  
    
    ax.set_title('$\chi^2$ Distribution', fontsize = FIG_FONT_SIZE+5);
    if om_or_w  == 'om':
        ax.set_xlabel('$\Omega_m$', size=FIG_FONT_SIZE+5); 
    elif om_or_w == 'w':
        ax.set_xlabel('$w$', size=FIG_FONT_SIZE+5); 
    ax.set_ylabel('$\chi^2$', fontsize=FIG_FONT_SIZE+5);
    xmin=min(X);xmax=max(X);ymin=min(Y);ymax=max(Y);
    ax.set_xlim(xmin,xmax);
    ax.set_ylim(ymin*0.99,ymax*1.01);
    legend_list = [str(data_index_list[row]) for rwo in range(len(data_index_list))]
    leg = ax.legend(legend_list, 'best');
    for label in leg.get_texts():
        label.set_fontsize(FIG_FONT_SIZE)
        leg.get_frame().set_alpha(0.5);
    textlocx = 0.2; textlocy = 0.5;
    ax.text(xmin+textlocx*(xmax-xmin),ymin+textlocy*(ymax-ymin),text,color='red',fontsize=FIG_FONT_SIZE-2); 

def mu_histogram(mu_data, mu_title='', nbins=15):
    fig=plt.figure(figsize=(10,5));
    muhist=fig.add_subplot(111);
    #Histogram Plotting
    x=mu_data
    muhist.hist(x, nbins, range=(-1,1));
    npoint=len(x)
    #Reference PDF
    X=np.linspace(-1,1,nbins); Y=X-X+npoint/(nbins+0.0)
    muhist.plot(X, Y, color='black', lw=3);
    muhist.set_title(mu_title);
    #Estimating Error Bars
    n_array, bins_array=np.histogram(x,nbins,range=(-1,1))
    X=[(bins_array[k]+bins_array[k+1])/2 for k in range(nbins)]
    Y=n_array
    Yerr=[n_array[k]/np.sqrt(0.0+n_array[k]) for k in range(nbins)]
    muhist.errorbar(X, Y, yerr=Yerr, color='red', lw=3, fmt='p');    
    plt.show();
    
    
def show_data(data, show_scatter = False):
    ndat = len(data)
    if ndat > 10000:
        ratio = 10000 / (ndat + 0.0)
    nowx = []; nowy = []; nowz = [];
    nowmu = [];
    for i in range(ndat):
        nowmu.append(data[i][8])
        x = random.uniform(0,1);
        if x < ratio:
            nowx.append(data[i][0])
            nowy.append(data[i][1])
            nowz.append(data[i][2])
    fig=plt.figure(figsize=(10,7));
    if show_scatter:
        ax=fig.add_subplot(111,projection='3d');
        ax.scatter(nowx,nowy,nowz);
        plt.show();
        plt.clf();
    mu_histogram(nowmu, nbins=15)
    abs_mu_histogram(nowmu, nbins=7)
    plt.clf();    

    
Do_Plottings_for_all = False

basename = 'chisqlike/'
```
