```python
### Functions for Getting beta-skeletons

# Dir/File names related with beta. 
#  DIR: directoryr of save the files.
#  IF: pair index file. 
#  CF: pair coord file.
def get_strings(beta):
    char_beta='%.1f'%beta
    DIR='beta_'+char_beta
    IF=DIR+'/pair_index_'+char_beta
    CF=DIR+'/pair_coord_'+char_beta
    return DIR, IF, CF

# Get the CF file based on the DF (datafile) and IF (index file)
def get_cor(DF, IF, CF):
    A=np.loadtxt(DF)
    B=np.loadtxt(IF)
    C=[[0 for col in range(3)] for row in range(len(B))]
    i=0; j=0; 
    while i<len(B):
        i1 = B[i,0]; i2 = B[i,1]
        if i1 < i2: # the pairs are double-counted by ngl-beta; only keep half of them
            z1=A[i1,2]; z2=A[i2,2];
            if z1<z2: #keep all vetors pointing "up" (from low-z to large-z)
                C[2*j] = A[i1]
                C[2*j+1] = A[i2]
            else:
                C[2*j+1] = A[i1]
                C[2*j] = A[i2]  
            j+=1
        i = i+1
    np.savetxt(CF,C)
    
```

```python
# Generating beta-skeletons for a beta array

# Generating the IF and CF for a series of beta; saving them in a directory for each beta    
for beta in beta_array:
    DIR, IF, CF = get_strings(beta)
    print 'Generating neighbour graph for beta = ', beta, '...'
    command='mkdir '+DIR
    !$command
    command='../ngl-beta/build/binsrc/getNeighborGraph -i'+' '+DF+' -d 3 -b'+' '+'%.lf'%beta+' >> '+IF
    !$command
    print 'Getting coordinates for beta = ', beta
    #Fortran code shall be much faster than python
    #!./getcor $DF $IF $CF 
    get_cor(DF, IF, CF)
```

```python
# Histogram Statistics of Spherical Cor r, theta, phi of a given set of x,y,z 

def getangle(x, y):
    theta = arcsin(y/sqrt(x**2+y**2))
    if x<0:
        theta = pi-theta
    if x>0 and y<0:
        theta = 2*pi+theta
    return theta

def getSC(x, y, z):
    r=sqrt(x**2+y**2+z**2)
    theta=getangle(z,sqrt(x**2+y**2))
    phi=getangle(x,y)
    return r, theta, phi

def SCSta(beta):
    DIR, IF, CF = get_strings(beta)
    CFdata=np.loadtxt(CF)
    ndat=len(CFdata)/2
    rdata=range(ndat)
    thetadata=range(ndat)
    phidata=range(ndat)
    for i in range(ndat):
        rdata[i], thetadata[i], phidata[i] = getSC(CFdata[2*i+1,0]-CFdata[2*i,0], CFdata[2*i+1,1]-CFdata[2*i,1], CFdata[2*i+1,2]-CFdata[2*i,2])
        #if thetadata[i] > pi/2:
        #    thetadata[i]= pi - thetadata[i]
        #if phidata[i] > pi:
        #    phidata[i] = phidata[i] - pi
    return rdata, thetadata, phidata

def get_bin_eb(xdata, nbins):
    ndat=len(xdata)
    nbins

def SCplot(rdata, thetadata, phidata): 
    print 'Number of data = ', len(rdata), len(thetadata), len(phidata)
    nbins = 20
    print 'Number of bins = ', nbins
    print 'Histogram for R (left), Cos(theta) (middle left), theta (middle right), phi (right): '
    
    
    #r
    fig=plt.figure(figsize=(20,5))
    rhist=fig.add_subplot(141)
    rhist.hist(rdata,  bins=nbins)
    x=rdata; np=len(rdata)
    #Estimate error bars
    n_array, bins_array=histogram(x,nbins)
    X=[(bins_array[k]+bins_array[k+1])/2 for k in range(nbins)]
    Y=n_array
    Yerr=[n_array[k]/sqrt(n_array[k]+0.0) for k in range(nbins)]
    plt.errorbar(X, Y, yerr=Yerr, color='red', lw=3, fmt='p')
    
    #cos(theta)
    thetahist=fig.add_subplot(142)
    thetahist.hist(cos(thetadata), range=(0,1), bins=nbins)
    x=cos(thetadata); np=len(thetadata)
    #Reference PDF
    X0=linspace(0,1,nbins); Y0=X0-X0+np/(nbins+0.0)
    thetahist.plot(X0, Y0, color='black', lw=3)
    #Estimate error bars
    n_array, bins_array=histogram(x,nbins,range=(0,1))
    X1=[(bins_array[k]+bins_array[k+1])/2 for k in range(nbins)]
    Y1=n_array
    Yerr=[n_array[k]/sqrt(n_array[k]+0.0) for k in range(nbins)]
    plt.errorbar(X1, Y1, yerr=Yerr, color='red', lw=3, fmt='p')
    chisq = sum([ ((Y0[i]-Y1[i])/Yerr[i])**2.0 for i in range(nbins)  ])
    print "chisq costheta = ", chisq
    
    #theta
    thetahist=fig.add_subplot(143)
    thetahist.hist(thetadata, range=(0,pi/2), bins=nbins)
    x=thetadata; np=len(thetadata)
    #Reference PDF
    X0=linspace(0,pi/2,nbins); Y0=sin(X0)*np*(pi/2)/nbins
    thetahist.plot(X0, Y0, color='black', lw=3)
    #Estimate error bars
    n_array, bins_array=histogram(x,nbins,range=(0,pi/2))
    X1=[(bins_array[k]+bins_array[k+1])/2 for k in range(nbins)]
    Y1=n_array
    Yerr=[n_array[k]/sqrt(n_array[k]+0.0) for k in range(nbins)]
    plt.errorbar(X1, Y1, yerr=Yerr, color='red', lw=3, fmt='p')
    chisq = sum([ ((Y0[i]-Y1[i])/Yerr[i])**2.0 for i in range(nbins)  ])
    print "chisq theta = ", chisq
    
    #phi
    phihist=fig.add_subplot(144)
    phihist.hist(phidata, range=(0,2*pi),  bins=nbins)
    x=phidata; np=len(phidata)
    #Reference PDF
    X0=linspace(0,2*pi,nbins); Y0=X0-X0+np/(nbins+0.0)
    phihist.plot(X0, Y0, color='black', lw=3)
    #Estimate error bars
    n_array, bins_array=histogram(x,nbins,range=(0,2*pi))
    X1=[(bins_array[k]+bins_array[k+1])/2 for k in range(nbins)]
    Y1=n_array
    Yerr=[n_array[k]/sqrt(n_array[k]+0.0) for k in range(nbins)]
    plt.errorbar(X1, Y1, yerr=Yerr, color='red', lw=3, fmt='p')
    X=linspace(0,2*pi,5); Y=X-X+np/(nbins+0.0)
    chisq = sum([ ((Y0[i]-Y1[i])/Yerr[i])**2.0 for i in range(nbins)  ])
    print "chisq phi = ", chisq
    
    plt.show()

# Histogram of a random gaussian field    
def RGSC(ndat):
    rdata=range(ndat)
    thetadata=range(ndat)
    phidata=range(ndat)
    for i in range(ndat):
        rdata[i], thetadata[i], phidata[i] = getSC(random.gauss(0,1), random.gauss(0,1), random.gauss(0,1))
        if thetadata[i] > pi/2:
            thetadata[i]= pi - thetadata[i]
        #if phidata[i] > pi:
        #    phidata[i] = phidata[i] - pi
    return rdata, thetadata, phidata
```
