## Sec 1. Useful Definitions

```python
### 

import matplotlib.pyplot as plt
import numpy as np
import commands
import mpl_toolkits.mplot3d.axes3d as Axes3D
%pylab inline
import os
import random
import scipy.integrate

#################################
# Useful Definiations
#################################

FIG_X_SIZE = 7
FIG_Y_SIZE = 5

CONST_C = 299792.458 #unit: km/s

PLOT_COLOR_ARRAY = ['k', 'b', 'g', 'r', 'c', 'm']
PLOT_STYLE_ARRAY = ['-',  '--', '-.',  'o', 'v', '+', 'x', 's', '*', 'D', ':']

NOW_PLOT_COLOR = 0
NOW_PLOT_STYLE = 0
NUM_PLOT_COLOR = len(PLOT_COLOR_ARRAY)
NUM_PLOT_STYLE = len(PLOT_STYLE_ARRAY)

def get_color_style():
    global NOW_PLOT_COLOR, NOW_PLOT_STYLE
    i = mod(NOW_PLOT_COLOR, NUM_PLOT_COLOR)
    j = mod(NOW_PLOT_COLOR, NUM_PLOT_STYLE)
    color_style = PLOT_COLOR_ARRAY[i]+PLOT_STYLE_ARRAY[j]
    NOW_PLOT_COLOR = NOW_PLOT_COLOR + 1
    NOW_PLOT_STYLE = NOW_PLOT_STYLE + 1
    return color_style

#################################
# get theta, phi according to x,y. 2D
#################################
def getangle(x, y):
    theta = np.arcsin(y/np.sqrt(x**2+y**2))
    if x<0:
        theta = pi-theta
    if x>0 and y<0:
        theta = 2*pi+theta
    return theta
#################################
# get r, theta, phi given x,y,z. 3D
#################################
def getSC(x, y, z):
    r=np.sqrt(x**2+y**2+z**2)
    theta=getangle(z,np.sqrt(x**2+y**2))
    phi=getangle(x,y)
    return r, theta, phi
#################################
# Find Root: f(x) == f0
#################################
def FindRoot( f, f0, xl, xr, tol = 1e-6, maxstep=100 ):
    fl = f(xl); fr = f(xr);
    if abs(fl-f0) < tol:
        return xl, fl-f0, 0, 'S'
    if abs(fr-f0) < tol:
        return xr, fr-f0, 0, 'S'
    if fl > f0 and fr > f0:
        print "Error! f(xl)/f(xr) = ", fl, fr, "both larger than ", f0, "!"
        return 0, 0, 0, 'E' # E means error
    if fl < f0 and fr < f0:
        print "Error! f(xl)/f(xr) = ", fl, fr, "both smaller than ", f0, "!"
        return 0, 0, 0, 'E'
    
    xm = xl+(xr-xl)*(f0-fl)/(fr-fl); fm = f(xm)
    step = 0
    while abs(fm-f0) > tol and step <= maxstep:
        if (fm > f0 and fl > f0) or (fm < f0 and fl < f0):
            xl = xm; fl = fm
        else:
            xr = xm; fr = fm
        xm = xl+(xr-xl)*(f0-fl)/(fr-fl); fm = f(xm)
        step +=1
    return xm, fm-f0, step, 'S' # S means success 

#################################
# Hubble parameter in unit of km/s/Mpc
#################################
def Hz(omegam, w, h, z):
    return 100*h*np.sqrt(omegam*(1.0+z)**3.0 + (1.0-omegam)*(1.0+z)**(3.0*(1+w)))

#################################
# comoving r in unit of Mpc/h
# in fact the result is independt on the input variable h
#################################
def comov_r(omegam, w, h, z):
    global CONST_C
    x, y = scipy.integrate.quad(lambda x: 1.0/Hz(omegam, w, h, x), 0, z)
    return CONST_C * x * h 
    
#################################
# invert of the comov_r function: 
# get z according to comoving_r 
# (in unit of Mpc/h)
#################################
def get_z(omegam, w, h, given_comov_r, zl=0.0, zr=5.0):
    z, deltar, num_step, flag = FindRoot(lambda z: comov_r(omegam, w, h, z), given_comov_r, zl, zr)
    return z
#################################
# unit transition: from km to Mpc/h
#################################
def km_to_Mpctoh(v, h):
    return v / (3.08567758e19) * h

#################################
## Sort a pair_index list
#################################

def sort_pair_index(pair_index):
    
    i = 0
    now_i1 = pair_index[i][0]
    i_begin = 0
    i_array = []
    
    while True:
        while i < len(pair_index) and pair_index[i][0] == now_i1:
            i += 1
        i_array.append([i_begin, i-1])
        
        if i < len(pair_index):
            now_i1 = pair_index[i][0]
            i_begin = i
            continue
        else:
            break
    
    for i in range(len(i_array)):
        i1 = i_array[i][0]
        i2 = i_array[i][1]
        now_segment = [ pair_index[row][1] for row in np.arange(i1, i2+1, 1)]
        
        now_segment.sort()
        for row in range(i1, i2+1):
            pair_index[row][1] = now_segment[row-i1]
            
    return pair_index

#################################
## Comparing two pair index list. 
## Find out the difference.
#################################

first_is_larger  = 1
first_is_smaller = -1
they_are_equal   = 0

def comp_pair(i1,i2, j1,j2):
    if i1==j1 and i2==j2:
        return they_are_equal
    if i1 < j1 or (i1 == j1 and i2 < j2):
        return first_is_smaller
    return first_is_larger

def diff_pair_index(pair_index1, pair_index2):
    pair_diff1 = []
    pair_diff2 = []
    i = 0
    j = 0
    
    while True:
        
        if i == len(pair_index1):
            while j < len(pair_index2):
                #print 'first reach boundary... i, j, append 2 = ', i, j, pair_index2[j]
                pair_diff2.append(pair_index2[j])
                j +=1
            break
        
        if j == len(pair_index2):
            while i < len(pair_index1):
                #print 'second reach boundary... i, j, append 1 = ', i, j, pair_index1[i]
                pair_diff1.append(pair_index1[i])
                i +=1
            break
            
        i1, i2 = pair_index1[i][0], pair_index1[i][1]
        j1, j2 = pair_index2[j][0], pair_index2[j][1]
        
        result = comp_pair(i1,i2, j1,j2)
        
        if result == they_are_equal:
            i +=1
            j +=1
            continue
        
        if result == first_is_smaller:
            #print 'fist_is_smaller: add', pair_index1[i], 'into diff1'
            pair_diff1.append(pair_index1[i])
            i +=1
            continue
            
        if result == first_is_larger:
            #print 'fist_is_larger: add', pair_index2[j], 'into diff2'
            pair_diff2.append(pair_index2[j])
            j +=1
            continue
    
    return pair_diff1, pair_diff2
```

## Sec 3. RSD & AP Effect on Single Halos

```python
##################################################################
# This defines the full information of a single halo,
#  including redshift and postions w/wo RSD or AP. 
# Position of the observer is x,y,z=0,0,0. 
# All lengths in unit Mpc/h; all velocities in unit km/s
# The structure is named as "point" rather than "halo", 
#  since it can be anything (e.g., particles, galaxies, ...)
##################################################################
class point:
    x =0.0;  y =0.0; z =0.0; theta=0.0; phi=0.0; unit_vector = [0.0,0.0,0.0]  
    vx=0.0;  vy=0.0; vz=0.0; vlos=0.0
    # resdhift and distance w/wo RSD
    z_real = 0.0;  z_obs = 0.0;  z_obs_diff = 0.0;  z_obs_discrep = 0.0
    r =0.0;        r_RSD = 0.0;  r_RSD_diff = 0.0;  r_RSD_discrep = 0.0;  xyz_RSD = [0,0,0]
    # cosmologicay information
    Hubble = 0.0
    comov_dist = 0.0
    D_A = 0.0
    # assumed cosmologies
    fidu_cosmo = [[0.0,0.0,0.0]]
    num_fidu_cosmo = 1
    # distances and postions with AP
    r_AP = [0.0]
    r_AP_diff = [0.0]
    r_AP_discrep = [0.0]
    xyz_AP = [[0,0,0]]
    r_AP_RSD = [0.0]
    r_AP_RSD_diff = [0.0]
    r_AP_RSD_discrep = [0.0]
    xyz_AP_RSD = [[0,0,0]]
    # initialize     
    def __init__(self, x,y,z, vx,vy,vz, fidu_cosmo = [[om_dft,w_dft,h_dft]] ): 
        global CONST_C
        # basic information
        self.x, self.y, self.z = x,y,z
        self.vx, self.vy, self.vz = vx,vy,vz
        self.r, self.theta, self.phi = getSC(self.x,self.y,self.z)
        self.unit_vector = [ self.x/self.r, self.y/self.r, self.z/self.r ]
        self.vlos = np.dot(self.unit_vector, [self.vx, self.vy, self.vz]) #los velocity in comoving space        
        # cosmological information
        self.z_real = get_z(om_dft, w_dft, h_dft, self.r)
        self.Hubble = Hz(om_dft, w_dft, h_dft, self.z_real)
        self.D_A = self.r / (1.0 + self.z_real)
        # observed redshift (considering RSD)
        self.z_obs = self.z_real + self.vlos / CONST_C  # observed, apparent redshift 
        self.z_obs_diff = self.z_obs - self.z_real 
        self.z_obs_discrep = self.z_obs_diff / self.z_real
        # Apparent Distances & Positions: RSD considered
        self.r_RSD = comov_r(om_dft, w_dft, h_dft, self.z_obs)
        self.r_RSD_diff = self.r_RSD - self.r
        self.r_RSD_discrep = self.r_RSD_diff / self.r
        ratio = self.r_RSD / self.r
        self.xyz_RSD = [ ratio*self.x, ratio*self.y, ratio*self.z ] # apparent x,y,z
        # Assumed Cosmologies
        self.fidu_cosmo = fidu_cosmo
        self.num_fidu_cosmo = len(fidu_cosmo)
        # Apparent Distances & Positions: AP & AP+RSD considered
        self.r_AP = range(self.num_fidu_cosmo)
        self.r_AP_RSD = range(self.num_fidu_cosmo)
        self.r_AP_diff = range(self.num_fidu_cosmo)
        self.r_AP_RSD_diff = range(self.num_fidu_cosmo)
        self.r_AP_discrep = range(self.num_fidu_cosmo)
        self.r_AP_RSD_discrep = range(self.num_fidu_cosmo)
        self.xyz_AP = [ [0.0, 0.0, 0.0] for row in range(self.num_fidu_cosmo) ]
        self.xyz_AP_RSD = [ [0.0, 0.0, 0.0] for row in range(self.num_fidu_cosmo) ]
        # Distance and postions with AP:
        for i in range(self.num_fidu_cosmo):
            om_fd, w_fd, h_fd = self.fidu_cosmo[i][0], self.fidu_cosmo[i][1], self.fidu_cosmo[i][2]
            # wo RSD
            self.r_AP[i] = comov_r(om_fd, w_fd, h_fd, self.z_real)
            self.r_AP_diff[i] = self.r_AP[i] - self.r
            self.r_AP_discrep[i] = (self.r_AP[i] - self.r) / self.r
            ratio = self.r_AP[i] / self.r
            self.xyz_AP[i] = [ ratio*self.x, ratio*self.y, ratio*self.z ]
            # w RSD
            self.r_AP_RSD[i] = comov_r(om_fd, w_fd, h_fd, self.z_obs)
            self.r_AP_RSD_diff[i] = self.r_AP_RSD[i] - self.r
            self.r_AP_RSD_discrep[i] = (self.r_AP_RSD[i] - self.r) / self.r
            ratio = self.r_AP_RSD[i] / self.r
            self.xyz_AP_RSD[i] = [ ratio*self.x, ratio*self.y, ratio*self.z ]
    # print out information    
    def speak(self):
        print '  Information of a point object (distances Mpc/h, velocities km/s)\n'
        print '     x, y, z                         =   ', self.x,self.y,self.z
        print '     comoving r                      =   ', self.r
        print '     theta, phi                      =   ', self.theta,self.phi
        print '     vx, vy, vz                      =   ', self.vx, self.vy, self.vz
        print '     line of sight speed             =   ', self.vlos
        print '     cosmological redshift z         =   ', self.z_real
        print '     Hubble (km/s/Mpc)               =   ', self.Hubble
        print '     D_A (Mpc/h)                     =   ', self.D_A
        print '\n  Observed quantities (considering RSD; discrep = (obs - real) / real)\n'
        print '     apparent redshift z (discrep)   =   ', self.z_obs, '  (', self.z_obs_discrep, ')'
        print '     apparent comoving r (discrep)   =   ', self.r_RSD, '  (', self.r_RSD_discrep, ')'
        for i in range(self.num_fidu_cosmo):
            print '\n  Inferred r  (assuming (om, w, h) =', self.fidu_cosmo[i][0], \
            self.fidu_cosmo[i][1], self.fidu_cosmo[i][2], '; discrep = (obs - real) / real) \n'
            print '     inferred comov r based on  cosmological  z (discrep)   =   ', \
            self.r_AP[i], '  (', self.r_AP_discrep[i] , ')'
            print '     inferred comov r based on observed RSDed z (discrep)   =   ', \
            self.r_AP_RSD[i], '  (', self.r_AP_RSD_discrep[i] , ')'

##################################################################
# Getting the full information of all halos. 
# May take ~1-2 minutes
##################################################################

halo_info = [ point( halo_data[row][0], halo_data[row][1], halo_data[row][2], \
                     halo_data[row][3], halo_data[row][4], halo_data[row][5], assumed_cosmo  )\
              for row in range(num_halo)]

##################################################################
# Minimal/maximal z, r, phi. 
#  shall be useful since we select a slice based on z/r/phi and plot
##################################################################

z_real_min = min(halo_info[row].z_real for row in range(num_halo))
z_real_max = max(halo_info[row].z_real for row in range(num_halo))
r_real_min = min(halo_info[row].r      for row in range(num_halo))
r_real_max = max(halo_info[row].r      for row in range(num_halo))
phi_min    = min(halo_info[row].phi    for row in range(num_halo))
phi_max    = max(halo_info[row].phi    for row in range(num_halo))
print "Information of the halo data:\n"
print "\tminimal redshift = ",z_real_min
print "\tmaximal redshift = ", z_real_max
print "\tminimal comoving r = ", r_real_min
print "\tmaximal comoving r = ", r_real_max
print "\tminimal phi        = ", phi_min
print "\tmaximal phi        = ", phi_max
```

## Sec 4. Calculating BSkeletons

```python
########################################################################

### Useful Definiations.

# Name of the file saving a beta-skeleton
def BSke_file_name(RSD, AP, beta):
    if AP=='F' and RSD == 'F':
        return output_dir+halo_data_name+'__beta='+str(beta)
    if AP=='F' and RSD == 'T':
        return output_dir+halo_data_name+'__RSD'+'__beta='+str(beta)
    if AP >= 0:
        cosmo_name = assumed_cosmo_name[AP]
    else:
        cosmo_name = 'assu_cosmo'
    if RSD == 'F':
        return output_dir+halo_data_name+'__'+cosmo_name+'__beta='+str(beta)
    if RSD == 'T':
        return output_dir+halo_data_name+'__'+cosmo_name+'_RSD'+'__beta='+str(beta)

# Get the scatter data based on RSD, AP  
def get_scat_data(RSD, AP):
    if AP=='F' and RSD == 'F':
        return [ [halo_info[row].x, halo_info[row].y, halo_info[row].z] for row in range(num_halo) ]
    if AP=='F' and RSD == 'T':
        return [ halo_info[row].xyz_RSD for row in range(num_halo) ]
    if AP!='F' and RSD == 'F':
        return [ halo_info[row].xyz_AP[AP] for row in range(num_halo) ]
    if AP!='F' and RSD == 'T':
        return [ halo_info[row].xyz_AP_RSD[AP] for row in range(num_halo) ]    

# Get the pair index data based on RSD, AP, beta.  Shall load a file.
def get_pair_index(RSD, AP, beta):
    pair_index = np.loadtxt(BSke_file_name(RSD, AP, beta))
    pair_index = [ [int(pair_index[row,0]+0.001), int(pair_index[row,1]+0.001)] for row in range(len(pair_index)) ]
    return pair_index

# Get the pair coordinate data based on RSD, AP, beta.
def get_pair_data(RSD, AP, beta):
    pair_index = get_pair_index(RSD, AP, beta)
    scat_data  = get_scat_data(RSD, AP)
    pair_data = [ [ scat_data[pair_index[row][0]], scat_data[pair_index[row][1]] ] for row in range(len(pair_index)) ]
    return pair_data

# Get the pair coordinate data, but based on given index list.
def get_pair_data_from_index(pair_index, RSD, AP, beta):
    scat_data  = get_scat_data(RSD, AP)
    pair_data = [ [ scat_data[pair_index[row][0]], scat_data[pair_index[row][1]] ] for row in range(len(pair_index)) ]
    return pair_data

########################################################################

# processing the pairs: remove redudant pairs in BSKe_file; sort the array.
def clean_pair(BSke_file):
    print '  Cleaning file ', BSke_file, '...\n'
    A=np.loadtxt(BSke_file)
    B=[]; i=0; j=0; 
    while i<len(A):
        i1 = A[i,0]; i2 = A[i,1]
        if i1 < i2: # the pairs are double-counted by ngl-beta; only keep half of them
            B.append([int(i1+0.001), int(i2+0.001)])
        i = i+1
    sort_pair_index(B)
    i=0
    INDEX_FILE = open(BSke_file, 'w')
    while i<len(B):
        i1 = B[i][0]; i2 = B[i][1]
        INDEX_FILE.write(str(i1)+'   '+str(i2)+"\n")
        i = i+1
    INDEX_FILE.close()

# Calculating BSKe by calling ngl, and clean the file using clear_pair()
def Calc_BSke(scat_data, BSke_file, beta):
        np.savetxt('tmp.txt', scat_data)
        print '  Generating neighbour graph for beta = ', beta, '...'
        command='rm '+BSke_file+';  '\
                   +'../ngl-beta/build/binsrc/getNeighborGraph -i tmp.txt -d 3 -b'+' '+str(beta)+' >> '+BSke_file
        print command
        print commands.getoutput(command)
        clean_pair(BSke_file)
        os.system('rm tmp.txt')
        return   

# Calculating all possible BSke, based on a list of given beta.
def Calc_all_BSke(beta_array):
        for RSD in ['F', 'T']:
            AP_range = range(len(assumed_cosmo))
            AP_range.insert(0, 'F')
            for AP in AP_range:
                scat_data = get_scat_data(RSD, AP)
                for beta in beta_array:
                    BSke_file = BSke_file_name(RSD, AP, beta)
                    print 'Calculating BSke: ', BSke_file, '...'
                    Calc_BSke(scat_data, BSke_file, beta)
```

### Sec 5.3 Directions of Piars: Angular Statistics (mu = cos theta)

```python
##########################################
# Useful functions for getting statistical
#  quantities: mu & theta & los/perp ratio;
#  Estimating the mean & variance of the 
#  sample.  
##########################################

def get_mu_of_pair(coord1, coord2):
    x1, y1, z1 = coord1[0:3]
    x2, y2, z2 = coord2[0:3]
    r = [(x1+x2)/2.0, (y1+y2)/2.0, (z1+z2)/2.0]
    len_r = np.sqrt(r[0]*r[0] + r[1]*r[1] + r[2]*r[2])
    dr = [x2-x1, y2-y1, z2-z1]
    len_dr = np.sqrt(dr[0]*dr[0] + dr[1]*dr[1] + dr[2]*dr[2])
    return abs(np.dot(r, dr)/len_r/len_dr)

# Get mu, theta, los/perp distance of pair
def get_info_of_pair(coord1, coord2):
    x1, y1, z1 = coord1[0:3]
    x2, y2, z2 = coord2[0:3]
    r = [(x1+x2)/2.0, (y1+y2)/2.0, (z1+z2)/2.0]
    len_r = np.sqrt(r[0]*r[0] + r[1]*r[1] + r[2]*r[2])
    dr = [x2-x1, y2-y1, z2-z1]
    len_dr = np.sqrt(dr[0]*dr[0] + dr[1]*dr[1] + dr[2]*dr[2])
    mu = abs(np.dot(r, dr)/len_r/len_dr)
    theta = np.arccos(mu)
    losdr  = len_dr * mu
    perpdr = len_dr * np.sin(theta)
    return mu, theta, losdr, perpdr

def get_mulist_from_pairlist(pair_data):
    return [ get_mu_of_pair(pair_data[row][0], pair_data[row][1]) for row in range(len(pair_data)) ]

def get_selected_mulist_from_pairlist(pair_data, delta_r_min, delta_r_max):
    mu_data = []
    for row in range(len(pair_data)/2):
        now_pair = pair(pair_data[row][0], pair_data[row][1])
        if now_pair.delta_r < delta_r_max and now_pair.delta_r > delta_r_min:
            mu_data.append(now_pair.mu)
    return mu_data

# Get the mean & variance, with errors
def get_stat_from_list(x):
    n = len(x)
    x_av    = sum(x)/len(x)
    x_var   = (sum(x[row]**2.0 for row in range(n)) - n * x_av**2.0) /(n-1.0) #correct the bias using n-1
    x_av_er = np.sqrt(x_var) / np.sqrt(n-1.0) # correct the bias using n-1 
    x_var_er = np.sqrt(2.0/(n-1.0)) * x_var
    return x_av, x_var, x_av_er, x_var_er

# Getting mu list from a given skeleton
def get_mu_data(RSD, AP, beta, min_delta_r=-1, max_delta_r=-1, given_pair_index = False):  
    # Calculating mu based on the BSke related to RSD, AP, beta
    if given_pair_index == False:
        if min_delta_r == -1 and max_delta_r==-1:
            mu_data = get_mulist_from_pairlist(get_pair_data(RSD,AP,beta))    
        else:
            mu_data = get_selected_mulist_from_pairlist(get_pair_data(RSD,AP,beta), \
                                             min_delta_r, max_delta_r)
        return mu_data 
    # Calculating mu based on given set of samples
    if given_pair_index != False:
        scat_data = get_scat_data(RSD, AP)
        pair_list = [ [scat_data[given_pair_index[row][0]], \
                       scat_data[given_pair_index[row][1]]] for row in range(len(given_pair_index))  ]
        if min_delta_r == -1 and max_delta_r==-1:
            mu_data = get_mulist_from_pairlist(pair_list)    
        else:
            mu_data = get_selected_mulist_from_pairlist(pair_list, \
                                             min_delta_r, max_delta_r)
        return mu_data

########################################################################
# This defines the class pair.
#  Information includes: positions; seperations;
#  distances; theta = los angles; mu = cos theat;...
########################################################################
class pair:
    # postions
    x1, y1, z1 = 0, 0, 0
    x2, y2, z2 = 0, 0, 0
    r1, r2 = 0, 0
    # seperation
    delta_r = 0
    delta_vector = [0,0,0]
    unit_delta_vector = [0,0,0]
    # middle point
    middle_point_r = 0
    middle_point_vector = [0,0,0]
    unit_middle_point_vector = [0, 0, 0]
    # los theta, cos theta, los & perp seperation
    mu = 0
    los_theta = 0
    losdr  = 0
    perpdr = 0
    # initialize
    def __init__(self, coord1, coord2):
        x1,y1,z1 = coord1[0], coord1[1], coord1[2]
        x2,y2,z2 = coord2[0], coord2[1], coord2[2]
        r1 = np.sqrt(x1**2+y1**2+z1**2)
        r2 = np.sqrt(x2**2+y2**2+z2**2)
        # always keep x1,y1,z1 the nearby one.
        if r1 < r2: 
            self.x1, self.y1, self.z1 = x1,y1,z1
            self.x2, self.y2, self.z2 = x2,y2,z2
            self.r1 = r1; self.r2 = r2
        else:
            self.x1, self.y1, self.z1 = x2,y2,z2
            self.x2, self.y2, self.z2 = x1,y1,z1
            self.r1 = r2; self.r2 = r1
        # seperation
        self.delta_vector = [ self.x2-self.x1, self.y2-self.y1, self.z2-self.z1 ]
        self.delta_r = np.sqrt(sum(self.delta_vector[row]**2.0 for row in range(3)))
        self.unit_delta_vector = [ self.delta_vector[row]/self.delta_r for row in range(3)]
        # middle point
        self.middle_point_vector = [ (self.x2+self.x1)/2.0, (self.y2+self.y1)/2.0, (self.z2+self.z1)/2.0 ]
        self.middle_point_r = np.sqrt(sum(self.middle_point_vector[row]**2.0 for row in range(3)))
        self.unit_middle_point_vector = [ self.middle_point_vector[row]/self.middle_point_r for row in range(3)]
        # mu, los_theta, los_dr, perp_dr
        self.mu = np.dot(self.unit_middle_point_vector, self.unit_delta_vector)
        self.los_theta = np.arccos(self.mu)
        self.losdr  = self.delta_r * self.mu
        self.perpdr = self.delta_r * np.sin(self.los_theta)
    # print out the information
    def speak(self):
        print 'Information for the pair:\n'
        print '  Nearby  point: ', (self.x1, self.y1, self.z1) 
        print '  Farther point: ', (self.x2, self.y2, self.z2) 
        print '  Distance of Nearby // Farther point   = ', self.r1, '//', self.r2
        print '  Distance between two points           = ', self.delta_r
        print '  Separation vector                     = ', self.delta_vector
        print '  Middle point                          = ', self.middle_point_vector
        print '  Distance of middle point              = ', self.middle_point_r
        print '  Angle between los/pair                = ', self.los_theta
        print '  Cos of angle between los/pair         = ', self.mu
        print '  Line-of-sight distance                = ', self.losdr
        print '  Perpendicular distance                = ', self.perpdr
### Code for Test
#pair_data = get_pair_data('F', 'F', 1.0);i=1;
#pair0 = pair(pair_data[i][0], pair_data[i][1]);pair0.speak()
########################################################################
# This defines the class beta-skeleton.
#  Information includes: basic information such as beta, RSD,
#  AP, File, as well as statistical information.
########################################################################    
class BSke:
    ### Basic information
    beta = 0
    RSD  = 'F'
    AP   = 'F'
    File = ''
    pair_index = []
    num_pair = 0 
    ### Statistic information
    have_statis_result = False
    # mu = cos(theta)
    mu_av     = 0;    mu_var    = 0;    mu_av_er  = 0;    mu_var_er = 0
    # theta and sin(theta)^2
    theta_av      = 0;    theta_var     = 0;    theta_av_er   = 0;    theta_var_er  = 0
    sinsq_av      = 0;    sinsq_var     = 0;    sinsq_av_er   = 0;    sinsq_var_er  = 0
    # los distance
    losdr_av     = 0;    losdr_var    = 0;    losdr_av_er  = 0;    losdr_var_er = 0
    # perp distance
    perpdr_av     = 0;    perpdr_var    = 0;    perpdr_av_er  = 0;    perpdr_var_er = 0
    # los/perp ratio 
    lostoperp_av     = 0;    lostoperp_var    = 0;    lostoperp_av_er  = 0;    lostoperp_var_er = 0
    ### initialize
    def __init__(self, RSD, AP, beta, calc_stat = False):
        self.RSD = RSD; self.AP = AP; self.beta=beta
        self.File = BSke_file_name(RSD, AP, beta)
        e = os.path.exists(self.File)
        if e == False:
            print 'BSke_file not exist. Now calculate the BSkeleton for RSD, AP, beta =', RSD, AP, beta
            scat_data = get_scat_data(RSD, AP)
            Calc_BSke(scat_data, self.File, beta)
        self.pair_index = sort_pair_index(get_pair_index(RSD, AP, beta))
        self.num_pair = len(self.pair_index)
        if calc_stat:
            self.do_statistic()
    ### calculate statistical quantities    
    def do_statistic(self, index_data = 0):
        if index_data == 0:
            index_data = self.pair_index
        pair_data = get_pair_data_from_index(index_data, self.RSD, self.AP, self.beta)
        num_pair = len(index_data)
        mu_list = range(num_pair);      theta_list = range(num_pair); 
        losdr_list = range(num_pair);  perpdr_list = range(num_pair);
        sinsq_list = range(num_pair);   lostoperp_list = range(num_pair);
        for i in range(num_pair):
            mu_list[i], theta_list[i], losdr_list[i], perpdr_list[i] = \
              get_info_of_pair(pair_data[i][0], pair_data[i][1])
            sinsq_list[i] = np.sin(theta_list[i])**2.0
            lostoperp_list[i] = losdr_list[i] / perpdr_list[i]        
        self.mu_av, self.mu_var, self.mu_av_er, self.mu_var_er = \
            get_stat_from_list(mu_list)
        self.theta_av, self.theta_var, self.theta_av_er, self.theta_var_er = \
            get_stat_from_list(theta_list)
        self.sinsq_av, self.sinsq_var, self.sinsq_av_er, self.sinsq_var_er = \
            get_stat_from_list(sinsq_list)
        self.losdr_av, self.losdr_var, self.losdr_av_er, self.losdr_var_er = \
            get_stat_from_list(losdr_list)
        self.perpdr_av, self.perpdr_var, self.perpdr_av_er, self.perpdr_var_er = \
            get_stat_from_list(perpdr_list)
        self.lostoperp_av, self.lostoperp_var, self.lostoperp_av_er, \
            self.lostoperp_var_er = get_stat_from_list(lostoperp_list)
        self.have_statis_result = True
    ### print the information
    def speak(self):
        print 'Information of the Bske:'
        print '  RSD, AP, beta      = ', (self.RSD, self.AP, self.beta)
        if self.AP != 'F':
            print '    Assuming cosmology', assumed_cosmo_name[self.AP],'; parameters = ', assumed_cosmo[self.AP]
        print '  File               = ', self.File
        print '  # of pairs         = ', self.num_pair
        if self.have_statis_result == False:
            print '  No stiatistic result. Return now.'
            return
        print 'Statistical information (may have signif error if highly non-gaussian): '
        print '  <mu>               =  %10.4f'%self.mu_av, ' +/-   %.4f'%self.mu_av_er
        print '  Var(mu)            =  %10.4f'%self.mu_var, ' +/-   %.4f'%self.mu_var_er
        x1=self.theta_av/np.pi*180.0; x2=self.theta_av_er/np.pi*180.0
        x3=self.theta_var/np.pi*180; x4=self.theta_var_er/np.pi*180
        print '  <theta>            =  %10.4f'%x1, ' +/-   %.4f'%x2
        print '  Var(theta)         =  %10.4f'%x3, ' +/-   %.4f'%x4
        print '  <(sin theta)^2>    =  %10.4f'%self.sinsq_av, ' +/-   %.4f'%self.sinsq_av_er
        print '  Var((sin theta)^2) =  %10.4f'%self.sinsq_var, ' +/-   %.4f'%self.sinsq_var_er
        print '  <los r>            =  %10.4f'%self.losdr_av, ' +/-   %.4f'%self.losdr_av_er
        print '  Var(los r)         =  %10.4f'%self.losdr_var, ' +/-   %.4f'%self.losdr_var_er
        print '  <perp r>           =  %10.4f'%self.perpdr_av, ' +/-   %.4f'%self.perpdr_av_er
        print '  Var(perp r)        =  %10.4f'%self.perpdr_var, ' +/-   %.4f'%self.perpdr_var_er
        print '  <los/perp>         =  %10.4f'%self.lostoperp_av, ' +/-   %.4f'%self.lostoperp_av_er
        print '  Var(los/perp)      =  %10.4f'%self.lostoperp_var, ' +/-   %.4f'%self.lostoperp_var_er    

########################################################################
# This defines the class beta-skeleton.
#  Information includes: basic information such as beta, RSD,
#   AP, File, as well as statistical information.
#   ...
########################################################################    
class BSke_list:  
    # list of beta
    beta_array = []
    # list of Bske
    Bske_name_list = []
    Bske_list = []
    # list of # of pairs
    num_pair_list =[]
    # list of # and ratio of removed/added pairs
    num_removed_pair_list = []
    num_added_pair_list = []
    ratio_removed_pair_list = []
    ratio_added_pair_list = []
    # list of list of removed pairs
    removed_pair_index_list = []
    # list of list of added pairs
    added_pair_index_list = []
    # list of Holdon Bske
    Holdon_Bske_list = []
    # list of statistical results
    mu_list = []
    theta_list = []
    sinsq_list = []
    losdr_list = []
    perpdr_list = []
    lostoperp_list = []
    Holdon_mu_list = []
    Holdon_theta_list = []
    Holdon_sinsq_list = []
    Holdon_losdr_list = []
    Holdon_perpdr_list = []
    Holdon_lostoperp_list = []
    # initialize
    def __init__(self, beta_array):
        self.self_clear()
        self.beta_array = beta_array
        print 'beta_array = ', beta_array,  '.'
        print '# of assumed cosmology = ', len(assumed_cosmo), '.'
        print '# of     Bskes    for each beta = ', 2*(1+len(assumed_cosmo)), '.'
        print '# of Holdon Bskes for each beta = ', 2*(1+len(assumed_cosmo)), '.'
        for beta in beta_array:
            print '     dealing with Bskes with beta = ', beta, '...'
            for RSD in ['F', 'T']:
                AP_range = range(len(assumed_cosmo))
                AP_range.insert(0, 'F')
                for AP in AP_range:
                    ### Bske
                    self.Bske_name_list.append([RSD, AP, beta])
                    Bske = BSke(RSD, AP, beta, True)
                    self.Bske_list.append(Bske)
                    # pair index
                    now_pair_index = Bske.pair_index
                    # # of pairs
                    num_pair = len(Bske.pair_index)
                    self.num_pair_list.append(num_pair)
                    # Real cosmology results as reference
                    if AP == 'F':
                        Holdon_pair_index = now_pair_index
                        num_holdon_pair = num_pair
                        mu, muer, theta, thetaer, sinsq, sinsqer, losdr, losdrer, perpdr, perpdrer, \
                            lostoperp, lostoperper = \
                            Bske.mu_av, Bske.mu_av_er, Bske.theta_av, Bske.theta_av_er,\
                            Bske.sinsq_av, Bske.sinsq_av_er, Bske.losdr_av, Bske.losdr_av_er,  \
                            Bske.perpdr_av, Bske.perpdr_av_er, Bske.lostoperp_av, Bske.lostoperp_av_er
                        # no changes compared with itself.
                        self.removed_pair_index_list.append([])
                        self.added_pair_index_list.append([])
                        self.num_removed_pair_list.append(0)
                        self.num_added_pair_list.append(0)
                        self.ratio_removed_pair_list.append(0)
                        self.ratio_added_pair_list.append(0)
                    else:
                        diff1, diff2 = diff_pair_index(Holdon_pair_index, now_pair_index)
                        self.removed_pair_index_list.append(diff1)
                        self.added_pair_index_list.append(diff2)
                        self.num_removed_pair_list.append(len(diff1))
                        self.num_added_pair_list.append(len(diff2))
                        self.ratio_removed_pair_list.append(len(diff1)/(num_holdon_pair+0.0))
                        self.ratio_added_pair_list.append(len(diff2)/(num_holdon_pair+0.0))
                    # Statisitcal quanitites                       
                    self.mu_list.append([Bske.mu_av, Bske.mu_av_er, Bske.mu_var, Bske.mu_var_er, \
                                         abs(Bske.mu_av-mu)/muer])
                    self.theta_list.append([Bske.theta_av, Bske.theta_av_er, Bske.theta_var, Bske.theta_var_er, \
                                            abs(Bske.theta_av-theta)/thetaer])
                    self.sinsq_list.append([Bske.sinsq_av, Bske.sinsq_av_er, Bske.sinsq_var, Bske.sinsq_var_er, \
                                            abs(Bske.sinsq_av-sinsq)/sinsqer])
                    self.losdr_list.append([Bske.losdr_av, Bske.losdr_av_er, Bske.losdr_var, Bske.losdr_var_er, \
                                            abs(Bske.losdr_av-losdr)/losdrer])
                    self.perpdr_list.append([Bske.perpdr_av, Bske.perpdr_av_er, \
                                              Bske.perpdr_var, Bske.perpdr_var_er, \
                                              abs(Bske.perpdr_av-perpdr)/perpdrer])
                    self.lostoperp_list.append([Bske.lostoperp_av, Bske.lostoperp_av_er, \
                                              Bske.lostoperp_var, Bske.lostoperp_var_er, \
                                              abs(Bske.lostoperp_av-lostoperp)/lostoperper])
                    
                    ### 'Hold on' Bske
                    Bske = BSke(RSD, AP, beta, False)
                    Bske.do_statistic(Holdon_pair_index)
                    self.Holdon_Bske_list.append(Bske)
                    # Real cosmology results as reference
                    if AP == 'F':
                        mu, muer, theta, thetaer, sinsq, sinsqer, losdr, losdrer, perpdr, perpdrer, \
                            lostoperp, lostoperper = \
                            Bske.mu_av, Bske.mu_av_er, Bske.theta_av, Bske.theta_av_er,\
                            Bske.sinsq_av, Bske.sinsq_av_er, Bske.losdr_av, Bske.losdr_av_er,  \
                            Bske.perpdr_av, Bske.perpdr_av_er, Bske.lostoperp_av, Bske.lostoperp_av_er
                    # Statisitcal quanitites                       
                    self.Holdon_mu_list.append([Bske.mu_av, Bske.mu_av_er, Bske.mu_var, Bske.mu_var_er, \
                                         abs(Bske.mu_av-mu)/muer])
                    self.Holdon_theta_list.append([Bske.theta_av, Bske.theta_av_er, Bske.theta_var, \
                                                   Bske.theta_var_er, abs(Bske.theta_av-theta)/thetaer])
                    self.Holdon_sinsq_list.append([Bske.sinsq_av, Bske.sinsq_av_er, Bske.sinsq_var, \
                                                   Bske.sinsq_var_er, abs(Bske.sinsq_av-sinsq)/sinsqer])
                    self.Holdon_losdr_list.append([Bske.losdr_av, Bske.losdr_av_er, Bske.losdr_var, \
                                                   Bske.losdr_var_er, abs(Bske.losdr_av-losdr)/losdrer])
                    self.Holdon_perpdr_list.append([Bske.perpdr_av, Bske.perpdr_av_er, \
                                              Bske.perpdr_var, Bske.perpdr_var_er, \
                                              abs(Bske.perpdr_av-perpdr)/perpdrer])
                    self.Holdon_lostoperp_list.append([Bske.lostoperp_av, Bske.lostoperp_av_er, \
                                              Bske.lostoperp_var, Bske.lostoperp_var_er, \
                                              abs(Bske.lostoperp_av-lostoperp)/lostoperper])
    def self_clear(self):
        self.beta_array = []
        self.Bske_name_list = []
        self.Bske_list = []
        self.num_pair_list =[]
        self.num_removed_pair_list = []
        self.num_added_pair_list = []
        self.ratio_removed_pair_list = []
        self.ratio_added_pair_list = []
        self.removed_pair_index_list = []
        self.added_pair_index_list = []
        self.Holdon_Bske_list = []
        self.mu_list = []
        self.theta_list = []
        self.sinsq_list = []
        self.losdr_list = []
        self.perpdr_list = []
        self.lostoperp_list = []
        self.Holdon_mu_list = []
        self.Holdon_theta_list = []
        self.Holdon_sinsq_list = []
        self.Holdon_losdr_list = []
        self.Holdon_perpdr_list = []
        self.Holdon_lostoperp_list = []

    def plot_num_pair_info(self, num_column = 2):
        global FIG_X_SIZE, FIG_Y_SIZE
        # setting figure structure.
        num_figure = 2 + 2*2  ## num of pair(2) + num/ratio of removed/added pair (2*2)
        num_row =  num_figure /num_column + min(1,np.mod(num_figure,num_column))
        single_plot_x_size = FIG_X_SIZE * 2.0 / (num_column+0.0)
        single_plot_y_size = single_plot_x_size * FIG_Y_SIZE / FIG_X_SIZE
        print 'num_row, num_column = ', num_row, num_column
        fig=plt.figure(figsize=(single_plot_x_size*num_column,single_plot_y_size*num_row))
        fig.subplots_adjust(hspace=0.6)
        # plottings of # pairs 
        i_fig = 0
        for RSD in ['F', 'T']:
            legend_array = []
            subfig = fig.add_subplot(num_row, num_column, i_fig+1, xlabel = 'beta', ylabel = 'num of pairs',\
                                     title = 'Number of Pairs: RSD = '+RSD)
            AP_range = range(len(assumed_cosmo))
            AP_range.insert(0, 'F')
            for AP in AP_range:
                if AP == 'F':
                    legend_array.append('No AP')
                else:
                    legend_array.append(assumed_cosmo_name[AP])
                X = []; Y = [];
                i = self.Bske_name_list.index([RSD, AP, beta_array[0]])
                for beta in self.beta_array:
                    X.append(beta); Y.append(self.num_pair_list[i])
                    i += 2*(num_assumed_cosmo+1)
                subfig.plot(X, Y, get_color_style())
            leg=subfig.legend(legend_array, 'upper right', shadow=False, fancybox=True);
            leg.get_frame().set_alpha(0.5)
            i_fig += 1
        # plottings of # removed / added pairs
        for RSD in ['F', 'T']:
            legend_array1 = []
            legend_array2 = []
            subfig1 = fig.add_subplot(num_row, num_column, i_fig+1, xlabel = 'beta', \
                                     ylabel = 'num of removed/added pairs',\
                                     title = 'Number of Removed/Added Pairs: RSD = '+RSD)
            subfig2 = fig.add_subplot(num_row, num_column, i_fig+2, xlabel = 'beta', \
                                     ylabel = 'ratio of removed/added pairs',\
                                     title = 'Ratio of Removed/Added Pairs: RSD = '+RSD)
            AP_range = range(len(assumed_cosmo))
            for AP in AP_range:
                X = []; Y1 = []; Y2 = []; Y3 = []; Y4 = []; 
                i = self.Bske_name_list.index([RSD, AP, beta_array[0]])
                for beta in self.beta_array:
                    X.append(beta); 
                    Y1.append(self.num_removed_pair_list[i]); 
                    Y2.append(self.num_added_pair_list[i])
                    Y3.append(self.ratio_removed_pair_list[i]); 
                    Y4.append(self.ratio_added_pair_list[i])
                    i += 2*(num_assumed_cosmo+1)
                subfig1.plot(X, Y1, get_color_style())
                legend_array1.append('# of removed pairs in '+assumed_cosmo_name[AP])
                subfig1.plot(X, Y2, get_color_style())
                legend_array1.append('# of  added  pairs in '+assumed_cosmo_name[AP])
                subfig2.plot(X, Y3, get_color_style())
                legend_array2.append('Ratio of removed pairs in '+assumed_cosmo_name[AP])
                subfig2.plot(X, Y4, get_color_style())
                legend_array2.append('Ratio of  added  pairs in '+assumed_cosmo_name[AP])
            leg=subfig1.legend(legend_array1, 'upper right', shadow=False, fancybox=True);
            leg.get_frame().set_alpha(0.5)
            leg=subfig2.legend(legend_array2, 'upper right', shadow=False, fancybox=True);
            leg.get_frame().set_alpha(0.5)
            i_fig += 2
    
    def print_skeleton_change_info(self, beta):
        i0 = self.Bske_name_list.index(['F', 'F', beta])
        i1 = i0 + len(assumed_cosmo) + 1
        print '  If no RSD,    assumed cosmo .vs. real cosmo:'
        for i in range(num_assumed_cosmo):
            print '\tAssumed cosmology ',i,': # of Removed/Added pairs are  ', \
                self.num_removed_pair_list[i0+i+1], self.num_added_pair_list[i0+i+1]
            print '\tRatios =  '+'%.4f'%self.ratio_removed_pair_list[i0+i+1]+\
                '  %.4f'%self.ratio_added_pair_list[i0+i+1]
        print '\n  If with RSD,  assumed cosmo .vs. real cosmo:'
        for i in range(num_assumed_cosmo):
            print '\tAssumed cosmology ',i,': # of Removed/Added pairs are  ', \
                self.num_removed_pair_list[i1+i+1], self.num_added_pair_list[i1+i+1]
            print '\tRatios =  '+'%.4f'%self.ratio_removed_pair_list[i1+i+1]+\
                '  %.4f'%self.ratio_added_pair_list[i1+i+1]
    
    def list_print_skeleton_change_info(self):
        for beta in self.beta_array:
            print '\nInformation of skeleton with beta = ', beta, ':\n'
            self.print_skeleton_change_info(beta)
            
    # Plot the # pairs information
    def plot_stat_info(self, keyworld = 'mu', num_column = 2):
        global FIG_X_SIZE, FIG_Y_SIZE
        # setting figure structure.
        num_figure = 2*len(assumed_cosmo)  ## num of pair(2) + num/ratio of removed/added pair (2*2)
        num_row =  num_figure /num_column + min(1,np.mod(num_figure,num_column))
        single_plot_x_size = FIG_X_SIZE * 2.0 / (num_column+0.0)
        single_plot_y_size = single_plot_x_size * FIG_Y_SIZE / FIG_X_SIZE
        print 'num_row, num_column = ', num_row, num_column
        fig=plt.figure(figsize=(single_plot_x_size*num_column,single_plot_y_size*num_row))
        fig.subplots_adjust(hspace=0.6)
        # make plottings
        i_fig = 0
        for RSD in ['F', 'T']:
            for i_cosmo in range(len(assumed_cosmo)):
                legend_array = []
                subfig = fig.add_subplot(num_row, num_column, i_fig+1, xlabel = 'beta', ylabel = 'num of pairs',\
                                     title = 'Statis results of '+keyworld+'. RSD = '+RSD)
                ### TBA...
            
        
```

```python

########################################   

### Plotting information of BSkel pairs

def plot_BSke_info(RSD, AP, beta):
    global FIG_X_SIZE, FIG_Y_SIZE
    pair_data = get_pair_data(RSD, AP, beta)
    num_pair = len(pair_data)
    pair_info = [pair(pair_data[row][0], pair_data[row][1]) for row in range(num_pair)]
    fig = plt.figure(figsize=(FIG_X_SIZE*2,FIG_Y_SIZE))
    # histogram of delta_r
    delta_r_hist=fig.add_subplot(131, title = "delta r  [Mpc/h]");
    delta_r_hist.hist( [pair_info[row].delta_r for row in range(num_pair)]);
    # histogram of mu
    mu_hist=fig.add_subplot(132, title = "mu");
    mu_hist.hist( [pair_info[row].mu for row in range(num_pair)], range=(0,1));
    # hisgotram of r
    r_hist=fig.add_subplot(133, title = "Nearby Point Distance [Mpc/h]");
    r_hist.hist( [pair_info[row].r1 for row in range(num_pair)]);
    plt.suptitle('Information of Pairs: AP='+str(AP)+', RSD='+str(RSD),fontsize=14);
    plt.show()
    
########################################    

## Histogram for mu: Single Plot

def mu_histogram(mu_data, mu_title='', nbins=10):
    fig=plt.figure(figsize=(10,5));
    muhist=fig.add_subplot(111);
    
    #Histogram Plotting
    x=mu_data
    muhist.hist(x, nbins, range=(0,1));
    npoint=len(x)
    
    #Reference PDF
    X=np.linspace(0,1,nbins); Y=X-X+npoint/(nbins+0.0)
    muhist.plot(X, Y, color='black', lw=3);
    muhist.set_title(mu_title);
    
    #Estimating Error Bars
    n_array, bins_array=np.histogram(x,nbins,range=(0,1))
    X=[(bins_array[k]+bins_array[k+1])/2 for k in range(nbins)]
    Y=n_array
    Yerr=[n_array[k]/np.sqrt(0.0+n_array[k]) for k in range(nbins)]
    muhist.errorbar(X, Y, yerr=Yerr, color='red', lw=3, fmt='p');
    
    plt.show();

    
########################################    


## Histogram for mu: Multiple Plot
    
def list_mu_histogram(list_mu_data, Title_List=[], nbins=10, num_column=3, print_statis_result=False ):
    global FIG_X_SIZE, FIG_Y_SIZE    
    num_fig = len(list_mu_data)
    if(len(Title_List) != num_fig and len(Title_List) != 0):
        print 'ERROR! number of figure ', num_fig , 'and title',len(Title_List),'not match!'
        return
    
    num_row =  num_fig /num_column + min(1,np.mod(num_fig,num_column))
    fig=plt.figure(figsize=(FIG_X_SIZE*num_column,FIG_Y_SIZE*num_row));
    fig.subplots_adjust(hspace=0.6)
    
    statis_result = []
    
    
    for i_fig in range(num_fig):    
        muhist=fig.add_subplot(num_row,num_column,i_fig+1);
    
        #Histogram Plotting
        x=list_mu_data[i_fig]
        muhist.hist(x, nbins, range=(0,1));
        npoint=len(x)
        #Reference PDF
        X0=np.linspace(0,1,nbins); Y0=X0-X0+npoint/(nbins+0.0)
        muhist.plot(X0, Y0, color='black', lw=1);
    
        #Estimating Error Bars
        n_array, bins_array=np.histogram(x,nbins,range=(0,1))
        X1=[(bins_array[k]+bins_array[k+1])/2 for k in range(nbins)]
        Y1=n_array
        Yerr=[n_array[k]/np.sqrt(n_array[k]-1.0) for k in range(nbins)]
        muhist.errorbar(X1, Y1, yerr=Yerr, color='red', lw=3, fmt='p');
        
        # chisq
        chisq = sum([ ((Y0[row]-Y1[row])/Yerr[row])**2.0 for row in range(nbins)  ])
        
        # statistical quantities
        mu_av = sum(x)/len(x)
        mu_var = sum((x[row]-mu_av)**2.0 for row in range(len(x)))/(len(x)-1.0) #correct the bias using n-1
        mu_av_er = np.sqrt(mu_var) / np.sqrt(len(x)-1.0) # correct the bias using n-1 
        
        if len(Title_List) != 0:
            muhist.set_title(Title_List[i_fig]\
                             +'\n mu_av = %.4f'%mu_av+'+/-%.4f'%mu_av_er \
                             +'.\n chisq=%.2f'%chisq,fontsize=9);
        if len(Title_List) == 0:
            muhist.set_title(' mean = %.4f'%mu_av+'+/-%.4f'%mu_av_er \
                             +'.\n chisq=%.2f'%chisq,fontsize=9)
        
        now_statis_result =[ Title_List[i_fig]+': mu_av, mu_av_er, mu_var, chisq = ', \
                            [mu_av, mu_av_er, mu_var, chisq] ]
        if print_statis_result:
            print Title_List[i_fig]+': [<mu>, variance of mu, chisq] = \t[%.4f'%mu_av\
                    +'+/-%.4f'%mu_av_er+', %.4f'%mu_var+']'
        statis_result.append(now_statis_result)
        
    plt.show()  
    return statis_result
    
    
### mu-Histogram of all Skeletons (real, AP, RSD, AP+RSD) with given beta


def list_all_mu_histogram(beta, min_delta_r=-1, max_delta_r=-1, given_pair_index = False):    
    
    AP = 'F'
    RSD = 'F'
    
    mu_data1 = get_mu_data(RSD, AP, beta, min_delta_r, max_delta_r, given_pair_index)
        
    AP = 'F'
    RSD = 'T'
    
    mu_data2 = get_mu_data(RSD, AP, beta, min_delta_r, max_delta_r, given_pair_index)
        
    mu_data_list = [mu_data1, mu_data2]
    title_list = [ 'No   RSD', 'With RSD']
    
    for AP in range(len(assumed_cosmo)):
        
        RSD = 'F'

        mu_data = get_mu_data(RSD, AP, beta, min_delta_r, max_delta_r, given_pair_index)
        mu_data_list.append(mu_data)
        title_list.append('AP='+str(assumed_cosmo[AP])+', No  RSD')
        
        RSD = 'T'
        mu_data = get_mu_data(RSD, AP, beta, min_delta_r, max_delta_r, given_pair_index)
        mu_data_list.append(mu_data)
        title_list.append('AP='+str(assumed_cosmo[AP])+', With RSD')
    
    print 'number of mus: '
    print '\t', [len(mu_data_list[row]) for row in range(len(mu_data_list))], '\n'
    
    statis_result = list_mu_histogram(mu_data_list, title_list, \
                                      nbins=10, num_column=3, print_statis_result = True)
    [mu_av0, mu_av_er0, mu_var0, chisq] = statis_result[0][1]
    [mu_av1, mu_av_er1, mu_var1, chisq] = statis_result[1][1]
    print '\n'
    print ' RSD shift of <mu>:   value, sigma = %.4f'\
        %(mu_av1-mu_av0)+', %.2f'%(abs(mu_av1-mu_av0)/mu_av_er0)
    print '\n'
    for AP in range(len(assumed_cosmo)):
        print 'Assuming cosmology',assumed_cosmo[AP], ':\n'
        [mu_av, mu_av_er, mu_var, chisq] = statis_result[2+2*AP][1]
        print ' No   RSD, AP shift of <mu>:   value, sigma = %.4f'\
            %(mu_av-mu_av0)+', %.2f'%(abs(mu_av-mu_av0)/mu_av_er0)
        [mu_av, mu_av_er, mu_var, chisq] = statis_result[2+2*AP+1][1]
        print ' With RSD, AP shift of <mu>:   value, sigma = %.4f'\
            %(mu_av-mu_av1)+', %.2f'%(abs(mu_av-mu_av1)/mu_av_er0)
        print '\n'
```
