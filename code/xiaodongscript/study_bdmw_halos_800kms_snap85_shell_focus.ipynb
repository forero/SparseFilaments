## Sec 1. Useful Definitions

```python
### 

import matplotlib.pyplot as plt
import numpy as np
import commands
import mpl_toolkits.mplot3d.axes3d as Axes3D
%pylab inline
import os
import random
import scipy.integrate

#################################
# Useful Definiations

fig_x_size = 7
fig_y_size = 5

const_c = 299792.458 #unit: km/s

# get theta, phi according to x,y. 2D
def getangle(x, y):
    theta = np.arcsin(y/np.sqrt(x**2+y**2))
    if x<0:
        theta = pi-theta
    if x>0 and y<0:
        theta = 2*pi+theta
    return theta

# get r, theta, phi given x,y,z. 3D
def getSC(x, y, z):
    r=np.sqrt(x**2+y**2+z**2)
    theta=getangle(z,np.sqrt(x**2+y**2))
    phi=getangle(x,y)
    return r, theta, phi

# Find Root: f(x) == f0
def FindRoot( f, f0, xl, xr, tol = 1e-6, maxstep=100 ):
    fl = f(xl); fr = f(xr);
    if abs(fl-f0) < tol:
        return xl, fl-f0, 0, 'S'
    if abs(fr-f0) < tol:
        return xr, fr-f0, 0, 'S'
    if fl > f0 and fr > f0:
        print "Error! f(xl)/f(xr) = ", fl, fr, "both larger than ", f0, "!"
        return 0, 0, 0, 'E' # E means error
    if fl < f0 and fr < f0:
        print "Error! f(xl)/f(xr) = ", fl, fr, "both smaller than ", f0, "!"
        return 0, 0, 0, 'E'
    
    xm = xl+(xr-xl)*(f0-fl)/(fr-fl); fm = f(xm)
    step = 0
    while abs(fm-f0) > tol and step <= maxstep:
        if (fm > f0 and fl > f0) or (fm < f0 and fl < f0):
            xl = xm; fl = fm
        else:
            xr = xm; fr = fm
        xm = xl+(xr-xl)*(f0-fl)/(fr-fl); fm = f(xm)
        step +=1
    return xm, fm-f0, step, 'S' # S means success 
    
# Hubble parameter in unit of km/s/Mpc


def Hz(omegam, w, h, z):
    return 100*h*np.sqrt(omegam*(1.0+z)**3.0 + (1.0-omegam)*(1.0+z)**(3.0*(1+w)))

# comoving r in unit of Mpc/h
# in fact the result is independt on the input variable h

def comov_r(omegam, w, h, z):
     x, y = scipy.integrate.quad(lambda x: 1.0/Hz(omegam, w, h, x), 0, z)
     return const_c * x * h 

# invert of the above function: get z according to comoving_r (in unit of Mpc/h)

def get_z(omegam, w, h, given_comov_r, zl=0.0, zr=5.0):
    z, deltar, num_step, flag = FindRoot(lambda z: comov_r(omegam, w, h, z), given_comov_r, zl, zr)
    return z

def km_to_Mpctoh(v, h):
    return v / (3.08567758e19) * h
```

## Sec 3. RSD & AP Effect on Single Halos

```python
# This class describes the full information of a single halo: 
# position, velocity, redshift, RSD effect, AP effect, ...

# The position of the observer looking at the halos is x,y,z=0,0,0. 
#  This will be used to calculate the observed redshift z and comoving r

# All lengths are in unit Mpc/h; All velocities are in unit km/s

# The structure is named as "point" rather than "halo", since it can be anything (e.g., particles, galaxies, ...)

class point:
    x =0.0;  y =0.0; z =0.0; theta=0.0; phi=0.0; unit_vector = [0.0,0.0,0.0]  
    vx=0.0;  vy=0.0; vz=0.0; vlos=0.0
    
    z_real = 0.0;  z_obs = 0.0;  z_obs_diff = 0.0;  z_obs_discrep = 0.0
    
    r =0.0;        r_RSD = 0.0;  r_RSD_diff = 0.0;  r_RSD_discrep = 0.0;  xyz_RSD = [0,0,0]
    
    Hubble = 0.0
    comov_dist = 0.0
    D_A = 0.0
    fidu_cosmo = [[0.0,0.0,0.0]]
    num_fidu_cosmo = 1
    r_AP = [0.0]
    r_AP_diff = [0.0]
    r_AP_discrep = [0.0]
    xyz_AP = [[0,0,0]]
    r_AP_RSD = [0.0]
    r_AP_RSD_diff = [0.0]
    r_AP_RSD_discrep = [0.0]
    xyz_AP_RSD = [[0,0,0]]
    
    # initialize 
    
    def __init__(self, x,y,z, vx,vy,vz, fidu_cosmo = [[om_dft,w_dft,h_dft]] ): 
        
        # basic information
        self.x, self.y, self.z = x,y,z
        self.vx, self.vy, self.vz = vx,vy,vz
        self.r, self.theta, self.phi = getSC(self.x,self.y,self.z)
        self.unit_vector = [ self.x/self.r, self.y/self.r, self.z/self.r ]
        self.vlos = np.dot(self.unit_vector, [self.vx, self.vy, self.vz]) 
        #los velocity in comoving space, in unit of Mpc/h/s        
        
        # cosmological information
        self.z_real = get_z(om_dft, w_dft, h_dft, self.r)
        self.Hubble = Hz(om_dft, w_dft, h_dft, self.z_real)
        self.D_A = self.r / (1.0 + self.z_real)
        
        # observed redshift (considering RSD)
        self.z_obs = self.z_real + self.vlos / const_c  # observed, apparent redshift 
        self.z_obs_diff = self.z_obs - self.z_real 
        self.z_obs_discrep = self.z_obs_diff / self.z_real
        
        # Apparent Distances & Positions: RSD considered
        self.r_RSD = comov_r(om_dft, w_dft, h_dft, self.z_obs)
        self.r_RSD_diff = self.r_RSD - self.r
        self.r_RSD_discrep = self.r_RSD_diff / self.r
        ratio = self.r_RSD / self.r
        self.xyz_RSD = [ ratio*self.x, ratio*self.y, ratio*self.z ] # apparent x,y,z
        
        # Assumed Cosmologies
        self.fidu_cosmo = fidu_cosmo
        self.num_fidu_cosmo = len(fidu_cosmo)
        
        # Apparent Distances & Positions: AP & AP+RSD considered
        self.r_AP = range(self.num_fidu_cosmo)
        self.r_AP_RSD = range(self.num_fidu_cosmo)
        
        self.r_AP_diff = range(self.num_fidu_cosmo)
        self.r_AP_RSD_diff = range(self.num_fidu_cosmo)
        
        self.r_AP_discrep = range(self.num_fidu_cosmo)
        self.r_AP_RSD_discrep = range(self.num_fidu_cosmo)
        
        self.xyz_AP = [ [0.0, 0.0, 0.0] for row in range(self.num_fidu_cosmo) ]
        self.xyz_AP_RSD = [ [0.0, 0.0, 0.0] for row in range(self.num_fidu_cosmo) ]
        
        for i in range(self.num_fidu_cosmo):
            om_fd, w_fd, h_fd = self.fidu_cosmo[i][0], self.fidu_cosmo[i][1], self.fidu_cosmo[i][2]
            
            self.r_AP[i] = comov_r(om_fd, w_fd, h_fd, self.z_real)
            self.r_AP_diff[i] = self.r_AP[i] - self.r
            self.r_AP_discrep[i] = (self.r_AP[i] - self.r) / self.r
            ratio = self.r_AP[i] / self.r
            self.xyz_AP[i] = [ ratio*self.x, ratio*self.y, ratio*self.z ]
            
            self.r_AP_RSD[i] = comov_r(om_fd, w_fd, h_fd, self.z_obs)
            self.r_AP_RSD_diff[i] = self.r_AP_RSD[i] - self.r
            self.r_AP_RSD_discrep[i] = (self.r_AP_RSD[i] - self.r) / self.r
            ratio = self.r_AP_RSD[i] / self.r
            self.xyz_AP_RSD[i] = [ ratio*self.x, ratio*self.y, ratio*self.z ]
    
    # print information    
    def speak(self):
        print '  Information of a point object (distances Mpc/h, velocities km/s)\n'
        print '     x, y, z                         =   ', self.x,self.y,self.z
        print '     comoving r                      =   ', self.r
        print '     theta, phi                      =   ', self.theta,self.phi
        print '     vx, vy, vz                      =   ', self.vx, self.vy, self.vz
        print '     line of sight speed             =   ', self.vlos
        print '     cosmological redshift z         =   ', self.z_real
        print '     Hubble (km/s/Mpc)               =   ', self.Hubble
        print '     D_A (Mpc/h)                     =   ', self.D_A
        print '\n  Observed quantities (considering RSD; discrep = (obs - real) / real)\n'
        print '     apparent redshift z (discrep)   =   ', self.z_obs, '  (', self.z_obs_discrep, ')'
        print '     apparent comoving r (discrep)   =   ', self.r_RSD, '  (', self.r_RSD_discrep, ')'
        
        for i in range(self.num_fidu_cosmo):
            print '\n  Inferred r  (assuming (om, w, h) =', self.fidu_cosmo[i][0], \
            self.fidu_cosmo[i][1], self.fidu_cosmo[i][2], '; discrep = (obs - real) / real) \n'
            print '     inferred comov r based on  cosmological  z (discrep)   =   ', \
            self.r_AP[i], '  (', self.r_AP_discrep[i] , ')'
            print '     inferred comov r based on observed RSDed z (discrep)   =   ', \
            self.r_AP_RSD[i], '  (', self.r_AP_RSD_discrep[i] , ')'

####################################################

#### Getting the full information of all halos
#### may take ~1-2 minutes

####################################################

halo_info = [ point( halo_data[i][0], halo_data[i][1], halo_data[i][2], \
                     halo_data[i][3], halo_data[i][4], halo_data[i][5], assumed_cosmo  )\
              for i in range(num_halo)]

z_real_min = min(halo_info[i].z_real for i in range(num_halo))
z_real_max = max(halo_info[i].z_real for i in range(num_halo))
r_real_min = min(halo_info[i].r      for i in range(num_halo))
r_real_max = max(halo_info[i].r      for i in range(num_halo))

print "Information of the halo data:\n"
print "\tminimal redshift = ",z_real_min, '\n'
print "\tmaximal redshift = ", z_real_max, '\n'
print "\tminimal comoving r = ", r_real_min, '\n'
print "\tmaximal comoving r = ", r_real_max, '\n'
```

## Sec 4. Calculating BSkeletons

```python
########################################################################

### Useful Functions 

### Name of files; Where to save them.

########################################################################

# This defines the rule of file saving BSkeleton

def BSKe_file_name(RSD, AP, beta):
    if AP=='F' and RSD == 'F':
        return output_dir+halo_data_name+'__beta='+str(beta)
    if AP=='F' and RSD == 'T':
        return output_dir+halo_data_name+'__RSD'+'__beta='+str(beta)
    if AP!='F' and RSD == 'F':
        return output_dir+halo_data_name+'__'+assumed_cosmo_name[AP]+'__beta='+str(beta)
    if AP!='F' and RSD == 'T':
        return output_dir+halo_data_name+'__'+assumed_cosmo_name[AP]+'_RSD'+'__beta='+str(beta)


# This defines which scatter data are used to generating BSKeleton    
    
def get_scat_data(RSD, AP):
    if AP=='F' and RSD == 'F':
        return [ [halo_info[i].x, halo_info[i].y, halo_info[i].z] for i in range(num_halo) ]
    if AP=='F' and RSD == 'T':
        return [ halo_info[i].xyz_RSD for i in range(num_halo) ]
    if AP!='F' and RSD == 'F':
        return [ halo_info[i].xyz_AP[AP] for i in range(num_halo) ]
    if AP!='F' and RSD == 'T':
        return [ halo_info[i].xyz_AP_RSD[AP] for i in range(num_halo) ]
    
# This defines which scatter data are used to generating BSKeleton    
    
def get_pair_index(RSD, AP, beta):
    pair_index = np.loadtxt(BSKe_file_name(RSD, AP, beta))
    pair_index = [ [int(pair_index[i,0]+0.001), int(pair_index[i,1]+0.001)] for i in range(len(pair_index)) ]
    return pair_index

def get_pair_data(RSD, AP, beta):
    pair_index = get_pair_index(RSD, AP, beta)
    scat_data  = get_scat_data(RSD, AP)
    pair_data = [ [ scat_data[pair_index[i][0]], scat_data[pair_index[i][1]] ] for i in range(len(pair_index)) ]
    return pair_data

def get_pair_data_from_index(pair_index, RSD, AP, beta):
    scat_data  = get_scat_data(RSD, AP)
    pair_data = [ [ scat_data[pair_index[i][0]], scat_data[pair_index[i][1]] ] for i in range(len(pair_index)) ]
    return pair_data


# processing the pairs
# will remove redudant pairs in BSKe_file

def clean_pair(BSKe_file):
    print 'Cleaning file ', BSKe_file, '...\n'
    A=np.loadtxt(BSKe_file)
    i=0; j=0; 
    INDEX_FILE = open(BSKe_file, 'w')
    while i<len(A):
        i1 = A[i,0]; i2 = A[i,1]
        if i1 < i2: # the pairs are double-counted by ngl-beta; only keep half of them
            INDEX_FILE.write(str(int(i1+0.001))+'   '+str(int(i2+0.001))+"\n")
        i = i+1
    INDEX_FILE.close()

### Calculating BSKe by calling ngl, and clean the file using clear_pair()

def Calc_BSKe(data_array, BSKe_file, beta):
        np.savetxt('tmp.txt', data_array)
        RSD = 'F'
        AP  = 'F'
        print 'Generating neighbour graph for beta = ', beta, '...'
        command='rm '+BSKe_file+';  '\
                   +'../ngl-beta/build/binsrc/getNeighborGraph -i tmp.txt -d 3 -b'+' '+'%.lf'%beta+' >> '+BSKe_file
        print commands.getoutput(command)
        clean_pair(BSKe_file)
        os.system('rm tmp.txt')
        return    
```

### Sec 5.3 Directions of Piars: Angular Statistics (mu = cos theta)

```python
########################################

## Class pair

class pair:
    x1, y1, z1 = 0, 0, 0
    x2, y2, z2 = 0, 0, 0
    r1, r2 = 0, 0
    
    delta_r = 0
    delta_vector = [0,0,0]
    unit_delta_vector = [0,0,0]
    
    middle_point_r = 0
    middle_point_vector = [0,0,0]
    unit_middle_point_vector = [0, 0, 0]
    
    mu = 0
    
    def __init__(self, coord1, coord2):
        x1,y1,z1 = coord1[0], coord1[1], coord1[2]
        x2,y2,z2 = coord2[0], coord2[1], coord2[2]
        r1 = np.sqrt(x1**2+y1**2+z1**2)
        r2 = np.sqrt(x2**2+y2**2+z2**2)
        if r1 < r2: # always keep x1,y1,z1 the nearby one.
            self.x1, self.y1, self.z1 = x1,y1,z1
            self.x2, self.y2, self.z2 = x2,y2,z2
            self.r1 = r1; self.r2 = r2
        else:
            self.x1, self.y1, self.z1 = x2,y2,z2
            self.x2, self.y2, self.z2 = x1,y1,z1
            self.r1 = r2; self.r2 = r1
        
        self.delta_vector = [ self.x2-self.x1, self.y2-self.y1, self.z2-self.z1 ]
        self.delta_r = np.sqrt(sum(self.delta_vector[i]**2.0 for i in range(3)))
        self.unit_delta_vector = [ self.delta_vector[i]/self.delta_r for i in range(3)]
       
        self.middle_point_vector = [ (self.x2+self.x1)/2.0, (self.y2+self.y1)/2.0, (self.z2+self.z1)/2.0 ]
        self.middle_point_r = np.sqrt(sum(self.middle_point_vector[i]**2.0 for i in range(3)))
        self.unit_middle_point_vector = [ self.middle_point_vector[i]/self.middle_point_r for i in range(3)]
        
        self.mu = np.dot(self.unit_middle_point_vector, self.unit_delta_vector)
    def speak(self):
        print 'Information for the pair:\n'
        print '  Nearby  point: ', (self.x1, self.y1, self.z1) 
        print '  Farther point: ', (self.x2, self.y2, self.z2) 
        print '  Distance of Nearby // Farther point   = ', self.r1, '//', self.r2
        print '  Distance between two points           = ', self.delta_r
        print '  Separation vector                     = ', self.delta_vector
        print '  Middle point                          = ', self.middle_point_vector
        print '  Distance of middle point              = ', self.middle_point_r
        print '  Cos of angle between los/pair         = ', self.mu
#     def speak(self):
 #           print 'Information for the pair:\n'
  #          print '  Nearby point is', (self.x1, self.y1, self.y2)   

########################################    

## Get mus from pairs

def get_mu_of_pair(coord1, coord2):
    return pair(coord1,coord2).mu

def get_mulist_from_pairlist(pair_data):
    return [ get_mu_of_pair(pair_data[i][0], pair_data[i][1]) for i in range(len(pair_data)) ]

def get_selected_mulist_from_pairlist(pair_data, delta_r_min, delta_r_max):
    mu_data = []
    for i in range(len(pair_data)/2):
        now_pair = pair(pair_data[i][0], pair_data[i][1])
        if now_pair.delta_r < delta_r_max and now_pair.delta_r > delta_r_min:
            mu_data.append(now_pair.mu)
    return mu_data

########################################   

### Get mu list from a given skeleton

def get_mu_data(RSD, AP, beta, min_delta_r=-1, max_delta_r=-1, given_pair_index = False):  
    ### Calculating mu based on the BSke related to RSD, AP, beta
    if given_pair_index == False:
        
        if min_delta_r == -1 and max_delta_r==-1:
            mu_data = get_mulist_from_pairlist(get_pair_data(RSD,AP,beta))    
        else:
            mu_data = get_selected_mulist_from_pairlist(get_pair_data(RSD,AP,beta), \
                                             min_delta_r, max_delta_r)
        return mu_data
    
    #### Calculating mu based on given set of samples
    #### 
    if given_pair_index != False:
        scat_data = get_scat_data(RSD, AP)
        pair_list = [ [scat_data[given_pair_index[i][0]], \
                       scat_data[given_pair_index[i][1]]] for i in range(len(given_pair_index))  ]
        if min_delta_r == -1 and max_delta_r==-1:
            mu_data = get_mulist_from_pairlist(pair_list)    
        else:
            mu_data = get_selected_mulist_from_pairlist(pair_list, \
                                             min_delta_r, max_delta_r)
        return mu_data
```

```python

########################################   

### Plotting information of BSkel pairs

def plot_BSKe_info(RSD, AP, beta):

    pair_data = get_pair_data(RSD, AP, beta)
    num_pair = len(pair_data)
    pair_info = [pair(pair_data[i][0], pair_data[i][1]) for i in range(num_pair)]

    fig = plt.figure(figsize=(fig_x_size*2,fig_y_size))
    delta_r_hist=fig.add_subplot(131, title = "delta r  [Mpc/h]");
    delta_r_hist.hist( [pair_info[i].delta_r for i in range(num_pair)]);
    
    mu_hist=fig.add_subplot(132, title = "mu");
    mu_hist.hist( [pair_info[i].mu for i in range(num_pair)], range=(0,1));

    r_hist=fig.add_subplot(133, title = "Nearby Point Distance [Mpc/h]");
    r_hist.hist( [pair_info[i].r1 for i in range(num_pair)]);
    plt.suptitle('Information of Pairs: AP='+str(AP)+', RSD='+str(RSD),fontsize=14);
    
    plt.show()
    
########################################    

## Histogram for mu: Single Plot

def mu_histogram(mu_data, mu_title='', nbins=10):
    fig=plt.figure(figsize=(10,5));
    muhist=fig.add_subplot(111);
    
    #Histogram Plotting
    x=mu_data
    muhist.hist(x, nbins, range=(0,1));
    npoint=len(x)
    
    #Reference PDF
    X=np.linspace(0,1,nbins); Y=X-X+npoint/(nbins+0.0)
    muhist.plot(X, Y, color='black', lw=3);
    muhist.set_title(mu_title);
    
    #Estimating Error Bars
    n_array, bins_array=np.histogram(x,nbins,range=(0,1))
    X=[(bins_array[k]+bins_array[k+1])/2 for k in range(nbins)]
    Y=n_array
    Yerr=[n_array[k]/np.sqrt(0.0+n_array[k]) for k in range(nbins)]
    muhist.errorbar(X, Y, yerr=Yerr, color='red', lw=3, fmt='p');
    
    plt.show();

    
########################################    


## Histogram for mu: Multiple Plot
    
def list_mu_histogram(list_mu_data, Title_List=[], nbins=10, num_column=3, print_statis_result=False ):
        
    num_fig = len(list_mu_data)
    if(len(Title_List) != num_fig and len(Title_List) != 0):
        print 'ERROR! number of figure ', num_fig , 'and title',len(Title_List),'not match!'
        return
    
    num_row =  num_fig /num_column + min(1,np.mod(num_fig,num_column))
    fig=plt.figure(figsize=(fig_x_size*num_column,fig_y_size*num_row));
    fig.subplots_adjust(hspace=0.6)
    
    statis_result = []
    
    
    for i_fig in range(num_fig):    
        muhist=fig.add_subplot(num_row,num_column,i_fig+1);
    
        #Histogram Plotting
        x=list_mu_data[i_fig]
        muhist.hist(x, nbins, range=(0,1));
        npoint=len(x)
        #Reference PDF
        X0=np.linspace(0,1,nbins); Y0=X0-X0+npoint/(nbins+0.0)
        muhist.plot(X0, Y0, color='black', lw=1);
    
        #Estimating Error Bars
        n_array, bins_array=np.histogram(x,nbins,range=(0,1))
        X1=[(bins_array[k]+bins_array[k+1])/2 for k in range(nbins)]
        Y1=n_array
        Yerr=[n_array[k]/np.sqrt(n_array[k]-1.0) for k in range(nbins)]
        muhist.errorbar(X1, Y1, yerr=Yerr, color='red', lw=3, fmt='p');
        
        # chisq
        chisq = sum([ ((Y0[i]-Y1[i])/Yerr[i])**2.0 for i in range(nbins)  ])
        
        # statistical quantities
        mu_av = sum(x)/len(x)
        mu_var = sum((x[i]-mu_av)**2.0 for i in range(len(x)))/(len(x)-1.0) #correct the bias using n-1
        mu_av_er = np.sqrt(mu_var) / np.sqrt(len(x)-1.0) # correct the bias using n-1 
        
        if len(Title_List) != 0:
            muhist.set_title(Title_List[i_fig]\
                             +'\n mu_av = %.4f'%mu_av+'+/-%.4f'%mu_av_er \
                             +'.\n chisq=%.2f'%chisq,fontsize=9);
        if len(Title_List) == 0:
            muhist.set_title(' mean = %.4f'%mu_av+'+/-%.4f'%mu_av_er \
                             +'.\n chisq=%.2f'%chisq,fontsize=9)
        
        now_statis_result =[ Title_List[i_fig]+': mu_av, mu_av_er, mu_var, chisq = ', \
                            [mu_av, mu_av_er, mu_var, chisq] ]
        if print_statis_result:
            print Title_List[i_fig]+': [<mu>, variance of mu, chisq] = \t[%.4f'%mu_av\
                    +'+/-%.4f'%mu_av_er+', %.4f'%mu_var+']'
        statis_result.append(now_statis_result)
        
    plt.show()  
    return statis_result
    
    
### mu-Histogram of all Skeletons (real, AP, RSD, AP+RSD) with given beta


def list_all_mu_histogram(beta, min_delta_r=-1, max_delta_r=-1, given_pair_index = False):    
    
    AP = 'F'
    RSD = 'F'
    
    mu_data1 = get_mu_data(RSD, AP, beta, min_delta_r, max_delta_r, given_pair_index)
        
    AP = 'F'
    RSD = 'T'
    
    mu_data2 = get_mu_data(RSD, AP, beta, min_delta_r, max_delta_r, given_pair_index)
        
    mu_data_list = [mu_data1, mu_data2]
    title_list = [ 'No   RSD', 'With RSD']
    
    for AP in range(len(assumed_cosmo)):
        
        RSD = 'F'

        mu_data = get_mu_data(RSD, AP, beta, min_delta_r, max_delta_r, given_pair_index)
        mu_data_list.append(mu_data)
        title_list.append('AP='+str(assumed_cosmo[AP])+', No  RSD')
        
        RSD = 'T'
        mu_data = get_mu_data(RSD, AP, beta, min_delta_r, max_delta_r, given_pair_index)
        mu_data_list.append(mu_data)
        title_list.append('AP='+str(assumed_cosmo[AP])+', With RSD')
    
    print 'number of mus: '
    print '\t', [len(mu_data_list[i]) for i in range(len(mu_data_list))], '\n'
    
    statis_result = list_mu_histogram(mu_data_list, title_list, \
                                      nbins=10, num_column=3, print_statis_result = True)
    [mu_av0, mu_av_er0, mu_var0, chisq] = statis_result[0][1]
    [mu_av1, mu_av_er1, mu_var1, chisq] = statis_result[1][1]
    print '\n'
    print ' RSD shift of <mu>:   value, sigma = %.4f'\
        %(mu_av1-mu_av0)+', %.2f'%(abs(mu_av1-mu_av0)/mu_av_er0)
    print '\n'
    for AP in range(len(assumed_cosmo)):
        print 'Assuming cosmology',assumed_cosmo[AP], ':\n'
        [mu_av, mu_av_er, mu_var, chisq] = statis_result[2+2*AP][1]
        print ' No   RSD, AP shift of <mu>:   value, sigma = %.4f'\
            %(mu_av-mu_av0)+', %.2f'%(abs(mu_av-mu_av0)/mu_av_er0)
        [mu_av, mu_av_er, mu_var, chisq] = statis_result[2+2*AP+1][1]
        print ' With RSD, AP shift of <mu>:   value, sigma = %.4f'\
            %(mu_av-mu_av1)+', %.2f'%(abs(mu_av-mu_av1)/mu_av_er0)
        print '\n'
```

```python
#################################

## Comparing two pair index list. Find out the difference.



first_is_larger  = 1
first_is_smaller = -1
they_are_equal   = 0

def comp_pair(i1,i2, j1,j2):
    if i1==j1 and i2==j2:
        return they_are_equal
    if i1 < j1 or (i1 == j1 and i2 < j2):
        return first_is_smaller
    return first_is_larger

def diff_pair_index(pair_index1, pair_index2):
    pair_diff1 = []
    pair_diff2 = []
    i = 0
    j = 0
    
    while True:
        
        if i == len(pair_index1):
            while j < len(pair_index2):
                #print 'first reach boundary... i, j, append 2 = ', i, j, pair_index2[j]
                pair_diff2.append(pair_index2[j])
                j +=1
            break
        
        if j == len(pair_index2):
            while i < len(pair_index1):
                #print 'second reach boundary... i, j, append 1 = ', i, j, pair_index1[i]
                pair_diff1.append(pair_index1[i])
                i +=1
            break
            
        i1, i2 = pair_index1[i][0], pair_index1[i][1]
        j1, j2 = pair_index2[j][0], pair_index2[j][1]
        
        result = comp_pair(i1,i2, j1,j2)
        
        if result == they_are_equal:
            i +=1
            j +=1
            continue
        
        if result == first_is_smaller:
            #print 'fist_is_smaller: add', pair_index1[i], 'into diff1'
            pair_diff1.append(pair_index1[i])
            i +=1
            continue
            
        if result == first_is_larger:
            #print 'fist_is_larger: add', pair_index2[j], 'into diff2'
            pair_diff2.append(pair_index2[j])
            j +=1
            continue
    
    return pair_diff1, pair_diff2


#################################

## Sort a pair_index list


def sort_pair_index(pair_index):
    
    i = 0
    now_i1 = pair_index[i][0]
    i_begin = 0
    i_array = []
    
    while True:
        while i < len(pair_index) and pair_index[i][0] == now_i1:
            i += 1
        i_array.append([i_begin, i-1])
        
        if i < len(pair_index):
            now_i1 = pair_index[i][0]
            i_begin = i
            continue
        else:
            break
    
    for i in range(len(i_array)):
        i1 = i_array[i][0]
        i2 = i_array[i][1]
        now_segment = [ pair_index[row][1] for row in np.arange(i1, i2+1, 1)]
        
        now_segment.sort()
        for row in range(i1, i2+1):
            pair_index[row][1] = now_segment[row-i1]
            
    return pair_index
```
