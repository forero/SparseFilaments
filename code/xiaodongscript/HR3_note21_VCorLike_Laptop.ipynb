```python
#./calc_grad_chisqs chisqlike/smnum30_newdrop_sigis0.50__fixnuminx60__test 
#chisqlike/smnum30_newdrop_sigis0.50__fixnuminx60__test

import matplotlib.pyplot as plt
import numpy as np
import commands
import mpl_toolkits.mplot3d.axes3d as Axes3D
%pylab inline
import os
import random
import scipy.integrate
import gc
import re

#################################
# Useful Definiations
#################################

FIG_X_SIZE = 18
FIG_Y_SIZE = 6
FIG_FONT_SIZE = 18

plt.rc('xtick', labelsize=20) 
plt.rc('ytick', labelsize=20)

OM_DFT = 0.26
W_DFT = -1.0

CONST_C = 299792.458 #unit: km/s

#################################
# get theta, phi according to x,y. 2D
#################################
def getangle(x, y):
    theta = np.arcsin(y/np.sqrt(x**2+y**2))
    if x<0:
        theta = pi-theta
    if x>0 and y<0:
        theta = 2*pi+theta
    return theta
#################################
# get r, theta, phi given x,y,z. 3D
#################################
def getSC(x, y, z):
    r=np.sqrt(x**2+y**2+z**2)
    theta=getangle(z,np.sqrt(x**2+y**2))
    phi=getangle(x,y)
    return r, theta, phi
#################################
# Find Root: f(x) == f0
#################################
def FindRoot( f, f0, xl, xr, tol = 1e-6, maxstep=100 ):
    fl = f(xl); fr = f(xr);
    if abs(fl-f0) < tol:
        return xl, fl-f0, 0, 'S'
    if abs(fr-f0) < tol:
        return xr, fr-f0, 0, 'S'
    if fl > f0 and fr > f0:
        print "Error! f(xl)/f(xr) = ", fl, fr, "both larger than ", f0, "!"
        return 0, 0, 0, 'E' # E means error
    if fl < f0 and fr < f0:
        print "Error! f(xl)/f(xr) = ", fl, fr, "both smaller than ", f0, "!"
        return 0, 0, 0, 'E'
    
    xm = xl+(xr-xl)*(f0-fl)/(fr-fl); fm = f(xm)
    step = 0
    while abs(fm-f0) > tol and step <= maxstep:
        if (fm > f0 and fl > f0) or (fm < f0 and fl < f0):
            xl = xm; fl = fm
        else:
            xr = xm; fr = fm
        xm = xl+(xr-xl)*(f0-fl)/(fr-fl); fm = f(xm)
        step +=1
    return xm, fm-f0, step, 'S' # S means success 

#################################
# Hubble parameter in unit of km/s/Mpc
#################################
def Hz(omegam, w, h, z):
    return 100*h*np.sqrt(omegam*(1.0+z)**3.0 + (1.0-omegam)*(1.0+z)**(3.0*(1+w)))

#################################
# comoving r in unit of Mpc/h
# in fact the result is independt on the input variable h
#################################
def comov_r(omegam, w, h, z):
    global CONST_C
    x, y = scipy.integrate.quad(lambda x: 1.0/Hz(omegam, w, h, x), 0, z)
    return CONST_C * x * h 
    
#################################
# invert of the comov_r function: 
# get z according to comoving_r 
# (in unit of Mpc/h)
#################################
def get_z(omegam, w, h, given_comov_r, zl=0.0, zr=5.0):
    z, deltar, num_step, flag = FindRoot(lambda z: comov_r(omegam, w, h, z), given_comov_r, zl, zr)
    return z
#################################
# unit transition: from km to Mpc/h
#################################
def km_to_Mpctoh(v, h):
    return v / (3.08567758e19) * h

def w_kernel(q):
    if q < 0:
        return 0
    if q < 1.0:
        return  (0.25*(2.0-q)**3.0 - (1.0-q)**3.0)
    if q < 2.0:
        return  0.25*(2.0-q)**3.0
    return 0

def der_w_kernel(q):
    if q < 0:
        return 0
    if q < 1.0:
        return  (-0.75*(2.0-q)**2.0 + 3.0*(1.0-q)**2.0)
    if q < 2.0:
        return  -0.75*(2.0-q)**2.0
    return 0

def W_kernel(r, h):
    q = r / (h+0.0)
    return w_kernel(q) / (h+0.0)

def der_W_kernel(r, h):
    q = r / (h+0.0)
    return der_w_kernel(q) / (h+0.0)

def ilist(i, di, imin, imax):
    i1 = i - di
    i2 = i + di
    if i1 < imin and i2 > imax:
        return range(imin, imax+1)
    if i1 < imin:
        return range(imin, imin + 2*di +1)
    if i2 > imax:
        return range(imax - 2*di, imax +1)
    return range(i1, i2+1)

def find_mean_var(n_list):
    n = len(n_list)
    n_mean = sum(n_list) / (n+0.0)
    n_var = sum([(n_list[row]-n_mean)**2.0 for row in range(n)]) / (n-1.0)
    return n_mean, n_var

def rescale_to_norm(A):
    n = len(A)
    Y = [A[j][1] for j in range(n)]
    Ymax = max(Y)
    return [[A[j][0],(Y[j]/Ymax)] for j in range(n)]
    
    
def gen_ini(fortran_ini_file, smnum, num_in_x, auto_num_in_x=False, \
        check_boundary=True, print_info=False,  cb_adjust_ratio=1.0, \
        remov_dist_ratio=1.0,  use_num_density=True, nbins_av = 10, use_intpl_rho=True, numdrop=0,  \
        num_changenuminx = 10, amp_changenuminx = 0.1, num_om_w=0,\
        use_fixmd=True, use_segcut=True,fixmd=40, segcut=5,\
        dropval=[], lowdropvalratio=[], highdropvalratio=[],
        dropdval=[], lowdropdvalratio=[], highdropdvalratio=[],
        om_w_list=[] ):
    f = open(fortran_ini_file, "w")
    if len(dropval) != numdrop or len(dropdval)!= numdrop \
        or len(lowdropvalratio)!= numdrop or len(lowdropdvalratio)!= numdrop \
        or len(highdropvalratio)!= numdrop or len(highdropdvalratio)!= numdrop:
            print 'ERROR! # of drops mismatches with ', numdrop
            return
    if len(om_w_list) != num_om_w:
            print 'ERROR! # of om_w mismatches with ', num_om_w
            return
    # General info
    nowstr = str(smnum)+'  '+str(num_in_x)+'   '+str(auto_num_in_x)+'   '+str(check_boundary)+'   '\
        +str(print_info)+'   '+str(cb_adjust_ratio)+'   '+str(remov_dist_ratio)+'   '\
        +str(use_num_density)+'   '+str(nbins_av)+'   '+str(use_intpl_rho)+'   '+str(numdrop)+'   '\
        + str(num_changenuminx)+'   '+str(amp_changenuminx)+'   '+str(num_om_w)+'   '+str(use_fixmd)+'   '+\
        str(fixmd)+'   '+str(use_segcut)+'   '+str(segcut)\
        +'\n'
    f.write(nowstr)
    ## Many info about drops
    for i in range(numdrop):
        nowstr = str(dropval[i])+'  '+str(lowdropvalratio[i])+'  '+str(highdropvalratio[i])\
                +'  '+str(dropdval[i])+'  '+str(lowdropdvalratio[i])+'  '+str(highdropdvalratio[i])+'\n'
        f.write(nowstr)
    # om_w_list
    for i in range(len(om_w_list)):
        nowstr=str(om_w_list[i][0])+'  '+str(om_w_list[i][1])+'\n'
        f.write(nowstr)
    f.close()
    return

def gen_chisq_list(pmin, pmax, num_chisqs, om_or_w='om', om_dft = OM_DFT, w_dft = W_DFT):
    if num_chisqs == 1:
        if om_or_w == 'om':
            om_w_list = [ [pmin, w_dft] ] 
        else:
            om_w_list = [ [om_dft, pmin]] 
        return om_w_list
    dp = (pmax-pmin) / (num_chisqs - 1.0)
    if om_or_w == 'om':
        om_w_list = [ [pmin + dp*row, w_dft] for row in range(num_chisqs)] 
    else:
        om_w_list = [ [om_dft, pmin + dp*row] for row in range(num_chisqs)] 
    return om_w_list

def rm_nearby_element(A, tol):
    for i in range(len(A)):
        x = A[i]
        if x == 'NONE':
            continue
        for j in range(i+1,len(A)):
            y = A[j]
            if y =='NONE' :
                continue
            if abs(x-y) < tol:
                A[j] = 'NONE'
    A = filter(lambda x: x != 'NONE', A)
    return A

def mu_histogram(mu_data, mu_title='', nbins=15):
    fig=plt.figure(figsize=(10,5));
    muhist=fig.add_subplot(111);
    #Histogram Plotting
    x=mu_data
    muhist.hist(x, nbins, range=(-1,1));
    npoint=len(x)
    #Reference PDF
    X=np.linspace(-1,1,nbins); Y=X-X+npoint/(nbins+0.0)
    muhist.plot(X, Y, color='black', lw=3);
    muhist.set_title(mu_title);
    #Estimating Error Bars
    n_array, bins_array=np.histogram(x,nbins,range=(-1,1))
    X=[(bins_array[k]+bins_array[k+1])/2 for k in range(nbins)]
    Y=n_array
    Yerr=[n_array[k]/np.sqrt(0.0+n_array[k]) for k in range(nbins)]
    muhist.errorbar(X, Y, yerr=Yerr, color='red', lw=3, fmt='p');    
    plt.show();

def get_stat_from_list(x):
    n = len(x)
    x_av    = sum(x)/len(x)
    x_var   = (sum(x[row]**2.0 for row in range(n)) - n * x_av**2.0) /(n-1.0) #correct the bias using n-1
    if n == 1:
        return x_av, x_var, 0, 0
    else:
        x_av_er = np.sqrt(x_var) / np.sqrt(n-1.0) # correct the bias using n-1 
        x_var_er = np.sqrt(2.0/(n-1.0)) * x_var
        return x_av, x_var, x_av_er, x_var_er


def abs_mu_histogram(mu_data, mu_title='', nbins=20, figxsize=10, figysize=5, printfig=True):
    #Histogram Plotting
    x=mu_data
    x=[abs(x[i]) for i in range(len(x))]
    muav, muvar, muaver, muvarer = get_stat_from_list(x)
    chisq = ((muav-0.5)/muaver)**2.0
    if printfig:
        fig=plt.figure(figsize=(figxsize, figysize));
        muhist=fig.add_subplot(111);
        muhist.hist(x, nbins, range=(0,1));
        npoint=len(x)
        #Reference PDF
        X=np.linspace(0,1,nbins); Y=X-X+npoint/(nbins+0.0)
        muhist.plot(X, Y, color='black', lw=3);
        muhist.set_title(mu_title+' <|$\\mu$|>=%.5f'%muav+'$\\pm$%.5f'%muaver+'; $\\chi^2$=%.8f'%chisq);
        #Estimating Error Bars
        n_array, bins_array=np.histogram(x,nbins,range=(0,1))
        X=[(bins_array[k]+bins_array[k+1])/2 for k in range(nbins)]
        Y=n_array
        Yerr=[n_array[k]/np.sqrt(0.0+n_array[k]) for k in range(nbins)]
        muhist.errorbar(X, Y, yerr=Yerr, color='red', lw=3, fmt='p');    
        plt.show();
    return muav, muaver, chisq
    
def show_data(data, show_scatter = False):
    ndat = len(data)
    if ndat > 10000:
        ratio = 10000 / (ndat + 0.0)
    nowx = []; nowy = []; nowz = [];
    nowmu = [];
    for i in range(ndat):
        nowmu.append(data[i][8])
        x = random.uniform(0,1);
        if x < ratio:
            nowx.append(data[i][0])
            nowy.append(data[i][1])
            nowz.append(data[i][2])
    fig=plt.figure(figsize=(10,7));
    if show_scatter:
        ax=fig.add_subplot(111,projection='3d');
        ax.scatter(nowx,nowy,nowz);
        plt.show();
        plt.clf();
    mu_histogram(nowmu, nbins=15)
    abs_mu_histogram(nowmu, nbins=7)
    plt.clf();    

def key(xx,i=0):
    return xx[i]

def inplvl_from_sortedXY(x,X,Y):
    ndat = len(X);
    #if X[0] > x or X[ndat-1] < x:
    #    print 'WARNING (intplvl_from_sortedXY)!!! x out of range: ', x, X[0], X[1]
    diff = 1.0e30
    for i in range(ndat):
        if abs(x-X[i]) < diff:
            j = i; diff = abs(x-X[i]);
    if j == 0:
        j = 1
    elif j >= ndat-1:
        j = ndat-2
    coefficients = np.polyfit([X[j-1],X[j],X[j+1]], [Y[j-1],Y[j],Y[j+1]],2);
    polynomial = np.poly1d(coefficients);
    return j, polynomial(x)
    
def get_bf_error_from_like(xs,ys):
    bestfiti, bestfitx = inplvl_from_sortedXY(1.0,ys,xs);
    tmpy = [ys[row] for row in np.arange(len(ys)-1,min(bestfiti,len(ys)-5),-1)]
    tmpx = [xs[row] for row in np.arange(len(ys)-1,min(bestfiti,len(ys)-5),-1)]
    right1ssi, right1sx = inplvl_from_sortedXY(np.exp(-0.5),tmpy,tmpx);
    right2ssi, right2sx = inplvl_from_sortedXY(np.exp(-2.0),tmpy,tmpx);
    tmpy = [ys[row] for row in np.arange(0,max(bestfiti,4),1)] 
    tmpx = [xs[row] for row in np.arange(0,max(bestfiti,4),1)]
    left1ssi, left1sx = inplvl_from_sortedXY(np.exp(-0.5),tmpy,tmpx);
    left2ssi, left2sx = inplvl_from_sortedXY(np.exp(-2.0),tmpy,tmpx);
    return left2sx, left1sx, bestfitx, right1sx, right2sx

def XYfromdata(data,i=0,j=1):
    ndat = len(data)
    X = [data[row][i] for row in range(len(data))]
    Y = [data[row][j] for row in range(len(data))]
    return X,Y

def XYZfromdata(data,i=0,j=1,k=2):
    ndat = len(data)
    X = [data[row][i] for row in range(len(data))]
    Y = [data[row][j] for row in range(len(data))]
    Z = [data[row][k] for row in range(len(data))]
    return X,Y,Z

def XtoY(X,Y):
    Z = [];
    for row in range(len(X)):
        if Y[row] != 0.0:
            Z.append(X[row] / Y[row])
    return Z

def XtoYRef(X,Y,Ref1,Ref2):
    Z = [];
    for row in range(len(X)):
        if Y[row] != 0.0 and Ref1[row] > 0 and Ref2[row] > 0:
            Z.append(X[row] / Y[row])
    return Z

def XminusYRef(X,Y,Ref1,Ref2):
    Z = [];
    for row in range(len(X)):
        if Ref1[row] > 0 and Ref2[row] > 0:
            Z.append(X[row] - Y[row])
    return Z

def XmultY(X,Y):
    return  [X[row] * Y[row] for row in range(len(X))]

def aXplusbY(a,b,X,Y):
    return  [a*X[row] + b*Y[row] for row in range(len(X))]

Do_Plottings_for_all = False
DFT_COLOURLIST = ['b','g','r','c','m','k', 'y']

basename = 'chisqlike/'
```

```python
def multbf_plot(basestr='MPI_chisqlike_Desktop/NewCBPHR3LC',spestr='',iLClist=range(20),numcube=300,fixmd=30,segcut=0,
                smnum=30,usefixmd=True,do_vcor=True,numvcor=1,vcorsegcut=10,vcorfixmd=30,
                plotlablist=[0],kwlist=['rho','rhoRSD'],plot2s=True,showleg=False,
                droplist=np.linspace(0,75,31),labbase=31,plotallLCs = False):
    
    for nowkw in kwlist:
        bestfitlist=[];siglist1C=[];siglist1Err=[];siglist2C=[];siglist2Err=[];ax0_x=0;ax0_xlist=[];leg0=[];
        clist=[];
        for iLC in iLClist:
            if usefixmd:
                basename = basestr+'%02i'%iLC+'_zge0_rge500_om_smnum'+str(smnum)+'_'\
                    +str(numcube)+'cube_fixmd%.1f'%fixmd+'_segmincut%.1f'%segcut
            else:
                basename = basestr+'%02i'%iLC+'_zge0_rge500_om_smnum'+str(smnum)+'_'\
                    +str(numcube)+'cube'
            if do_vcor:
                basename = basename+'_'+str(numvcor)+'vcor%.1f'%vcorsegcut+'to%.1f'%vcorfixmd
            basename += spestr
            if plotallLCs:
                print basename
            numdrop=0;totlegendlist=[];
            dropval=[];lowdropvalratio=[];highdropvalratio=[];dropdval=[];lowdropdvalratio=[];highdropdvalratio=[]
            for lowratio in [0]:
                for highratio in droplist:
                    numdrop += 1; dropval.append(True); dropdval.append(True);
                    lowdropvalratio.append(0); lowdropdvalratio.append(lowratio)
                    highdropvalratio.append(highratio); highdropdvalratio.append(highratio)
                    totlegendlist.append('# of pixels = '+str(numcube)+' cube ; '+\
                        'Drop high |$n$| & |$\\nabla n$|=%5.1f'%highratio+'%')
            data=np.loadtxt(basename+'_'+nowkw+'.txt');dim1 = len(data); dim2 = len(data[0]);
            data2=[[data[row2,row1] for row1 in range(dim2)] for row2 in range(dim1)];data2.sort(key=key);
            data=data2;
            
            X=[data[row][0] for row in range(len(data))];
            nfig = len(plotlablist);
            if plotallLCs:
                fig = plt.figure(num=nfig, figsize=(FIG_X_SIZE,FIG_Y_SIZE));
                ax = fig.add_subplot(1,1,1);
                ax.set_title('Likelihood for '+nowkw+' (LC'+str(iLC)+')',fontsize=FIG_FONT_SIZE+5)
                ax.set_xlabel('$\Omega_m$',fontsize=FIG_FONT_SIZE);
                ax.set_ylabel('Likelihood',fontsize=FIG_FONT_SIZE);
            for ifig in range(nfig):
                # Plotting likelihood curve
                colour = DFT_COLOURLIST[mod(ifig,len(DFT_COLOURLIST))]
                Y = [data[row2][plotlablist[ifig]+2+labbase] for row2 in range(len(data))];
                coefficients=np.polyfit(X,Y,12);
                Ymin = min(Y);Y = [np.exp(-(Y[row]-Ymin)/2.0) for row in range(len(Y))];
                # Poly-fitted likelihood
                polynomial = np.poly1d(coefficients);
                xs = np.linspace(min(X),max(X),100); ys = polynomial(xs); ymin=min(ys);
                ys = [np.exp(-(ys[row]-ymin)/2.0) for row in range(len(ys))];
                if plotallLCs:
                    ax.plot(X,Y,c=colour,marker='o', lw=2, label = totlegendlist[ifig]);
                    grid(b=True, which='major', color='g', linestyle='-');
                    ax.plot(xs,ys,'--',c=colour,lw=2,label = totlegendlist[ifig]+' (fitted)');
                # bf & error bars
                left2sx, left1sx, bestfitx, right1sx, right2sx=get_bf_error_from_like(xs,ys)
                bestfitlist.append(bestfitx);
                siglist1C.append((left1sx+right1sx)/2.0);siglist1Err.append((right1sx-left1sx)/2.0);
                siglist2C.append((left2sx+right2sx)/2.0);siglist2Err.append((right2sx-left2sx)/2.0);
                leg0.append(nowkw+'; Dropping %5.1f'%highdropvalratio[plotlablist[ifig]]+'%; LC'+str(iLC))
                ax0_xlist.append(ax0_x);ax0_x+=1;clist.append(colour)
            if plotallLCs:
                print bestfitx, ' /+ ', right1sx-bestfitx, ' /- ', bestfitx-left1sx, ' (',(left1sx+right1sx)/2.0,'+/- ',\
                    (right1sx-left1sx)/2,' )'
                X = [0.26 for row in range(100)];Y = [0.01*row for row in range(100)]
                ax.plot(X,Y,c='k',lw=3);ax.legend(loc='best');plt.show();
        fig0 = plt.figure(num=1, figsize=(FIG_X_SIZE,FIG_Y_SIZE));
        ax0 = fig0.add_subplot(111);numfig0=len(bestfitlist);
        for i in range(len(ax0_xlist)):
            colour = DFT_COLOURLIST[mod(i,len(DFT_COLOURLIST))]
            ax0.scatter(ax0_xlist[i],bestfitlist[i],c=colour);
            ax0.errorbar(ax0_xlist[i],siglist1C[i],siglist1Err[i],c=colour,lw=3,capsize=7,label=leg0[i]);
            if plot2s:
                ax0.errorbar(ax0_xlist[i],siglist2C[i],siglist2Err[i],c=colour,lw=2,capsize=5);
        if showleg:
            axmin=-numfig0*0.35;axmax=int(numfig0*1.1)-0.5;
            leg=ax0.legend(loc='best', fancybox=True);
            leg.get_frame().set_alpha(0.5)
        else:
            axmin=-numfig0*0.1;axmax=int(numfig0*1.1)-0.5;
        ax0.set_xlim(axmin,axmax);
        numfig0=len(bestfitlist);
        bfav, bfvar, bfaver, bfvarer = get_stat_from_list(bestfitlist)
        # Default cosmology
        X = np.linspace(axmin,axmax,100);
        Y = [0.26 for row in range(100)];ax0.plot(X,Y,c='k',lw=5);
        # errorbar estimated from Var(bestfit)
        Y = [bfav for row in range(100)];ax0.plot(X,Y,c='g',ls='-',lw=2);
        Y = [bfav+sqrt(bfvar) for row in range(100)];ax0.plot(X,Y,c='g',ls='--',lw=2);
        Y = [bfav-sqrt(bfvar) for row in range(100)];ax0.plot(X,Y,c='g',ls='--',lw=2);
        if nowkw == 'rho':
            str0 = 'NO RSD'
        elif nowkw == 'rhoRSD':
            str0 = 'With RSD'
        elif nowkw == 'rhoRSD_vcor':
            str0 = 'With RSD (Vel-Cor)'
        else:
            str0 = nowkw
        if plotlablist[0] != 0:
            str0 +=' (dropping %5.1f'%highdropvalratio[plotlablist[0]]+'%)'
        if usefixmd:
            if segcut != 0:
                str0 += ', r='+str(segcut)+'-'+str(fixmd)+'Mpc/h, '
            else:
                str0 += ', r='+str(fixmd)+'Mpc/h, '
        else:
            str0 += ', # of sm = '+str(smnum)+', '
        ax0.set_title(str0+str(numcube)+'cube.\n Statistics of '+str(numfig0)+\
                      ' LC best-fits: %6.3f'%bfav+' +/-%6.3f'%sqrt(bfvar),fontsize=30)
        plt.show();
```
