```python
### SPH smoothed particle hydrodynamics paper will give yo;u the functional form of Spline Kernel. 
# Adopt n=30, for example, and use the variable h. Actually you can calculate the density and density
# gradiant at the same time using the derivative of the kernel. Monaghan ...

import matplotlib.pyplot as plt
import numpy as np
import commands
import mpl_toolkits.mplot3d.axes3d as Axes3D
%pylab inline
import os
import random
import scipy.integrate

#################################
# Useful Definiations
#################################

FIG_X_SIZE = 7
FIG_Y_SIZE = 5

CONST_C = 299792.458 #unit: km/s

PLOT_COLOR_ARRAY = ['k', 'b', 'g', 'r', 'c', 'm', 'y']
PLOT_STYLE_ARRAY = ['-',  '--', '-.' ]#,  'o', 'v', '+', 'x', 's', '*', 'D', ':']

NOW_PLOT_COLOR = 0
NOW_PLOT_STYLE = 0
NUM_PLOT_COLOR = len(PLOT_COLOR_ARRAY)
NUM_PLOT_STYLE = len(PLOT_STYLE_ARRAY)

def get_color_style():
    global NOW_PLOT_COLOR, NOW_PLOT_STYLE
    i = mod(NOW_PLOT_COLOR, NUM_PLOT_COLOR)
    j = mod(NOW_PLOT_COLOR, NUM_PLOT_STYLE)
    color_style = PLOT_COLOR_ARRAY[i]+PLOT_STYLE_ARRAY[j]
    NOW_PLOT_COLOR = NOW_PLOT_COLOR + 1
    NOW_PLOT_STYLE = NOW_PLOT_STYLE + 1
    return color_style

#################################
# get theta, phi according to x,y. 2D
#################################
def getangle(x, y):
    theta = np.arcsin(y/np.sqrt(x**2+y**2))
    if x<0:
        theta = pi-theta
    if x>0 and y<0:
        theta = 2*pi+theta
    return theta
#################################
# get r, theta, phi given x,y,z. 3D
#################################
def getSC(x, y, z):
    r=np.sqrt(x**2+y**2+z**2)
    theta=getangle(z,np.sqrt(x**2+y**2))
    phi=getangle(x,y)
    return r, theta, phi
#################################
# Find Root: f(x) == f0
#################################
def FindRoot( f, f0, xl, xr, tol = 1e-6, maxstep=100 ):
    fl = f(xl); fr = f(xr);
    if abs(fl-f0) < tol:
        return xl, fl-f0, 0, 'S'
    if abs(fr-f0) < tol:
        return xr, fr-f0, 0, 'S'
    if fl > f0 and fr > f0:
        print "Error! f(xl)/f(xr) = ", fl, fr, "both larger than ", f0, "!"
        return 0, 0, 0, 'E' # E means error
    if fl < f0 and fr < f0:
        print "Error! f(xl)/f(xr) = ", fl, fr, "both smaller than ", f0, "!"
        return 0, 0, 0, 'E'
    
    xm = xl+(xr-xl)*(f0-fl)/(fr-fl); fm = f(xm)
    step = 0
    while abs(fm-f0) > tol and step <= maxstep:
        if (fm > f0 and fl > f0) or (fm < f0 and fl < f0):
            xl = xm; fl = fm
        else:
            xr = xm; fr = fm
        xm = xl+(xr-xl)*(f0-fl)/(fr-fl); fm = f(xm)
        step +=1
    return xm, fm-f0, step, 'S' # S means success 

#################################
# Hubble parameter in unit of km/s/Mpc
#################################
def Hz(omegam, w, h, z):
    return 100*h*np.sqrt(omegam*(1.0+z)**3.0 + (1.0-omegam)*(1.0+z)**(3.0*(1+w)))

#################################
# comoving r in unit of Mpc/h
# in fact the result is independt on the input variable h
#################################
def comov_r(omegam, w, h, z):
    global CONST_C
    x, y = scipy.integrate.quad(lambda x: 1.0/Hz(omegam, w, h, x), 0, z)
    return CONST_C * x * h 
    
#################################
# invert of the comov_r function: 
# get z according to comoving_r 
# (in unit of Mpc/h)
#################################
def get_z(omegam, w, h, given_comov_r, zl=0.0, zr=5.0):
    z, deltar, num_step, flag = FindRoot(lambda z: comov_r(omegam, w, h, z), given_comov_r, zl, zr)
    return z
#################################
# unit transition: from km to Mpc/h
#################################
def km_to_Mpctoh(v, h):
    return v / (3.08567758e19) * h

#################################
## Sort a pair_index list
#################################

def sort_pair_index(pair_index):
    
    i = 0
    now_i1 = pair_index[i][0]
    i_begin = 0
    i_array = []
    
    while True:
        while i < len(pair_index) and pair_index[i][0] == now_i1:
            i += 1
        i_array.append([i_begin, i-1])
        
        if i < len(pair_index):
            now_i1 = pair_index[i][0]
            i_begin = i
            continue
        else:
            break
    
    for i in range(len(i_array)):
        i1 = i_array[i][0]
        i2 = i_array[i][1]
        now_segment = [ pair_index[row][1] for row in np.arange(i1, i2+1, 1)]
        
        now_segment.sort()
        for row in range(i1, i2+1):
            pair_index[row][1] = now_segment[row-i1]
            
    return pair_index

#################################
## Comparing two pair index list. 
## Find out the difference.
#################################

first_is_larger  = 1
first_is_smaller = -1
they_are_equal   = 0

def comp_pair(i1,i2, j1,j2):
    if i1==j1 and i2==j2:
        return they_are_equal
    if i1 < j1 or (i1 == j1 and i2 < j2):
        return first_is_smaller
    return first_is_larger

def diff_pair_index(pair_index1, pair_index2):
    pair_diff1 = []
    pair_diff2 = []
    i = 0
    j = 0
    
    while True:
        
        if i == len(pair_index1):
            while j < len(pair_index2):
                #print 'first reach boundary... i, j, append 2 = ', i, j, pair_index2[j]
                pair_diff2.append(pair_index2[j])
                j +=1
            break
        
        if j == len(pair_index2):
            while i < len(pair_index1):
                #print 'second reach boundary... i, j, append 1 = ', i, j, pair_index1[i]
                pair_diff1.append(pair_index1[i])
                i +=1
            break
            
        i1, i2 = pair_index1[i][0], pair_index1[i][1]
        j1, j2 = pair_index2[j][0], pair_index2[j][1]
        
        result = comp_pair(i1,i2, j1,j2)
        
        if result == they_are_equal:
            i +=1
            j +=1
            continue
        
        if result == first_is_smaller:
            #print 'fist_is_smaller: add', pair_index1[i], 'into diff1'
            pair_diff1.append(pair_index1[i])
            i +=1
            continue
            
        if result == first_is_larger:
            #print 'fist_is_larger: add', pair_index2[j], 'into diff2'
            pair_diff2.append(pair_index2[j])
            j +=1
            continue
    
    return pair_diff1, pair_diff2

###############
# distance of two point
def distance(p0, p1):
    if len(p0) == len(p1):
        x = 0
        for i in range(len(p0)):
            x += (p0[i] - p1[i])**2.0
        return np.sqrt(x)
    print 'error! dim of two vectors not match'

###############
# cubic spline kernel and their derivatives.
# truncated at 2*h
def w_kernel(q):
    sigma = 1.0/np.pi
    if q < 0:
        return 0
    if q < 1.0:
        return sigma * (0.25*(2.0-q)**3.0 - (1.0-q)**3.0)
    if q < 2.0:
        return sigma * 0.25*(2.0-q)**3.0
    return 0

def der_w_kernel(q):
    sigma = 1.0/np.pi
    if q < 0:
        return 0
    if q < 1.0:
        return sigma * (-0.75*(2.0-q)**2.0 + 3.0*(1.0-q)**2.0)
    if q < 2.0:
        return sigma * -0.75*(2.0-q)**2.0
    return 0

def W_kernel(r, h):
    q = r / (h+0.0)
    return w_kernel(q) / (h+0.0)**3.0

def der_W_kernel(r, h):
    q = r / (h+0.0)
    return der_w_kernel(q) / (h+0.0)**4.0

def gauss(r, sigma):
    return exp(-r**2.0/(2.0*sigma**2.0))/(np.sqrt(2.0*np.pi*sigma**2.0)**3.0)
    
def der_gauss(r, sigma):
    return (-r/sigma**2.0)*exp(-r**2.0/(2.0*sigma**2.0))/(np.sqrt(2.0*np.pi*sigma**2.0)**3.0)
# Test the kernel
if False:
    h = 5
    sigma = 0.6 * h
    print scipy.integrate.quad(lambda q: 4*np.pi*q**2.0*W_kernel(q,h), 0, 2.0*h)
    X = np.linspace(0,10,200)
    Y0 = [W_kernel(X[row], h) for row in range(len(X))]
    Y1 = [gauss(X[row], sigma) for row in range(len(X))]
    plt.plot(X, Y0)
    plt.plot(X, Y1)
    plt.show()
    Y0 = [der_W_kernel(X[row], h) for row in range(len(X))]
    Y1 = [der_gauss(X[row], sigma) for row in range(len(X))]
    plt.clf()
    plt.plot(X, Y0)
    plt.plot(X, Y1)
    plt.show()
```

```python
##################################################################
# This defines the full information of a single halo,
#  including redshift and postions w/wo RSD or AP. 
# Position of the observer is x,y,z=0,0,0. 
# All lengths in unit Mpc/h; all velocities in unit km/s
# The structure is named as "point" rather than "halo", 
#  since it can be anything (e.g., particles, galaxies, ...)
##################################################################
class point:
    mass = 0.0;
    massvir = 0.0;
    masstot = 0.0;
    x =0.0;  y =0.0; z =0.0; theta=0.0; phi=0.0; unit_vector = [0.0,0.0,0.0]  
    vx=0.0;  vy=0.0; vz=0.0; vlos=0.0
    # resdhift and distance w/wo RSD
    z_real = 0.0;  z_obs = 0.0;  z_obs_diff = 0.0;  z_obs_discrep = 0.0
    r =0.0;        r_RSD = 0.0;  r_RSD_diff = 0.0;  r_RSD_discrep = 0.0;  xyz_RSD = [0,0,0]
    # cosmologicay information
    Hubble = 0.0
    comov_dist = 0.0
    D_A = 0.0
    # assumed cosmologies
    fidu_cosmo = [[0.0,0.0,0.0]]
    num_fidu_cosmo = 1
    # distances and postions with AP
    r_AP = [0.0]
    r_AP_diff = [0.0]
    r_AP_discrep = [0.0]
    xyz_AP = [[0,0,0]]
    r_AP_RSD = [0.0]
    r_AP_RSD_diff = [0.0]
    r_AP_RSD_discrep = [0.0]
    xyz_AP_RSD = [[0,0,0]]
    # initialize     
    def __init__(self, x,y,z, vx,vy,vz, fidu_cosmo = [[om_dft,w_dft,h_dft]], massvir = 0.0, masstot = 0.0 ): 
        global CONST_C
        # basic information
        self.mass = masstot
        self.masstot = masstot
        self.massvir = massvir
        self.x, self.y, self.z = x,y,z
        self.vx, self.vy, self.vz = vx,vy,vz
        self.r, self.theta, self.phi = getSC(self.x,self.y,self.z)
        self.unit_vector = [ self.x/self.r, self.y/self.r, self.z/self.r ]
        self.vlos = np.dot(self.unit_vector, [self.vx, self.vy, self.vz]) #los velocity in comoving space        
        # cosmological information
        self.z_real = get_z(om_dft, w_dft, h_dft, self.r)
        self.Hubble = Hz(om_dft, w_dft, h_dft, self.z_real)
        self.D_A = self.r / (1.0 + self.z_real)
        # observed redshift (considering RSD)
        self.z_obs = self.z_real + self.vlos / CONST_C  # observed, apparent redshift 
        self.z_obs_diff = self.z_obs - self.z_real 
        self.z_obs_discrep = self.z_obs_diff / self.z_real
        # Apparent Distances & Positions: RSD considered
        self.r_RSD = comov_r(om_dft, w_dft, h_dft, self.z_obs)
        self.r_RSD_diff = self.r_RSD - self.r
        self.r_RSD_discrep = self.r_RSD_diff / self.r
        ratio = self.r_RSD / self.r
        self.xyz_RSD = [ ratio*self.x, ratio*self.y, ratio*self.z ] # apparent x,y,z
        # Assumed Cosmologies
        self.fidu_cosmo = fidu_cosmo
        self.num_fidu_cosmo = len(fidu_cosmo)
        # Apparent Distances & Positions: AP & AP+RSD considered
        self.r_AP = range(self.num_fidu_cosmo)
        self.r_AP_RSD = range(self.num_fidu_cosmo)
        self.r_AP_diff = range(self.num_fidu_cosmo)
        self.r_AP_RSD_diff = range(self.num_fidu_cosmo)
        self.r_AP_discrep = range(self.num_fidu_cosmo)
        self.r_AP_RSD_discrep = range(self.num_fidu_cosmo)
        self.xyz_AP = [ [0.0, 0.0, 0.0] for row in range(self.num_fidu_cosmo) ]
        self.xyz_AP_RSD = [ [0.0, 0.0, 0.0] for row in range(self.num_fidu_cosmo) ]
        # Distance and postions with AP:
        for i in range(self.num_fidu_cosmo):
            om_fd, w_fd, h_fd = self.fidu_cosmo[i][0], self.fidu_cosmo[i][1], self.fidu_cosmo[i][2]
            # wo RSD
            self.r_AP[i] = comov_r(om_fd, w_fd, h_fd, self.z_real)
            self.r_AP_diff[i] = self.r_AP[i] - self.r
            self.r_AP_discrep[i] = (self.r_AP[i] - self.r) / self.r
            ratio = self.r_AP[i] / self.r
            self.xyz_AP[i] = [ ratio*self.x, ratio*self.y, ratio*self.z ]
            # w RSD
            self.r_AP_RSD[i] = comov_r(om_fd, w_fd, h_fd, self.z_obs)
            self.r_AP_RSD_diff[i] = self.r_AP_RSD[i] - self.r
            self.r_AP_RSD_discrep[i] = (self.r_AP_RSD[i] - self.r) / self.r
            ratio = self.r_AP_RSD[i] / self.r
            self.xyz_AP_RSD[i] = [ ratio*self.x, ratio*self.y, ratio*self.z ]
    # print out information    
    def speak(self):
        print '  Information of a point object (distances Mpc/h, velocities km/s)\n'
        print '     mass (vir)                      =   ', self.massvir
        print '     mass (tot)                      =   ', self.masstot
        print '     x, y, z                         =   ', self.x,self.y,self.z
        print '     comoving r                      =   ', self.r
        print '     theta, phi                      =   ', self.theta,self.phi
        print '     vx, vy, vz                      =   ', self.vx, self.vy, self.vz
        print '     line of sight speed             =   ', self.vlos
        print '     cosmological redshift z         =   ', self.z_real
        print '     Hubble (km/s/Mpc)               =   ', self.Hubble
        print '     D_A (Mpc/h)                     =   ', self.D_A
        print '\n  Observed quantities (considering RSD; discrep = (obs - real) / real)\n'
        print '     apparent redshift z (discrep)   =   ', self.z_obs, '  (', self.z_obs_discrep, ')'
        print '     apparent comoving r (discrep)   =   ', self.r_RSD, '  (', self.r_RSD_discrep, ')'
        for i in range(self.num_fidu_cosmo):
            print '\n  Inferred r  (assuming (om, w, h) =', self.fidu_cosmo[i][0], \
            self.fidu_cosmo[i][1], self.fidu_cosmo[i][2], '; discrep = (obs - real) / real) \n'
            print '     inferred comov r based on  cosmological  z (discrep)   =   ', \
            self.r_AP[i], '  (', self.r_AP_discrep[i] , ')'
            print '     inferred comov r based on observed RSDed z (discrep)   =   ', \
            self.r_AP_RSD[i], '  (', self.r_AP_RSD_discrep[i] , ')'

##################################################################
# Getting the full information of all halos. 
# May take ~1-2 minutes
##################################################################

halo_info = [ point( halo_data[row][0], halo_data[row][1], halo_data[row][2], \
                     halo_data[row][3], halo_data[row][4], halo_data[row][5], assumed_cosmo, \
                     halo_data[row][6], halo_data[row][7]  )\
              for row in range(num_halo)]

##################################################################
# Minimal/maximal z, r, phi. 
#  shall be useful since we select a slice based on z/r/phi and plot
##################################################################

z_real_min = min(halo_info[row].z_real for row in range(num_halo))
z_real_max = max(halo_info[row].z_real for row in range(num_halo))
r_real_min = min(halo_info[row].r      for row in range(num_halo))
r_real_max = max(halo_info[row].r      for row in range(num_halo))
phi_min    = min(halo_info[row].phi    for row in range(num_halo))
phi_max    = max(halo_info[row].phi    for row in range(num_halo))
print "Information of the halo data:\n"
print "\tminimal redshift = ",z_real_min
print "\tmaximal redshift = ", z_real_max
print "\tminimal comoving r = ", r_real_min
print "\tmaximal comoving r = ", r_real_max
print "\tminimal phi        = ", phi_min
print "\tmaximal phi        = ", phi_max
#x= [ halo_info[row].massvir for row in range(num_halo)]
#plt.hist(x, bins = 100);
#x= [ halo_info[row].masstot for row in range(num_halo)]
#plt.hist(x, bins = 100);
#x= [ halo_info[row].masstot / halo_info[row].massvir for row in range(num_halo)]
#plt.hist(x, bins = 100, range=(1,1.5));
```

```python
# dividing halos into cells.
# dividing according to cosmological (not observational) position
def do_cell_initialize(RSD, AP, num_in_x = 10, print_info = False):
    global xyz_list, xmin, xmax, ymin, ymax, zmin, zmax, \
           deltax, deltay, deltaz, n_cellx, n_celly, n_cellz, \
           cell_vol, cell_mat, mass_mat
    if print_info:
        print 'Initializing grids of cells with RSD, AP = ', RSD, AP, '...'
    if RSD == 'F' and AP == 'F':
        xyz_list = [[halo_info[i].x, halo_info[i].y, halo_info[i].z] for i in range(num_halo)]
    elif RSD == 'T' and AP == 'F':
        xyz_list = [halo_info[i].xyz_RSD for i in range(num_halo)]
    elif RSD == 'F':
        xyz_list = [halo_info[i].xyz_AP[AP] for i in range(num_halo)]
    elif RSD == 'T':
        xyz_list = [halo_info[i].xyz_AP_RSD[AP] for i in range(num_halo)]
    # ranges of box
    x_list = [xyz_list[row][0] for row in range(num_halo)]
    y_list = [xyz_list[row][1] for row in range(num_halo)]
    z_list = [xyz_list[row][2] for row in range(num_halo)]
    xmin = min(x_list); xmax = max(x_list);
    ymin = min(y_list); ymax = max(y_list);
    zmin = min(z_list); zmax = max(z_list);
    #X = [halo_info[row].z for row in range(num_halo)]
    #plt.hist(X )
    # number of cells in each direction
    unit_len = (xmax - xmin) / (num_in_x + 0.0)
    n_cellx = int(round((xmax - xmin)/unit_len))
    n_celly = int(round((ymax - ymin)/unit_len))
    n_cellz = int(round((zmax - zmin)/unit_len))
    deltax = ((xmax - xmin)/(0.0+n_cellx))
    deltay = ((ymax - ymin)/(0.0+n_celly))
    deltaz = ((zmax - zmin)/(0.0+n_cellz))
    # volome of one cell
    cell_vol = deltax * deltay * deltaz
    # which halo is in which cell
    cell_mat = [[[ [] for row in range(n_cellx)] for row2 in range(n_celly)] for row3 in range(n_cellz) ]
    for i in range(num_halo):
        x, y, z = xyz_list[i][0], xyz_list[i][1], xyz_list[i][2]
        ix = max(min(int((x-xmin) / deltax), n_cellx - 1), 0)
        iy = max(min(int((y-ymin) / deltay), n_celly - 1), 0)
        iz = max(min(int((z-zmin) / deltaz), n_cellz - 1), 0)
        cell_mat[ix][iy][iz].append(i) 
    # total mass in all cells
    mass_mat = [[[ 0 for row in range(n_cellx)] for row2 in range(n_celly)] for row3 in range(n_cellz) ]
    for i in range(n_cellx):
        for j in range(n_celly):
            for k in range(n_cellz):
                halo_list = cell_mat[i][j][k]
                now_num_halo = len(halo_list)
                mass = 0
                for i_halo in halo_list:
                    mass += halo_info[i_halo].mass
                mass_mat[i][j][k] = mass
    if print_info:
        print 'Initalization finish. \n\tCell num in x/y/z are: ', n_cellx, n_celly, n_cellz, \
            '. \n\tLength in x/y/z are: ', deltax, deltay, deltaz            
                

# central position of the cell            
def cell_pos(ix, iy, iz):
    global xmin, ymin, zmin, deltax, deltay, deltaz
    return [xmin + (ix+0.5)*deltax, ymin + (iy+0.5)*deltay, zmin + (iz+0.5)*deltaz]          

# average rho of a cell
def cell_rho(ix, iy, iz):
    global mass_mat, cell_vol
    mass = mass_mat[ix][iy][iz] 
    return mass / cell_vol

# return a list, with center i, number of elements 2*di+1, 
# but confined in the range (imin, imax)
def ilist(i, di, imin, imax):
    i1 = i - di
    i2 = i + di
    if i1 < imin and i2 > imax:
        return range(imin, imax+1)
    if i1 < imin:
        return range(imin, imin + 2*di +1)
    if i2 > imax:
        return range(imax - 2*di, imax +1)
    return range(i1, i2+1)

# gradient of rho in the cell.
def cell_drho(ix, iy, iz):
    ix_list = ilist(ix, 1, 0, n_cellx-1)
    iy_list = ilist(iy, 1, 0, n_celly-1)
    iz_list = ilist(iz, 1, 0, n_cellz-1)
    rho2 = cell_rho(ix_list[2],iy,iz)
    rho1 = cell_rho(ix_list[0],iy,iz)
    drhodx = (rho2 - rho1) / (2*deltax)
    rho2 = cell_rho(ix,iy_list[2],iz)
    rho1 = cell_rho(ix,iy_list[0],iz)
    drhody = (rho2 - rho1) / (2*deltay)
    rho2 = cell_rho(ix,iy,iz_list[2])
    rho1 = cell_rho(ix,iy,iz_list[0])
    drhodz = (rho2 - rho1) / (2*deltaz)
    return drhodx, drhody, drhodz

### example
if False:
    do_cell_initialize(RSD = 'F', AP = 'F', print_info = True)
    ix, iy, iz = 3,4,5
    print cell_pos(ix,iy,iz)
    print cell_rho(ix,iy,iz)
    print cell_drho(ix,iy,iz)
    X = []
    for i in range(n_cellx):
      for j in range(n_celly):
        for k in range(n_cellz):
            X.append(len(cell_mat[i][j][k]))
    plt.hist(X, bins = 22);
```

```python
# select halos in the cells nearby the position.
# make sure the number is far larger than the required num
def nb_select(x, y, z, num = 30, ra_ratio = 4):
    global xmin, deltax, ymin, deltay, zmin, deltaz, \
           n_cellx, n_celly, n_cellz
    ix = max(min(int((x-xmin) / deltax), n_cellx - 1), 0)
    iy = max(min(int((y-ymin) / deltay), n_celly - 1), 0)
    iz = max(min(int((z-zmin) / deltaz), n_cellz - 1), 0)
    dev = max(abs(xmin + ix*deltax - x)/deltax, abs(ymin + iy*deltay - y)/deltay,  abs(zmin + iz*deltaz - z)/deltaz)
    ratio = ra_ratio * (1 + dev * 1.5)
    requirednum = int(num * ratio)
    di = 0
    while True:
        ix_list = ilist(ix, di, 0, n_cellx-1)
        iy_list = ilist(iy, di, 0, n_celly-1)
        iz_list = ilist(iz, di, 0, n_cellz-1)
        now_num = 0
        for i in ix_list:
            for j in iy_list:
                for k in iz_list:
                    now_num += len(cell_mat[i][j][k])
        if now_num > requirednum: 
            break
        di += 1
    halo_list = []        
    for i in ix_list:
            for j in iy_list:
                for k in iz_list:
                    halo_list += cell_mat[i][j][k]
    return halo_list
# select halos in the cells nearby the position.
def nb_di_select(x, y, z, di):
    ix = max(min(int((x-xmin) / deltax), n_cellx - 1), 0)
    iy = max(min(int((y-ymin) / deltay), n_celly - 1), 0)
    iz = max(min(int((z-zmin) / deltaz), n_cellz - 1), 0)
    if True:
        ix_list = ilist(ix, di, 0, n_cellx-1)
        iy_list = ilist(iy, di, 0, n_celly-1)
        iz_list = ilist(iz, di, 0, n_cellz-1)
    halo_list = []        
    for i in ix_list:
            for j in iy_list:
                for k in iz_list:
                    halo_list += cell_mat[i][j][k]
    return halo_list

### Estimating rho and gradient of rho based on cubic spline kernel.
#
def nb_list(x, y, z, num = 30):
    global xyz_list, halo_info
    halo_list = nb_select(x, y, z, num, ra_ratio=4)
    tmp = []
    for i in range(len(halo_list)):
        now_halo = halo_info[halo_list[i]]
        now_xyz = xyz_list[halo_list[i]]
        tmp.append([ distance([x,y,z],now_xyz), halo_list[i], now_xyz, \
                    now_halo.mass])
    def sort_key(x):
        return x[0]
    tmp.sort(key=sort_key)
    # quantities
    distance_array = [tmp[row][0] for row in range(num)]
    index_array = [tmp[row][1] for row in range(num)]
    xyz_array = [tmp[row][2] for row in range(num)]
    mass_array = [tmp[row][3] for row in range(num)]
    # kernel
    max_dist = distance_array[num-1]
    h = max_dist / 2.0
    W_array = [W_kernel(distance_array[row], h) for row in range(num)]
    dWdx_array = [-(xyz_array[row][0]-x) / distance_array[row] *\
                  der_W_kernel(distance_array[row], h) for row in range(num)]
    dWdy_array = [-(xyz_array[row][1]-y) / distance_array[row] *\
                  der_W_kernel(distance_array[row], h) for row in range(num)]
    dWdz_array = [-(xyz_array[row][2]-z) / distance_array[row] *\
                  der_W_kernel(distance_array[row], h) for row in range(num)]
    # estimated quantities
    rho = sum( [mass_array[row]*W_array[row] for row in range(num)] )
    drhodx = sum( [mass_array[row]*dWdx_array[row] for row in range(num)] )
    drhody = sum( [mass_array[row]*dWdy_array[row] for row in range(num)] )
    drhodz = sum( [mass_array[row]*dWdz_array[row] for row in range(num)] )
    #print distance_array
    #print index_array
    #print xyz_array        
    #print W_array
    #print dWdx_array
    #print dWdy_array
    #print dWdz_array
    ##print 'estimated rho = ', rho
    #print 'estimated gradient of rho: ', drhodx, drhody, drhodz
    return rho, drhodx, drhody, drhodz

### example
if True:
    RSD = 'F'; AP = 0;
    do_cell_initialize(RSD, AP)
    ix, iy, iz = 3,8,5
    x,y,z = cell_pos(ix,iy,iz)
    print 'position:', x,y,z
    print '\nrho and drho estimated by using cell averaging: \n\t', cell_rho(ix,iy,iz), cell_drho(ix,iy,iz)
    print '\nrho and drho estimated by using cubic splining: \n\t', nb_list(x,y,z,  num = 100)
```

```python
### Calculating rho and drho in some grid points

def grid_rho_drho_list(RSD, AP, num=30, num_in_x = 10, print_info = False, boundary_remove = 0, \
                       only_cell_average = False):
    do_cell_initialize(RSD, AP, num_in_x, print_info=print_info)
    pos_list = []
    rho_list0 = []
    rho_list1 = []
    drho_list0 = []
    drho_list1 = []
    for ix in range(boundary_remove, n_cellx-boundary_remove):
      if print_info:
        print 'ix = ', ix, '...'
      for iy in range(boundary_remove, n_celly-boundary_remove):
        for iz in range(boundary_remove, n_cellz-boundary_remove):
            [x,y,z] = cell_pos(ix,iy,iz)
            pos_list.append([x,y,z])
            rho_list0.append(cell_rho(ix,iy,iz))
            drhodx, drhody, drhodz = cell_drho(ix,iy,iz)
            drho_list0.append([drhodx, drhody, drhodz])
            if only_cell_average:
                continue
            rho, drhox, drhoy, drhoz = nb_list(x,y,z, num)
            rho_list1.append(rho)
            drho_list1.append([drhox,drhoy,drhoz])
    return pos_list, rho_list0,  drho_list0, rho_list1, drho_list1
```

## Sec 4. mu statistics

```python
# los mu = cos theta, from position & gradient vector
def get_mu_of_gradient(pos, gradient, abs_mu = False):
    [fx, fy, fz] = gradient
    r = pos
    len_r = np.sqrt(r[0]*r[0] + r[1]*r[1] + r[2]*r[2])
    dr = gradient
    len_dr = np.sqrt(dr[0]*dr[0] + dr[1]*dr[1] + dr[2]*dr[2])
    if abs_mu:
        return abs(np.dot(r, dr)/len_r/len_dr)
    else:
        return np.dot(r, dr)/len_r/len_dr

def get_mu_from_list(pos_list, gradient_list, abs_mu = False):
    if len(pos_list) != len(gradient_list):
        print 'length of pos_list and gradient_list not matched!'
        print '   Their lengths = ', len(pos_list), len(gradient_list)
        return 
    return [get_mu_of_gradient(pos_list[row], gradient_list[row], abs_mu) for row in range(len(pos_list))]

    
# Getting mu list from a given skeleton
def get_mu_data(RSD, AP, num=30, num_in_x = 10, print_info = True, \
                boundary_remove = 0, only_cell_average = False, abs_mu = False):
    pos_list, rho_list0, drho_list0, rho_list1, drho_list1 = \
            grid_rho_drho_list(RSD, AP, num, num_in_x, print_info, boundary_remove, only_cell_average)
    mu_data0 = get_mu_from_list(pos_list, drho_list0, abs_mu)
    mu_data1 = []
    if not only_cell_average:
        mu_data1 = get_mu_from_list(pos_list, drho_list1, abs_mu)
    return mu_data0, mu_data1

# Histogram of mu: Single Plot
def mu_histogram(mu_data, mu_title='', nbins=10, only_do_statis = False, abs_mu = False):
    global FIG_X_SIZE, FIG_Y_SIZE
    fig=plt.figure(figsize=(FIG_X_SIZE,FIG_Y_SIZE));
    muhist=fig.add_subplot(111);
    #Histogram Plotting
    x=mu_data
    #abs_x = [abs(x[row]) for row in range(len(x))]
    # statistical quantities
    mu_av = sum(x)/len(x)
    mu_var = sum((x[row]-mu_av)**2.0 for row in range(len(x)))/(len(x)-1.0)
    mu_av_er = np.sqrt(mu_var) / np.sqrt(len(x)-1.0) # correct the bias using n-1 
    statis_result =[mu_av, mu_av_er, mu_var]
    if only_do_statis:
        return statis_result
    if abs_mu:
        xmin = 0; xmax = 1
    else:
        #nbins *= 2; 
        xmin = -1; xmax = 1
    muhist.hist(x, nbins, range=(xmin, xmax));
    npoint=len(x)
    #Reference PDF
    X0=np.linspace(xmin,xmax,nbins); Y0=X0-X0+npoint/(nbins+0.0)
    muhist.plot(X0, Y0, color='black', lw=1);
    #Estimating Error Bars
    n_array, bins_array=np.histogram(x,nbins,range=(xmin, xmax))
    X1=[(bins_array[k]+bins_array[k+1])/2 for k in range(nbins)]
    Y1=n_array
    Yerr=[n_array[k]/np.sqrt(n_array[k]-1.0) for k in range(nbins)]
    muhist.errorbar(X1, Y1, yerr=Yerr, color='red', lw=3, fmt='p');
    # chisq
    chisq = sum([ ((Y0[row]-Y1[row])/Yerr[row])**2.0 for row in range(nbins)  ])  
    # title
    muhist.set_title(mu_title    +'\n <mu>= %.4f'%mu_av+'+/-%.4f'%mu_av_er \
                                 +'.\n chisq=%.2f'%chisq,fontsize=9);
    plt.show();
    return statis_result
```

```python
### list comparing all <mu>s

# Histogram of mu: Multiple Plot
def list_mu_histogram(list_mu_data, Title_List=[], nbins=10, num_column=3, print_statis_result=False, \
                      only_do_statis = False, abs_mu = False):
    global FIG_X_SIZE, FIG_Y_SIZE  
    num_fig = len(list_mu_data)
    if only_do_statis == False:
        if(len(Title_List) != num_fig and len(Title_List) != 0):
            print 'ERROR! number of figure ', num_fig , 'and title',len(Title_List),'not match!'
            return
        # figure settings
        num_row =  num_fig /num_column + min(1,np.mod(num_fig,num_column))
        fig=plt.figure(figsize=(FIG_X_SIZE*num_column,FIG_Y_SIZE*num_row));
        fig.subplots_adjust(hspace=0.6)
    # array which saves statis result 
    statis_result = []
    # making plottings    
    for i_fig in range(num_fig):  
        x=list_mu_data[i_fig]
        #abs_x = [abs(x[row]) for row in range(len(x))]
        # statistical quantities
        mu_av = sum(x)/len(x)
        mu_var = sum((x[row]-mu_av)**2.0 for row in range(len(x)))/(len(x)-1.0) #correct the bias using n-1
        mu_av_er = np.sqrt(mu_var) / np.sqrt(len(x)-1.0) # correct the bias using n-1 
        # saving statistical results
        now_statis_result =[ Title_List[i_fig]+': mu_av, mu_av_er, mu_var = ', \
                                        [mu_av, mu_av_er, mu_var] ]
        statis_result.append(now_statis_result)
        if only_do_statis == False:
            if abs_mu:
                xmin = 0; xmax = 1
            else:
                xmin = -1; xmax = 1
            muhist=fig.add_subplot(num_row,num_column,i_fig+1);
            #Histogram Plotting   
            muhist.hist(x, nbins, range=(xmin,xmax));
            npoint=len(x)
            #Reference PDF
            X0=np.linspace(xmin,xmax,nbins); Y0=X0-X0+npoint/(nbins+0.0)
            muhist.plot(X0, Y0, color='black', lw=1);
            #Estimating Error Bars
            n_array, bins_array=np.histogram(x,nbins,range=(xmin,xmax))
            X1=[(bins_array[k]+bins_array[k+1])/2 for k in range(nbins)]
            Y1=n_array
            Yerr=[n_array[k]/np.sqrt(n_array[k]-1.0) for k in range(nbins)]
            muhist.errorbar(X1, Y1, yerr=Yerr, color='red', lw=3, fmt='p');
            # chisq
            chisq = sum([ ((Y0[row]-Y1[row])/Yerr[row])**2.0 for row in range(nbins)  ]) 
            # titles
            if len(Title_List) != 0:
                muhist.set_title(Title_List[i_fig]\
                                 +'\n <mu> = %.4f'%mu_av+'+/-%.4f'%mu_av_er \
                                 +'.\n chisq=%.2f'%chisq,fontsize=14);
            if len(Title_List) == 0:
                muhist.set_title(' <mu> = %.4f'%mu_av+'+/-%.4f'%mu_av_er \
                                     +'.\n chisq=%.2f'%chisq,fontsize=14)
        if print_statis_result:
            print Title_List[i_fig]+': [<mu>, variance of mu, chisq] = \t[%.4f'%mu_av\
                    +'+/-%.4f'%mu_av_er+', %.4f'%mu_var+']'            
    if only_do_statis == False:
            plt.show()  
    return statis_result

# mu-Histogram of all Skeletons (real, AP, RSD, AP+RSD), with given beta
def list_all_mu_histogram(num = 30, num_in_x = 10, boundary_remove = 0, cell_average = False, \
                          abs_mu = False, nbins=10, num_column=3, only_do_statis = False,\
                          print_statis_result = True):    
    # real cosmology, wo/with RSD
    AP = 'F'
    RSD = 'F'
    mu_data01, mu_data1 = get_mu_data(RSD, AP,  num, num_in_x, print_info=False, \
                                    boundary_remove=boundary_remove, only_cell_average = cell_average,\
                                    abs_mu = abs_mu)
    AP = 'F'
    RSD = 'T'
    mu_data02, mu_data2 = get_mu_data(RSD, AP,  num, num_in_x, print_info=False,\
                                    boundary_remove=boundary_remove, only_cell_average = cell_average,\
                                    abs_mu = abs_mu)
    if cell_average:
        mu_data_list = [mu_data01, mu_data02]
    else:
        mu_data_list = [mu_data1, mu_data2]
    title_list = [ 'No   RSD', 'With RSD']
    # with AP
    for AP in range(len(assumed_cosmo)):
        RSD = 'F'
        mu_data0, mu_data = get_mu_data(RSD, AP,  num, num_in_x, print_info=False,\
                                        boundary_remove=boundary_remove, only_cell_average = cell_average,\
                                        abs_mu = abs_mu)
        if cell_average:
            mu_data_list.append(mu_data0)
        else:
            mu_data_list.append(mu_data)
        title_list.append('AP='+str(assumed_cosmo[AP])+', No  RSD')
        RSD = 'T'
        mu_data0, mu_data = get_mu_data(RSD, AP,  num, num_in_x, print_info=False,\
                                        boundary_remove=boundary_remove, only_cell_average = cell_average,\
                                        abs_mu = abs_mu)
        if cell_average:
            mu_data_list.append(mu_data0)
        else:
            mu_data_list.append(mu_data)
        title_list.append('AP='+str(assumed_cosmo[AP])+', With RSD')

    # get statistical results
    statis_result = list_mu_histogram(mu_data_list, title_list, \
                                      nbins, num_column, print_statis_result = print_statis_result, \
                                      only_do_statis = only_do_statis, abs_mu=abs_mu)
    [mu_av0, mu_av_er0, mu_var0] = statis_result[0][1]
    [mu_av1, mu_av_er1, mu_var1] = statis_result[1][1]
    
    if print_statis_result:
        # print information
        print 'number of mus: '
        print '\t', [len(mu_data_list[row]) for row in range(len(mu_data_list))], '\n'
        print '\n'
        print ' RSD shift of <|mu|>:   value, sigma = %.4f'\
        %(mu_av1-mu_av0)+', %.2f'%(abs(mu_av1-mu_av0)/mu_av_er0)
        print '\n'
        for AP in range(len(assumed_cosmo)):
            print 'Assuming cosmology',assumed_cosmo[AP], ':\n'
            [mu_av, mu_av_er, mu_var] = statis_result[2+2*AP][1]
            print ' No   RSD, AP shift of <mu>:   value, sigma = %.4f'\
            %(mu_av-mu_av0)+', %.2f'%(abs(mu_av-mu_av0)/mu_av_er0)
            [mu_av, mu_av_er, mu_var] = statis_result[2+2*AP+1][1]
            print ' With RSD, AP shift of <mu>:   value, sigma = %.4f'\
            %(mu_av-mu_av1)+', %.2f'%(abs(mu_av-mu_av1)/mu_av_er0)
            print '\n'
    return statis_result
```
