Visualization of the network data produced by ngl-beta

```python
%pylab inline
rcParams['figure.figsize'] = 8, 8
```

```python
point_data_path="../../data/input/"
link_data_path="../../data/output/"
#point_filename="Pos_MD_halos_1000kms_z0.dat"
point_filename="Pos_MD_full_halos_900-1000kms_z0.dat"
```

```python
link_method="BSkeleton"
beta_value=1.0
link_filename="%s_b%.1f_%s"%(link_method, beta_value, point_filename)
```

```python
link_data = loadtxt(link_data_path+link_filename)
point_data =  loadtxt(point_data_path+point_filename)
```

```python
#here we clean the link data. all the edges are listed twice.
def clean_links(link_data):
    n_edges = size(link_data[:,0])
    clean_link_data_A = empty((0))
    clean_link_data_B = empty((0))

    clean_link_data_A = append(clean_link_data_A, link_data[0,0])
    clean_link_data_B = append(clean_link_data_B, link_data[0,1])
    print clean_link_data_A, clean_link_data_B
    for i_edge in range(1,n_edges):
        tmp_edge = link_data[i_edge,:]
#    print clean_link_data
        tmp_index = where((clean_link_data_A[:]==tmp_edge[1]) & (clean_link_data_B[:]==tmp_edge[0]))
        if not size(tmp_index):
            clean_link_data_A = append(clean_link_data_A, tmp_edge[0])
            clean_link_data_B = append(clean_link_data_B, tmp_edge[1])
    clean_link_data_A = int_(clean_link_data_A)
    clean_link_data_B = int_(clean_link_data_B)

    print size(clean_link_data_A), size(link_data[:,0])
    new_size = size(clean_link_data_A)
    clean_link_data = ones((new_size,2))
    clean_link_data[:,0] = clean_link_data_A[:]
    clean_link_data[:,1] = clean_link_data_B[:]
    return clean_link_data

clean_link_data = clean_links(link_data)
```

```python
n_points = size(clean_link_data[:,0])
for i_link in range(n_points):
    point_A_ID = clean_link_data[i_link,0]
    point_B_ID = clean_link_data[i_link,1]
    points_x = array([point_data[point_A_ID,0], point_data[point_B_ID,0]])
    points_y = array([point_data[point_A_ID,1], point_data[point_B_ID,1]])
    points_z = array([point_data[point_A_ID,2], point_data[point_B_ID,2]])
    if((points_z[0]<750.0) & (points_z[0]>700.0) & (points_z[1]<750.0) & (points_z[1]>700.0)):
        plot(points_x, points_y, linestyle="-", color="b")
        #scatter(points_x, points_z)
```

```python
volume = 1000.0**3.0
n_halos = size(point_data[:,0])
average_length = (volume/n_halos)**(1.0/3.0)
print average_length, volume/n_halos
length_list = empty((0))

for point in link_data[:,]:
    point_A_ID = point[0]
    point_B_ID = point[1]
    points_x = array([point_data[point_A_ID,0], point_data[point_B_ID,0]])
    points_y = array([point_data[point_A_ID,1], point_data[point_B_ID,1]])
    points_z = array([point_data[point_A_ID,2], point_data[point_B_ID,2]])
    #print points_x, points_y, points_z
    length = (points_x[0]- points_x[1])**2 + (points_y[0]- points_y[1])**2 + (points_z[0]- points_z[1])**2
    length = sqrt(length)
    length_list = append(length_list, length)
```

```python
n, bins, patches = hist((length_list/average_length), 20, normed=0, color = 'silver')
ax = axes()
ax.set_xlabel("$l_{\mathrm{edge}}/ l_{\mathrm{mean}}$",fontsize=25)
ax.set_ylabel("$\mathrm{N_{edges}}$",fontsize=25)
ticklabels_x = ax.get_xticklabels()
ticklabels_y = ax.get_yticklabels()

for label_x in ticklabels_x:
    label_x.set_fontsize(18)
    label_x.set_family('serif')
for label_y in ticklabels_y:
    label_y.set_fontsize(18)
    label_y.set_family('serif')
```
